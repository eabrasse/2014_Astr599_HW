```python
print "Number of missing values: " + str(mask.sum())
```

#### Next we need to find the rows where there are missing data. To do this, we can set a new mask called bad that checks for any missing values in each column ('False'). Then if any are False, the row is deemed False and we subscript our old data with the opposite of the 'bad' mask.

```python
bad = np.isnan(data[:,0] + data[:,1] + data[:,2] + data[:,3] + data[:,4] +data[:,5] + data[:,6])

newdata = data[~bad]
print "Number of complete rows: " + str(len(newdata))
```

```python
 #axis allows you to make calculations for all columns
maxarr = np.amax(newdata, axis=0)
minarr = np.amin(newdata, axis=0)
meanarr = np.mean(newdata, axis=0)
stdarr = np.std(newdata, axis=0)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sini = np.sin(newdata[:,2] * (np.pi/180.0))
```

```python
plt.scatter(newdata[:,0] , sini)
plt.xlabel('a (AU)')
plt.ylabel('sin(i)')
```

#### The distribution of orbital inclination as a function of semimajor axis. We can clearly see the Kirkwood Gaps (due to resonances with Jupiter) at ~3.25, ~3.75 and ~4.25 AU and the majority of asteroids have sin(i) values < 0.3. This makes sense considering our Solar System is relatively flat, but there are a few outliers that could have high inclinations due to gravitational interactions with other asteroids or planets.

```python
plt.scatter(newdata[:,5], newdata[:,3])
plt.xlabel('B-V')
plt.ylabel('H')
plt.axis([0.5, 1.2, 14, 2])

# overplot a line at the mean of B-V
plt.plot( [np.mean(newdata[:,5]), np.mean(newdata[:,5])], [14,2])
```

#### Color Magnitude Diagram (CMD) shown above

```python
# just set a mask to true/false the data if they're below/above B-V mean
fam1 = (newdata[:,5] < np.mean(newdata[:,5]))

plt.scatter(newdata[:,0] , sini, c=fam1, cmap='autumn')
plt.xlabel('a (AU)')
plt.ylabel('sin(i)')

# true = 1 = bluer B-V = yellow dot
# false = 0 = redder B-V = red dot
```

#### The redder objects appear to orbit at closer distances than the bluer objects. There may also be a slight trend in inclination, where redder objects are found at lower inclinations than those of their bluer counterparts (beyond ~3 AU). This is consistent with the plots in Parker et al. 2008

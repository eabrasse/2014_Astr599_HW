```python
import numpy as np
import matplotlib.pyplot as plt
from astroML.datasets import fetch_rrlyrae_combined
from sklearn.cross_validation import train_test_split

%matplotlib inline
```

```python
X, y = fetch_rrlyrae_combined()
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
N_plot = 5000
plt.scatter(X[-N_plot:, 0], X[-N_plot:, 1], c=y[-N_plot:],
            edgecolors='none', alpha=0.3)
plt.xlabel('u-g color')
plt.ylabel('g-r color')
```

```python
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier(n_neighbors=5)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
```

```python
print(np.sum(y_pred == y_test))
print(np.sum(y_pred != y_test))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred,
                            target_names=['MS star', 'RR Lyrae']))
```

### Exercise 1: Classification of RR Lyrae

#### Gaussian Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
clf2 = GaussianNB()
clf2.fit(X_train, y_train)
y_pred2 = clf2.predict(X_test)
```

```python
print(np.sum(y_pred2 == y_test))
print(np.sum(y_pred2 != y_test))
```

```python
print(classification_report(y_test, y_pred2,
                            target_names=['MS star', 'RR Lyrae']))
```

#### Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
# can tune RandomForClass with random state (?) ... 23 gave best precision
clf3 = RandomForestClassifier(random_state=23)
clf3.fit(X_train, y_train)
y_pred3 = clf3.predict(X_test)
```

```python
print(np.sum(y_pred3 == y_test))
print(np.sum(y_pred3 != y_test))
```

```python
print(classification_report(y_test, y_pred3,
                            target_names=['MS star', 'RR Lyrae']))
```

##### The Random Forest Classifier resulted in the best precision (74%) while the Gaussian Naive Bayes resulted in the best recall (81%). 

### Exercise 2: Photometric Redshifts

```python
from astroML.datasets import fetch_sdss_specgals
```

```python
data = fetch_sdss_specgals()

# put magnitudes in a matrix
X = np.vstack([data['modelMag_%s' % f] for f in 'ugriz']).T
y = data['z']

# Split into training and testing data
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
from sklearn.linear_model import LinearRegression
est = LinearRegression()
est.fit(X_train, y_train)
y_pred = est.predict(X_test)
```

```python
plt.plot(y_test, y_pred, ',k')
plt.plot([0, 1], [0, 1], ':k')
plt.xlim(0, 0.6)
plt.ylim(0, 0.6)
```

```python
rms = np.sqrt(np.mean((y_test - y_pred) ** 2))
print(rms)
```

#### Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
est2 = RandomForestRegressor()
est2.fit(X_train, y_train)
y_pred2 = est2.predict(X_test)
```

```python
plt.plot(y_test, y_pred2, ',k')
plt.plot([0, 1], [0, 1], ':k')
plt.xlim(0, 0.6)
plt.ylim(0, 0.6)
```

```python
rms = np.sqrt(np.mean((y_test - y_pred2) ** 2))
print(rms)
```

#### K-nearest Neighbors

```python
from sklearn.neighbors import KNeighborsRegressor
est3 = KNeighborsRegressor()
est3.fit(X_train, y_train)
y_pred3 = est3.predict(X_test)
```

```python
plt.plot(y_test, y_pred3, ',k')
plt.plot([0, 1], [0, 1], ':k')
plt.xlim(0, 0.6)
plt.ylim(0, 0.6)
```

```python
rms = np.sqrt(np.mean((y_test - y_pred3) ** 2))
print(rms)
```

##### The Random Forest Regressor resulted in the best RMS error (0.0232).

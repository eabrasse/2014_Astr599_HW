1: Load asteroids data from "asteroids5000.txt"

2: Create mask to determine how many values are missing from dataset
and then determine how many complete rows there are given 7 data
points per row. 

3: New array containing complete rows

4: Column-wise stats

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

#1
data = np.genfromtxt("asteroids5000.txt",delimiter=',',skip_header=1)

#2
mask = np.isnan(data)
print "Total number of missing values =",mask.sum()

#Sum will be !nan if it's complete
mask = (~np.isnan(data[:,0]+data[:,1]+data[:,2]+data[:,3]+data[:,4]+data[:,5]+data[:,6]))
print "Number of complete rows = ",mask.sum()

#3
cleanData = data[mask]

#4
colMaxArr = np.amax(cleanData,axis=0)
colMinArr = np.amin(cleanData,axis=0)
colMeanArr = np.mean(cleanData,axis=0)
colStdArr = np.std(cleanData,axis=0)
print "Maximum in each column:",colMaxArr
print "Minimum in each column:",colMinArr
print "Mean in each column:",colMeanArr
print "Standard deviation is each column:",colStdArr
```

```python
#PART 2

#1: header info
!head -1 asteroids5000.txt
#2 :Matplot lib stuff done above

#3: Plot a vs. sin(i) where i = 3rd column...cleanData[:,2]
#convert i from deg->
conv = np.pi/180
cleanData[:,2] = cleanData[:,2] * conv
cleanData[:,2] = np.sin(cleanData[:,2])
plt.scatter(cleanData[:,0],cleanData[:,2]);
plt.ylabel('sin(i)');
plt.xlabel('Semimajor Axis (AU)');
```

It appears that the asteroids seem to be clumped in the range (2,3.25) AU for the semimajor axes and for sins of inclinaiton in the range (0,0.3).  From this, we can see that in general, the asteroids tend to be constained to low inclinations (<~17.5 degrees).  This makes sense since when they formed, they likely formed from a thin circumstellar disk which via conservation of angular momentum remains thin, hence small inclinations. Some reached higher inclinations through high velocity collisions giving the resultant spread.
The Asteroid Belt is located between ~2 and ~3.5 AU and this agrees well with the location of the asteroid clusters in the plot.  Also apparent are pronounced gaps in the plot around 2.5AU, 3AU, 3.25AU and 3.5-4Au.  These gaps are known as Kirkwood gaps are due to orbital resonances with Jupiter.  Their presence in the data confirms the plots accuracy.
source: https://en.wikipedia.org/wiki/Kirkwood_gap).

```python
#4: Color-magnitude diagram
plt.scatter(cleanData[:,3],cleanData[:,5]);
plt.ylabel('B-V');
plt.xlabel('H');
#Overplot line separating 2 clumps
#use B-V mean as demarcation line
sampleX = np.linspace(0,16,50)
sampleY = np.ones_like(sampleX)*colMeanArr[5]
plt.plot(sampleX,sampleY)
plt.xlim([0,16]);
```

```python
#5
clump1 = (cleanData[:,5] > colMeanArr[5])
clump2 = (cleanData[:,5] <= colMeanArr[5])
#Make new subarrays
clump1Arr = cleanData[clump1]
clump2Arr = cleanData[clump2]
#Plot
plt.scatter(clump1Arr[:,0],clump1Arr[:,2],color='red',alpha=0.5);
plt.scatter(clump2Arr[:,0],clump2Arr[:,2],color='blue',alpha=0.5);
plt.ylabel('sin(i)');
plt.xlabel('Semimajor Axis (AU)');
```

Redder objects (B-V > mean) tend to have lower semimajor axes while bluer objects (B-v < mean)
tend to be located farther out as shown in the plot below.  There is, however, a sizeable overlap
region ~2.25-2.75 AU where one can find red and blue objects.  Also, it seems that redder objects
on average have slightly lower inclinations than their bluer counterparts.

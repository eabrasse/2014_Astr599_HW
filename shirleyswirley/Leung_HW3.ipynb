Import libraries.

```python
%matplotlib inline
import astroML.datasets
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize
```

Fetch a star.

```python
lcs = astroML.datasets.fetch_LINEAR_sample()
time, flux, dflux = lcs[10003298].T
```

Plot the raw lightcurve.

```python
plt.errorbar(time,flux,yerr=dflux, fmt="ro")
plt.gca().invert_yaxis()
plt.xlabel("Mean Julian Day")
plt.ylabel("Mag")
plt.show()
```

```python
from astroML.time_series import lomb_scargle
periods = np.logspace(-1, 0, 10000, base=10)
periodogram = lomb_scargle(time, flux, dflux, omega=2 * np.pi / periods, generalized=True)
plt.plot(periods, periodogram)
plt.semilogx()
plt.xlabel("Period (days)")
plt.ylabel("Power")
plt.show()
```

Find the peak of this periodogram using argmax.

```python
idx = np.argmax(periodogram)
print periods[idx], periodogram[idx]
```

Now fold the lightcurve at this best period and plot it.

```python
period = periods[idx]
phase = time/period - time//period
print min(phase), max(phase)
plt.errorbar(phase, flux, yerr=dflux, fmt="ro")
plt.gca().invert_yaxis()
plt.xlabel("Phase")
plt.ylabel("Mag")
plt.show()
```

Create a class to implement the model we want to fit to our data.

```python
class Fourier(object):
    def __init__(self, phase, flux, dflux, nterms):
        self.phase = phase
        self.flux = flux
        self.dflux = dflux
        self.nterms = nterms
        assert(self.nterms > 0)
        
    def evaluate(self, phase, terms):
        assert(len(terms) == 2 * self.nterms - 1)
        model = terms[0] * np.ones(len(phase))
        for i in range(self.nterms-1):
            model += terms[2*i + 1] * np.cos(2 * np.pi * (i+1) * phase + terms[2*i + 2])
        return model
    
    def chi2(self, args):
        model = self.evaluate(self.phase, args)
        chi   = (model - self.flux) / self.dflux
        return np.sum(chi**2)

    def makeGuess(self):
        guess = []
        guess.append(np.mean(self.flux))
        if self.nterms > 1:
            guess.append(np.std(self.flux))
            guess.append(2 * np.pi * (1 - self.phase[np.argsort(self.flux)[-1]]))
            for i in range(2, self.nterms):
                guess.append(0.0)
                guess.append(2 * np.pi * (1 - self.phase[np.argsort(self.flux)[-1]]))
        return guess
```

Try a range of models and plot how chi-squared decreases as you increase the model parameters.

```python
nterms = np.arange(1, 10)
chi2   = []
BICpenalty = []
AICpenalty = []
for nterm in nterms:
    fourier = Fourier(phase, flux, dflux, nterm)
    guess = fourier.makeGuess()
    chi2.append(optimize.fmin_bfgs(fourier.chi2, x0=[guess], disp=0, full_output=1)[1])
    BICpenalty.append(len(guess)*np.log(len(fourier.phase)))
    AICpenalty.append(len(guess)*2.0)
chi2 = np.array(chi2)
BICpenalty = np.array(BICpenalty)
AICpenalty = np.array(AICpenalty)
print "chi-sq:",chi2
print "BIC penalty:",BICpenalty
print "AIC penalty:",AICpenalty
plt.plot(nterms, chi2, label="chi2")
plt.plot(nterms, chi2 + BICpenalty, label="BIC")
plt.plot(nterms, chi2 + AICpenalty, label="AIC")
plt.xlabel("N Terms")
plt.ylabel("Chi2")
plt.semilogy()
plt.legend()
plt.show()
```

### Method 1: Find the optimal model based on chi-sq differences.

```python
delta_chi2 = chi2[:-1] - chi2[1:]
plt.plot(nterms[1:], delta_chi2)
plt.xlabel("N terms")
plt.ylabel("Delta chi2")
plt.show()
print "delta chi-sq:",delta_chi2
```

From the table given at the beginning of the 14_Optimization notebook combined with a little online background reading, delta chi-sq must be greater than ~6.17 for p < 0.05 (we add one term or two parameters for each new model we test and compare this new model to the previous one with one less term or two less parameters - thus, N (dfs of delta chi-sq) = 2).
<br>So we look for the number of terms such that delta chi-sq drops below ~6.17.

```python
print "chi-sq diffs optimal model order:",nterms[1:][delta_chi2 < 6.17]-1
```

### Methods 2 and 3: Find the optimal model based on minimum BIC and AIC.

```python
print "BIC optimal model order:",nterms[np.argmin(chi2+BICpenalty)]
print "AIC optimal model order:",nterms[np.argmin(chi2+AICpenalty)]
```

Plot all the different optimal models with the data.

```python
mphase = np.arange(0, 1, 0.01)

# 4-term model (BIC)
fourier4 = Fourier(phase, flux, dflux, 4)
guess4 = fourier4.makeGuess()
result4 = optimize.fmin_bfgs(fourier4.chi2, x0=[guess4], disp=0, full_output=1)
best_fit4 = result4[0]
print "Best-fit parameters:",best_fit4
plt.errorbar(fourier4.phase, fourier4.flux, yerr=fourier4.dflux, fmt="ro")
plt.plot(mphase, fourier4.evaluate(mphase, guess4), "b--", label="Initial guess")
plt.plot(mphase, fourier4.evaluate(mphase, best_fit4), "g-", label="Best fit")
plt.gca().invert_yaxis()
plt.xlabel("Phase")
plt.ylabel("Mag")
plt.title("4-term model - BIC")
plt.legend(loc=4)
plt.show()

# 5-term model (delta chi-sq)
fourier5 = Fourier(phase, flux, dflux, 5)
guess5 = fourier5.makeGuess()
result5 = optimize.fmin_bfgs(fourier5.chi2, x0=[guess5], disp=0, full_output=1)
best_fit5 = result5[0]
print "Best-fit parameters:",best_fit5
plt.errorbar(fourier5.phase, fourier5.flux, yerr=fourier5.dflux, fmt="ro")
plt.plot(mphase, fourier5.evaluate(mphase, guess5), "b--", label="Initial guess")
plt.plot(mphase, fourier5.evaluate(mphase, best_fit5), "g-", label="Best fit")
plt.gca().invert_yaxis()
plt.xlabel("Phase")
plt.ylabel("Mag")
plt.title("5-term model - chi-sq diffs")
plt.legend(loc=4)
plt.show()

# 9 term-model (AIC)
fourier9 = Fourier(phase, flux, dflux, 9)
guess9 = fourier9.makeGuess()
result9 = optimize.fmin_bfgs(fourier9.chi2, x0=[guess9], disp=0, full_output=1)
best_fit9 = result9[0]
print "Best-fit parameters:",best_fit9
plt.errorbar(fourier9.phase, fourier9.flux, yerr=fourier9.dflux, fmt="ro")
plt.plot(mphase, fourier9.evaluate(mphase, guess9), "b--", label="Initial guess")
plt.plot(mphase, fourier9.evaluate(mphase, best_fit9), "g-", label="Best fit")
plt.gca().invert_yaxis()
plt.xlabel("Phase")
plt.ylabel("Mag")
plt.title("9-term model - AIC")
plt.legend(loc=4)
plt.show()
```

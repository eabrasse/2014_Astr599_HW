```python
import numpy as np
```

## Part 1: Loading and exploring the data

```python
# 1. Load data from the asteroids file
asteroids=np.genfromtxt("asteroids5000.csv",delimiter=",")
```

```python
# 2. How many values are missing in this data? How many asteroids have no missing values?
print 'Number of missing values:',np.isnan(asteroids).sum()
[m,n]=np.shape(asteroids)
print 'Number of asteroids without missing values:',m-np.isnan(asteroids).any(axis=1).sum()
```

```python
# 3. Create a new array containing only the rows with no missing values.
asteroidsvalid=asteroids[~np.isnan(asteroids).any(axis=1)]
```

```python
# 4. Compute the max, min, mean, stdev of the values in each column.
print 'Max of each column:'
print np.max(asteroidsvalid,axis=0)
print 'Min of each column:'
print np.min(asteroidsvalid,axis=0)
print 'Mean of each column:'
print np.mean(asteroidsvalid,axis=0)
print 'Stdev of each column:'
print np.std(asteroidsvalid,axis=0)
```

## Part 2: Plotting the data

```python
# 1. Display the first line of the data file.
!head -1 asteroids5000.csv
```

```python
# 2. Make figures appear inline in the notebook.
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# 3. Plot the semi-major axis vs. the sine of the inclination angle
semimajoraxis = asteroidsvalid[:,0]
sininclangle = np.sin(np.radians(asteroidsvalid[:,2]))
plt.scatter(sininclangle,semimajoraxis)
plt.title('Semi-major axis vs. sin of inclination angle')
plt.xlabel('Sin of inclination angle')
plt.ylabel('Semi-major axis')
```

There seem to be 5 or 6 clumps of asteroids orbiting with 5 or 6 different average semi-major axis lengths. Perhaps this corresponds to orbits of some of the planets? I don't know astronomy...

```python
# 4. Plot H vs. B-V
H = asteroidsvalid[:,3]
BV = asteroidsvalid[:,5]
plt.scatter(BV,H)
plt.title('H vs. B-V')
plt.xlabel('B-V')
plt.ylabel('H')
plt.xlim([np.min(BV),np.max(BV)])
plt.ylim([np.min(H),np.max(H)])
# Overplot a line that divides two distinct families of asteroids
dividingBV=0.78; # eye-balled
plt.plot([dividingBV,dividingBV],[np.min(H),np.max(H)])
```

```python
# 5. Repeat #3 but plot the two families from #4 in different colors.
plt.scatter(sininclangle[BV>dividingBV],semimajoraxis[BV>dividingBV],c='b',alpha=0.7,label=r'larger B-V')
plt.scatter(sininclangle[BV<=dividingBV],semimajoraxis[BV<=dividingBV],c='r',alpha=0.7,label=r'smaller B-V')
plt.title('Semi-major axis vs. sin of inclination angle')
plt.xlabel('Sin of inclination angle')
plt.ylabel('Semi-major axis')
plt.grid()
plt.legend()
```

Redder objects seem to have larger semi-major axes than bluer ones.

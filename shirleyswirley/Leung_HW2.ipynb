# Part 1

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

# Define a function for sin(x)
def func(x):
    return np.sin(x)
```

# Part 2 and 3

```python
class Integrator(object):
    def __init__(self, a, b, n):
        self.a, self.b, self.n = a, b, n
        self.points, self.weights = self.method()
        
    def method(self):
        raise NotImplementedError('no rule in class %s' %self.__class__.__name__)
        
    def integrate(self,func):
        return np.multiply(self.weights,func(self.points)).sum()
    
    def indefintegrate(self,func):
        return np.multiply(self.weights,func(self.points)).cumsum()
```

# Part 4

```python
class Trapezoidal(Integrator):
    def __init__(self,*args):
        super(Trapezoidal, self).__init__(*args)

    def method(self):
        x = np.linspace(self.a,self.b,self.n)
        w = np.zeros_like(x)
        w[:] = (self.b-self.a)/self.n
        w[0] = (self.b-self.a)/self.n/2
        w[-1] = (self.b-self.a)/self.n/2
        return (x,w)
```

```python
trap1 = Trapezoidal(0,2*np.pi,100)
print(trap1.integrate(func)) # should be zero-ish
trap2 = Trapezoidal(0,np.pi,100)
print(trap2.integrate(func)) # should be 2-ish
```

# Part 5

```python
class Midpoint(Integrator):
    def __init__(self,*args):
        super(Midpoint, self).__init__(*args)

    def method(self):
        x = np.arange(self.a+(self.b-self.a)/self.n/2,self.b-(self.b-self.a)/self.n/2,(self.b-self.a)/self.n)
        w = np.zeros_like(x)
        w[:] = (self.b-self.a)/self.n
        return (x,w)
```

```python
mid1 = Midpoint(0,2*np.pi,100)
print(mid1.integrate(func)) # should be zero-ish
mid2 = Midpoint(0,np.pi,100)
print(mid2.integrate(func)) # should be 2-ish
```

# Part 6

```python
def integfuncMidpoint(x,n):
    xpts = np.linspace(0,x,n)
    y = np.zeros_like(xpts)
    for i in range(0,len(xpts)):
        y[i] = Midpoint(0,xpts[i],n).integrate(func)
    plt.plot(xpts,y,label=r'midpt integr')
    plt.plot(xpts,1-np.cos(xpts),label=r'1-cos(x)')
    plt.title('Comparison of midpt integration of sin(x) and its exact solution = 1-cos(x)')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid()
    plt.legend()
    plt.axis([0, x, 0, 2.1])
    
integfuncMidpoint(4*np.pi,200)
```

```python
def integfuncTrapezoidal(x,n):
    xpts = np.linspace(0,x,n)
    y = np.zeros_like(xpts)
    for i in range(0,len(xpts)):
        y[i] = Trapezoidal(0,xpts[i],n).integrate(func)
    plt.plot(xpts,y,label=r'trap integr')
    plt.plot(xpts,1-np.cos(xpts),label=r'1-cos(x)')
    plt.title('Comparison of trap integration of sin(x) and its exact solution = 1-cos(x)')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid()
    plt.legend()
    plt.axis([0, x, 0, 2.1])
    
integfuncTrapezoidal(4*np.pi,200)
```

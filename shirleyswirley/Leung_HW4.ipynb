```python
# From Global Health Data Exchange, load PHMRC VA adult data, CSV format:
url = 'http://ghdx.healthdata.org/sites/default/files/'\
      'record-attached-files/IHME_PHMRC_VA_DATA_ADULT_Y2013M09D11_0.csv'
df = pd.read_csv(url, low_memory=False)
# make the row names more interesting than numbers starting from zero
df.index = ['person %d'%(i+1) for i in df.index]
df.head()
```

```python
# Also load codebook
# Couldn't get excel file to load, so just saved and converted it to a csv file
cb = pd.read_csv('codebook.csv')
cb.head()
```

# Use the codebook to find the column which corresponds to the question "Did __ have a fever?"

```python
cb[cb.question.str.contains('fever')]
```

```python
var = cb.loc[79, 'variable']
```

#  Find the cause with the highest endorsement rate for the symptom "fever" among Adult deaths

```python
df['fever'] = df[var] == 'Yes'
```

```python
df.groupby('gs_text34').word_fever.mean().order(ascending=False)
```

So <b>malaria</b> has the highest endorsement rate for the symptom "fever" among Adult deaths.

# Display the cause-specific endorsement rates visually.

```python
import matplotlib.pyplot as plt
%matplotlib inline
(100*df.groupby('gs_text34').word_fever.mean()).order().plot(kind='barh', figsize=(10,14))
plt.xlabel('Endorsement Rate (%)')
plt.ylabel('Cause')
```

# Make a 2x2 table showing the number of adult deceased with and without this question endorsed for decedents with underlying cause AIDS and not AIDS.

```python
df['AIDSendorsed'] = df[var] == 'Yes'
df['AIDScaused'] = df['gs_text34'] == 'AIDS'
```

```python
df.groupby(['AIDSendorsed','AIDScaused']).size().unstack()
```

# Compare this 2x2 table across study sites.

```python
tablebysite=pd.crosstab(df['site'], [df['AIDSendorsed'], df['AIDScaused']])
tablebysite
```

# Now do a series of 2x2 tables comparing the percent of deaths truly due to AIDS for which "Had AIDS" question was endorsed.

```python
func = lambda x: (100 * x / x.sum()).round(2)
tablebysite.apply(func, axis=1)
```

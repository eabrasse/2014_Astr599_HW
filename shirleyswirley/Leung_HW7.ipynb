# Exercise 1

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
from astroML.datasets import fetch_rrlyrae_combined
from sklearn.cross_validation import train_test_split

X, y = fetch_rrlyrae_combined()

# For now, we'll only fit the first two colors
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
N_plot = 5000
plt.scatter(X[-N_plot:, 0], X[-N_plot:, 1], c=y[-N_plot:],
            edgecolors='none')
plt.xlabel('u-g color')
plt.ylabel('g-r color')
```

### Classification 1: K-neighbors

```python
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier(n_neighbors=5)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

print('y_pred==y_test',np.sum(y_pred == y_test))
print('y_pred!=y_test',np.sum(y_pred != y_test))

from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred,
                             target_names=['MS star', 'RR Lyrae']))
```

### Classification 2: Gaussian Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
clf = GaussianNB()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

print('y_pred==y_test',np.sum(y_pred == y_test))
print('y_pred!=y_test',np.sum(y_pred != y_test))

print(classification_report(y_test, y_pred,
                             target_names=['MS star', 'RR Lyrae']))
```

### Classification 3: LDA

```python
from sklearn.lda import LDA
clf = LDA()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

print('y_pred==y_test',np.sum(y_pred == y_test))
print('y_pred!=y_test',np.sum(y_pred != y_test))

print(classification_report(y_test, y_pred,
                             target_names=['MS star', 'RR Lyrae']))
```

### Classification 4: SVC with kernel='rbf'

```python
from sklearn.svm import SVC
clf = SVC()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

print('y_pred==y_test',np.sum(y_pred == y_test))
print('y_pred!=y_test',np.sum(y_pred != y_test))

print(classification_report(y_test, y_pred,
                             target_names=['MS star', 'RR Lyrae']))
```

### Classification 5: Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

print('y_pred==y_test',np.sum(y_pred == y_test))
print('y_pred!=y_test',np.sum(y_pred != y_test))

print(classification_report(y_test, y_pred,
                             target_names=['MS star', 'RR Lyrae']))
```

#### Conclusions: K-Neighbors and Random Forest Classifier methods seem to produce the best precision/recall for the RR-Lyrae data.

# Exercise 2

```python
from astroML.datasets import fetch_sdss_specgals

data = fetch_sdss_specgals()

# put magnitudes in a matrix
X = np.vstack([data['modelMag_%s' % f] for f in 'ugriz']).T
y = data['z']

# Split into training and testing data
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

### Regression 1: Linear Regression

```python
from sklearn.linear_model import LinearRegression
est = LinearRegression()
est.fit(X_train, y_train)
y_pred = est.predict(X_test)
plt.plot(y_test, y_pred, ',k')
plt.plot([0, 1], [0, 1], ':k')
plt.xlim(0, 0.6)
plt.ylim(0, 0.6)
plt.xlabel('actual data')
plt.ylabel('predicted data')
rms = np.sqrt(np.mean((y_test - y_pred) ** 2))
print(rms)
```

### Regression 2: K-Neighbors Regressor

```python
from sklearn.neighbors import KNeighborsRegressor
est = KNeighborsRegressor()
est.fit(X_train, y_train)
y_pred = est.predict(X_test)
plt.plot(y_test, y_pred, ',k')
plt.plot([0, 1], [0, 1], ':k')
plt.xlim(0, 0.6)
plt.ylim(0, 0.6)
plt.xlabel('actual data')
plt.ylabel('predicted data')
rms = np.sqrt(np.mean((y_test - y_pred) ** 2))
print(rms)
```

### Regression 3: Support Vector Regression

```python
from sklearn.svm import SVR
est = SVR()
est.fit(X_train, y_train)
y_pred = est.predict(X_test)
plt.plot(y_test, y_pred, ',k')
plt.plot([0, 1], [0, 1], ':k')
plt.xlim(0, 0.6)
plt.ylim(0, 0.6)
plt.xlabel('actual data')
plt.ylabel('predicted data')
rms = np.sqrt(np.mean((y_test - y_pred) ** 2))
print(rms)
```

### Regression 4: Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
est = RandomForestRegressor()
est.fit(X_train, y_train)
y_pred = est.predict(X_test)
plt.plot(y_test, y_pred, ',k')
plt.plot([0, 1], [0, 1], ':k')
plt.xlim(0, 0.6)
plt.ylim(0, 0.6)
plt.xlabel('actual data')
plt.ylabel('predicted data')
rms = np.sqrt(np.mean((y_test - y_pred) ** 2))
print(rms)
```

### Regression 5: Gradient Boosting Regressor

```python
from sklearn.ensemble import GradientBoostingRegressor
est = GradientBoostingRegressor()
est.fit(X_train, y_train)
y_pred = est.predict(X_test)
plt.plot(y_test, y_pred, ',k')
plt.plot([0, 1], [0, 1], ':k')
plt.xlim(0, 0.6)
plt.ylim(0, 0.6)
plt.xlabel('actual data')
plt.ylabel('predicted data')
rms = np.sqrt(np.mean((y_test - y_pred) ** 2))
print(rms)
```

#### Conclusions: The Random Forest Regressor seems to generate the smallest RMS deviations.

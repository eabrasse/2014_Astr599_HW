```python
var = 'a1_01_14'
df['hadAIDS'] = df[var] == 'Yes'
df['diedAIDS']=df['gs_text34']=='AIDS'
```

```python
df1 = pd.DataFrame({'reported AIDS': [np.sum(df.hadAIDS & df.diedAIDS), 
                                      np.sum(df.hadAIDS & (df.diedAIDS==False))],
                   ' didnt report AIDS': [np.sum((df.hadAIDS==False) & df.diedAIDS), 
                                                 np.sum((df.hadAIDS==False) & (df.diedAIDS==False))]})
```

```python
df1.index=['killed by AIDS', 'not killed by AIDS']
df1
```

```python
def comparison(location, df):
    for loc in location:
        print 'site = ', loc
        dfs=df[df.site==loc]
        df2 = pd.DataFrame({'reported AIDS': [np.sum(dfs.hadAIDS & dfs.diedAIDS), 
                                      np.sum(dfs.hadAIDS & (dfs.diedAIDS==False))],
                   ' didnt report AIDS': [np.sum((dfs.hadAIDS==False) & dfs.diedAIDS), 
                                                 np.sum((dfs.hadAIDS==False) & (dfs.diedAIDS==False))]})
        df2.index=['killed by AIDS', 'not killed by AIDS']

        print df2, '\n'
```

```python
comparison(np.unique(df.site), df)
```

```python
def comparisonPercent(location, df):
    for loc in location:
        print 'site = ', loc
        dfs=df[df.site==loc]
        df2 = pd.DataFrame({'reported AIDS': [100*np.sum(dfs.hadAIDS & dfs.diedAIDS)/(dfs.hadAIDS.sum()), 
                                      100*np.sum(dfs.hadAIDS & (dfs.diedAIDS==False))/(dfs.hadAIDS.sum())],
                   ' didnt report AIDS': [100*np.sum(((dfs.hadAIDS==False) & dfs.diedAIDS))/((dfs.hadAIDS==False).sum()), 
                                          100*np.sum((dfs.hadAIDS==False) & (dfs.diedAIDS==False))/((dfs.hadAIDS==False).sum())]})
        
        df2.index=['killed by AIDS', 'not killed by AIDS']

        print df2, '\n'
```

```python
comparisonPercent(np.unique(df.site), df)
```

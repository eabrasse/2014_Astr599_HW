## Part 1: loading and exploring the data

##### 1. load asteroids500.csv

```python
%matplotlib inline
import numpy as np
import matplotlib as plt
ast=np.genfromtxt('../data/asteroids5000.csv',delimiter=',')
```

##### 2.a. How many missing values are in the data?

```python
nNans=np.isnan(ast).sum()
print('There are '+str(nNans)+' nans in the data.')
```

##### 2.b. How many row have no nans?

```python
nanRows=np.isnan(ast).sum(axis=1)
isFullRow=[not(x) for x in nanRows]
nNoNanRows=sum(isFullRow)
print('There are '+str(nNoNanRows)+' rows with no missing data.')
```

##### 3. Make a new matrix with only objects without missing data

```python
goodAst=ast[np.where(isFullRow),:].squeeze()
```

##### 4. compute the min, max, mean, and std of each column

```python
x=np.random.random([2,3])
colMin =np.min (goodAst,axis=0)
colMax =np.max (goodAst,axis=0)
colMean=np.mean(goodAst,axis=0)
colStd =np.std (goodAst,axis=0)

print 'the column minimums are '+', '.join(map(str,(round(x,3) for x in colMin)))
print 'the column maximums are '+', '.join(map(str,(round(x,3) for x in colMax)))
print 'the column means are '   +', '.join(map(str,(round(x,3) for x in colMean)))
print 'the column stds are '    +', '.join(map(str,(round(x,3) for x in colStd)))
```

##### 3. make the first orbital parameter plot

```python
plt.pyplot.scatter(goodAst[:,0],np.sin(goodAst[:,2]*np.pi/180),alpha=.35,linewidths=0);
plt.pyplot.title ('a vs sin(i)');
plt.pyplot.xlabel('a(AU)');
plt.pyplot.ylabel('sin(i)');
```

There seem to be density waves in the asteroids as a function of distance.  This could be caused by some kind of orbital responance with the orbits of the planets (interaction with inclination is presumably negligible due to the concentration near the ecliptic plane).  The pattern could also be related to an artifact in sampling or estimation (e.g. if distance were estimated with a probabistic model which used observed intensitiy and there were quasi-categorical 'types' of asteroids with different albeido).

##### 4. plot the color-magnitude diagram

```python
plt.pyplot.scatter(goodAst[:,3],np.sin(goodAst[:,5]*np.pi/180),alpha=.35,linewidths=0);
plt.pyplot.title ('Color-Magnitude Diagram');
plt.pyplot.xlabel('H');
plt.pyplot.ylabel('BV');
plt.pyplot.plot([2,14],[.01375,.01375],'r');
```

##### 5. plot the 'families' in the orbital parameter plot

```python
grp1=(np.sin(goodAst[:,5]*np.pi/180)>=.01375)
grp2=(np.sin(goodAst[:,5]*np.pi/180)< .01375)
plt.pyplot.scatter(goodAst[grp2][:,0],np.sin(goodAst[grp2][:,2]*np.pi/180),c=[.75,  0,1],alpha=.35,linewidths=0);
plt.pyplot.scatter(goodAst[grp1][:,0],np.sin(goodAst[grp1][:,2]*np.pi/180),c=[  0,.75,0],alpha=.35,linewidths=0);
plt.pyplot.title ('a vs sin(i)');
plt.pyplot.xlabel('a(AU)');
plt.pyplot.ylabel('sin(i)');
plt.pyplot.plot([2.50,2.50],[-.1,.7],'r--');
plt.pyplot.plot([2.82,2.82],[-.1,.7],'r--');
```

There is definitely a trend for objects with greater BV to be in slightly closer and those with a lower BV to be further out, despite a  substantial overlap, particularly in the second density band.  There may also be a trend for the lower BV objects to stray further form the ecliptic plane.

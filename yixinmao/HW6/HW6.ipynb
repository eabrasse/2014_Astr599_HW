```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
def mandel(x, y, max_iter):
    """
    Given z = x + iy and max_iter, determine whether the candidate
    is in the mandelbrot set for the given number of iterations
    """
    c = complex(x, y)
    z = 0.0j
    for i in range(max_iter):
        z = z*z + c
        if (z.real*z.real + z.imag*z.imag) >= 4:
            return i

    return max_iter

def create_fractal(Nx, Ny, xmin, xmax, ymin, ymax, max_iter):
    """Create and return a fractal image"""
    image = np.zeros((Ny, Nx), dtype=float)
    dx = (xmax - xmin) * 1. / Nx
    dy = (ymax - ymin) * 1. / Ny
    
    for x in range(Nx):
        rpart = xmin + x * dx
        for y in range(Ny):
            ipart = ymin + y * dy
            color = mandel(rpart, ipart, max_iter)
            image[y, x] = color
    return image
```

```python
image = create_fractal(300, 200, -2, 1, -1, 1, 20)
plt.imshow(image, cmap=plt.cm.jet)
```

## Part 1 - speeding up using Cython

```python
%load_ext cythonmagic
```

```python
%%cython
import numpy as np
def mandel_cython(x, y, max_iter):
    """
    Given z = x + iy and max_iter, determine whether the candidate
    is in the mandelbrot set for the given number of iterations
    """
    cdef complex c = complex(x, y)
    cdef complex z = 0.0j
    for i in range(max_iter):
        z = z*z + c
        if (z.real*z.real + z.imag*z.imag) >= 4:
            return i

    return max_iter

def create_fractal_cython(Nx, Ny, xmin, xmax, ymin, ymax, max_iter):
    """Create and return a fractal image"""
    image = np.zeros((Ny, Nx), dtype=float)
    cdef float dx = (xmax - xmin) * 1. / Nx
    cdef float dy = (ymax - ymin) * 1. / Ny
    cdef float rpart
    cdef float ipart
    cdef int x
    cdef int y
    
    for x in range(Nx):
        rpart = xmin + x * dx
        for y in range(Ny):
            ipart = ymin + y * dy
            color = mandel_cython(rpart, ipart, max_iter)
            image[y, x] = color
    return image
```

## Part 2 - speeding up with numba

```python
from numba import autojit
def mandel_nb1(x, y, max_iter):
    """
    Given z = x + iy and max_iter, determine whether the candidate
    is in the mandelbrot set for the given number of iterations
    """
    c = complex(x, y)
    z = 0.0j
    for i in range(max_iter):
        z = z*z + c
        if (z.real*z.real + z.imag*z.imag) >= 4:
            return i

    return max_iter
mandel_nb = autojit(mandel_nb1)

def create_fractal_nb1(Nx, Ny, xmin, xmax, ymin, ymax, max_iter):
    """Create and return a fractal image"""
    image = np.zeros((Ny, Nx), dtype=float)
    dx = (xmax - xmin) * 1. / Nx
    dy = (ymax - ymin) * 1. / Ny
    
    for x in range(Nx):
        rpart = xmin + x * dx
        for y in range(Ny):
            ipart = ymin + y * dy
            color = mandel_nb(rpart, ipart, max_iter)
            image[y, x] = color
    return image
create_fractal_nb = autojit(create_fractal_nb1)
```

## Part 3 - time it

```python
print "Regular:"
%timeit create_fractal(300, 200, -2, 1, -1, 1, 20)
print "Cython:"
%timeit create_fractal_cython(300, 200, -2, 1, -1, 1, 20)
print "Numba:"
%timeit create_fractal_nb(300, 200, -2, 1, -1, 1, 20)
```

```python
# Cython speeds up the running by a factor of 5-10; Numba speeds up the running by a factor of 50-100.
```

```python
%matplotlib inline
import pandas as pd
import numpy
import matplotlib.pyplot as plt
```

```python
################ import data #################
url = 'http://ghdx.healthdata.org/sites/default/files/'\
      'record-attached-files/IHME_PHMRC_VA_DATA_ADULT_Y2013M09D11_0.csv'
df = pd.read_csv(url)
cb = pd.read_csv('IHME_PHMRC_VA_DATA_CODEBOOK_Y2013M09D11_0.csv') 
# Because I am using pandas version 0.10.0 which cannot directly read .xlsx files, I saved it as csv file and read it.
```

```python
################ Question 1 Did __ have a fever? #############
temp1 = cb.question.str[0:3]=="Did"
temp2 = cb.ix[temp1]
temp3 = temp2.question.str[-13:]=="have a fever?"
answer = temp2.ix[temp3]
var = answer.ix[int(answer.index),0]
print var
```

```python
################ Question 2 the highest endorsement ratio #############
df['fever'] = df[var] == 'Yes'
df['count'] = 1
a = df.groupby('gs_text34').fever.sum()
b = df.groupby('gs_text34').count.sum()
end_ratio = a/b
print end_ratio.index[end_ratio.argmax()], end_ratio[end_ratio.argmax()]
```

```python
############ Question 3 Display the cause-specific endorsement rates visually ##########
(df.groupby('gs_text34').fever.sum()/df.groupby('gs_text34').count.sum()).order().plot(kind='barh', figsize=(12,12))
```

```python
######### Question 4 Find the question that asks if the deceased had AIDS ########
temp1 = cb.question.str[0:3]=="Did"
temp2 = cb.ix[temp1]
temp3 = temp2.question.str[-10:]=="Have AIDS?"
answer = temp2.ix[temp3]
var2 = answer.ix[int(answer.index),0]
print var2
```

```python
######## Question 5 Make a 2x2 table showing the number of adult deceased with and without this question endorsed for decedents with underlying cause AIDS and not AIDS #############
dfa = df[df.module == "Adult"]
dfa['AIDS'] = df[var2] == 'Yes'
# with AIDS
dfAIDS = dfa[dfa.gs_text34=='AIDS']
a = dfAIDS.groupby('count').AIDS.sum()
b = dfAIDS.groupby('count').count.sum()
count_wAIDS_end = a
count_wAIDS_nonend = b-a
# without AIDS
dfWoAIDS = dfa[dfa.gs_text34!='AIDS']
a = dfWoAIDS.groupby('count').AIDS.sum()
b = dfWoAIDS.groupby('count').count.sum()
count_woAIDS_end = a
count_woAIDS_nonend = b-a
AIDS_table = pd.DataFrame({'With_AIDS': [int(count_wAIDS_end), int(count_wAIDS_nonend)],
         'Without_AIDS': [int(count_woAIDS_end), int(count_woAIDS_nonend)]})
AIDS_table.index = ['Endorsed', 'Not endorsed']
print AIDS_table
```

```python
########## Question 6 Now do a series of 2x2 tables comparing the percent of deaths truely due to AIDS for which "Had AIDS" question was endorsed #########
AIDS_percent_table_00 = float(AIDS_table.ix[0,0])/(AIDS_table.ix[0,0]+AIDS_table.ix[0,1])
AIDS_percent_table_01 = float(AIDS_table.ix[0,1])/(AIDS_table.ix[0,0]+AIDS_table.ix[0,1])
AIDS_percent_table_10 = float(AIDS_table.ix[1,0])/(AIDS_table.ix[1,0]+AIDS_table.ix[1,1])
AIDS_percent_table_11 = float(AIDS_table.ix[1,1])/(AIDS_table.ix[1,0]+AIDS_table.ix[1,1])
AIDS_percent_table = pd.DataFrame({'Truly_With_AIDS': [AIDS_percent_table_00, AIDS_percent_table_10],
         'Actually_without_AIDS': [AIDS_percent_table_01, AIDS_percent_table_11]})
AIDS_percent_table.index = ['Endorsed', 'Not endorsed']
print AIDS_percent_table
```

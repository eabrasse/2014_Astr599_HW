```python
def drumhead_height(n, k, r, theta, t):
    import scipy
    from scipy.special import jn, jn_zeros
    nth_zero = jn_zeros(n, k)[k-1]
    return (scipy.cos(t)*scipy.cos(n*theta)*jn(n, r*nth_zero))
```

```python
def plot_Bessel(t):
    import matplotlib.pyplot as plt
    import numpy as np
    from mpl_toolkits.mplot3d import Axes3D
    
    # create a 3*3 grid object
    gs = plt.GridSpec(3, 3, wspace=0.2, hspace=0.2)
    
    # create a figure
    fig = plt.figure(figsize=(16,10))
    
    # create a grid
    r = np.linspace(0, 1, 101)
    theta = np.linspace(0, 2*np.pi, 361)
    grid_polar = np.meshgrid(r, theta)
    rr = grid_polar[0]
    thetatheta = grid_polar[1]
    xx = rr*np.cos(thetatheta)
    yy = rr*np.sin(thetatheta)
    
    # set specs for plotting
    specs = {'cmap':'seismic', 'vmin':-0.5, 'vmax':0.5}
    
    # Subplot 1
    ax = fig.add_subplot(gs[0:3, 0:2], projection='3d')
    elev = drumhead_height(n=2, k=3, r=rr, theta=thetatheta, t=t)
    cs = ax.plot_surface(xx, yy, elev, linewidth=0, **specs)
    #cset = ax.contour(xx, yy, elev, zdir='elev', **specs)
    ax.set_zlim([-0.5, 0.5])
    plt.colorbar(cs, shrink=0.7)
    plt.title('n=2, k=3', fontsize=16)
    
    # Subplot 2
    ax = fig.add_subplot(gs[0, 2], projection='3d')
    elev = drumhead_height(n=2, k=5, r=rr, theta=thetatheta, t=t)
    cs = ax.plot_surface(xx, yy, elev, linewidth=0, **specs)
    #cset = ax.contour(xx, yy, elev, zdir='elev', **specs)
    ax.set_zlim([-0.5, 0.5])
    plt.title('n=2, k=5', fontsize=20)
    
    # Subplot 3
    ax = fig.add_subplot(gs[1, 2], projection='3d')
    elev = drumhead_height(n=2, k=10, r=rr, theta=thetatheta, t=t)
    cs = ax.plot_surface(xx, yy, elev, linewidth=0, **specs)
    #cset = ax.contour(xx, yy, elev, zdir='elev', **specs)
    ax.set_zlim([-0.5, 0.5])
    plt.title('n=2, k=10', fontsize=16)
    
    # Subplot 4
    ax = fig.add_subplot(gs[2, 2], projection='3d')
    elev = drumhead_height(n=5, k=3, r=rr, theta=thetatheta, t=t)
    cs = ax.plot_surface(xx, yy, elev, linewidth=0, **specs)
    #cset = ax.contour(xx, yy, elev, zdir='elev', **specs)
    ax.set_zlim([-0.5, 0.5])
    plt.title('n=5, k=3', fontsize=16)
    fig.savefig('drum_t%.1f.png' %t, format='png')
    plt.show()
```

```python
plot_Bessel(t=0)
plot_Bessel(t=0.5)
plot_Bessel(t=1)
plot_Bessel(t=1.5)
plot_Bessel(t=2)

# For making video part, because I'm using Python on my lab system and I don't have administrator rights,
# so I didn't do it in Python. Instead, I saved all the figures and used alternative software to generate
# a short video, see submission.
```

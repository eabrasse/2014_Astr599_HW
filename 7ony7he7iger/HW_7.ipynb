### Classification: Labeling Photometric Sources

Recall the packages I need

```python
%matplotlib inline
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from astroML.datasets import fetch_rrlyrae_combined
from sklearn.cross_validation import train_test_split

X, y = fetch_rrlyrae_combined()

# For now, we'll only fit the first two colors
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
N_plot = 5000
plt.scatter(X[-N_plot:, 0], X[-N_plot:, 1], c=y[-N_plot:],
            edgecolors='none')
plt.xlabel('u-g color')
plt.ylabel('g-r color');
```

```python
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier(n_neighbors=5)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
```

```python
print(np.sum(y_pred == y_test))
print(np.sum(y_pred != y_test))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred,
                            target_names=['MS star', 'RR Lyrae']))
```

### Regression: Photometric Redshifts

```python
from astroML.datasets import fetch_sdss_specgals
```

```python
data = fetch_sdss_specgals()

# put magnitudes in a matrix
X = np.vstack([data['modelMag_%s' % f] for f in 'ugriz']).T
y = data['z']

# Split into training and testing data
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
from sklearn.linear_model import LinearRegression
est = LinearRegression()
est.fit(X_train, y_train)
y_pred = est.predict(X_test)
```

```python
plt.plot(y_test, y_pred, ',k')
plt.plot([0, 1], [0, 1], ':k')
plt.xlim(0, 0.6)
plt.ylim(0, 0.6)
plt.title('Linear Regression');
```

```python
rms = np.sqrt(np.mean((y_test - y_pred) ** 2))
print(rms)
```

### Exercise 2: Improving the regression

##### K-neighbors regressor

```python
from sklearn.neighbors import KNeighborsRegressor
est = KNeighborsRegressor()
est.fit(X_train, y_train)
y_pred = est.predict(X_test)
```

```python
plt.plot(y_test, y_pred, ',k')
plt.plot([0, 1], [0, 1], ':k')
plt.xlim(0, 0.6)
plt.ylim(0, 0.6)
plt.title('K-Neighbors Regression');
```

```python
rms = np.sqrt(np.mean((y_test - y_pred) ** 2))
print(rms)
```

More accurate than the normal linear regression

##### epsilon-Support Vector Regression

```python
from sklearn.svm import SVR
est = SVR()
est.fit(X_train[::5], y_train[::5])
y_pred = est.predict(X_test)
```

```python
plt.plot(y_test, y_pred, ',k')
plt.plot([0, 1], [0, 1], ':k')
plt.xlim(0, 0.6)
plt.ylim(0, 0.6)
plt.title('SVR');
```

```python
rms = np.sqrt(np.mean((y_test - y_pred) ** 2))
print(rms)
```

Showed quite high rms. Skipped the entire data to save time.

##### Random Forest Regression

```python
from sklearn.ensemble import RandomForestRegressor
est = RandomForestRegressor()
est.fit(X_train, y_train)
y_pred = est.predict(X_test)
```

```python
plt.plot(y_test, y_pred, ',k')
plt.plot([0, 1], [0, 1], ':k')
plt.xlim(0, 0.6)
plt.ylim(0, 0.6)
plt.title('Random Forest Regression');
```

```python
rms = np.sqrt(np.mean((y_test - y_pred) ** 2))
print(rms)
```

##### Gradient Boosting Regression

```python
from sklearn.ensemble import GradientBoostingRegressor
est = GradientBoostingRegressor()
est.fit(X_train[::5], y_train[::5])
y_pred = est.predict(X_test)
```

```python
plt.plot(y_test, y_pred, ',k')
plt.plot([0, 1], [0, 1], ':k')
plt.xlim(0, 0.6)
plt.ylim(0, 0.6)
plt.title('Gradient Boosting Regression');
```

1 Create a function for drumhead height

```python
from scipy import *
import numpy as np
from scipy.special import jn, jn_zeros
import matplotlib.pyplot as plt


def drumhead_height(n,k,r,theta,t):

    u = np.cos(t)*np.cos(n*theta)*jn(n,jn_zeros(n,k)[k-1]*r)
    return u    

```

2 Create function for plot_Bessel, 3 create arrays for the grid, 4, calculate the drumhead height for various n,k values, 5, use plot_surface to 3d plot

```python
from mpl_toolkits.mplot3d import axes3d
from matplotlib import cm
def plot_Bessel(t):
    
    theta = np.linspace(0,2*np.pi, 100)
    r = np.linspace(-2, 2, 100) 
    rm, thetm = meshgrid(r, theta)
    
    x=rm*np.cos(thetm)
    y=rm*np.sin(thetm)
    time=t/10.
    
    gs = plt.GridSpec(3, 3) 
    fig = plt.figure(figsize=(10,8))
    
    ax1 = fig.add_subplot(gs[0:, 0:2],projection='3d')
    u=drumhead_height(2.,3.,rm,thetm,time)    
    ax1.plot_surface(x, y, u, rstride=1, cstride=1, alpha=0.4, cmap=cm.seismic)
    ax1.set_zlim(-0.7, 0.7)
    ax1.set_axis_off()
    
    ax2 = fig.add_subplot(gs[0,2],projection='3d')
    u=drumhead_height(3,4,rm,thetm,time)
    ax2.plot_surface(x, y, u, rstride=1, cstride=1, linewidth=0, alpha=0.4,cmap=cm.seismic)
    ax2.set_zlim(-0.7, 0.7)
    ax2.set_axis_off()
    
    ax3 = fig.add_subplot(gs[1,2],projection='3d')
    u=drumhead_height(2,5,rm,thetm,time)
    ax3.plot_surface(x, y, u, rstride=1, cstride=1, linewidth=0, alpha=0.4,cmap=cm.seismic)
    ax3.set_zlim(-0.7, 0.7)
    ax3.set_axis_off()
    
    ax4 = fig.add_subplot(gs[2,2],projection='3d')
    u=drumhead_height(5,5,rm,thetm,time)
    ax4.plot_surface(x, y, u, rstride=1, cstride=1, linewidth=0, alpha=0.4,cmap=cm.seismic)
    ax4.set_zlim(-0.7, 0.7)
    ax4.set_axis_off()
    
    plt.savefig('animation_frames/frame{0:07d}.png'.format(t+1));
    plt.close()
```

6, use JSAnimation to plot 5 timesteps

```python
#%matplotlib inline
from JSAnimation import IPython_display, HTMLWriter
from matplotlib import animation
import Image
import glob

n = 5 # Use 100 to produce bigger animation
print('Frame number:')
for i in xrange(0,n,1):
    plot_Bessel(i)
    print i
```

```python
import JSAnimation
filenames=sorted(glob.glob('animation_frames/frame*.png'))
print(filenames)
fig = plt.figure(figsize=(16,6))
ax = fig.add_axes([0, 0, 1, 1])
ax.axis('off')
im = plt.imshow(Image.open(filenames[0]))

#initialization function
def init():
    im.set_data(Image.open(filenames[0]))
    return im,

#animation function
def animate(i):
    image=Image.open(filenames[i])
    im.set_data(image)
    return im,

anim = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=5, interval=20, blit=True)

#set embed_frames=True to embed base64-encoded frames directly in the HTML
anim.save('animation_frames/animation.html', writer=HTMLWriter(embed_frames=True))

```

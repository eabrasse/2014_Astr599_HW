```python
import numpy as np
import matplotlib.pyplot as plt
%pylab inline
```

# Part 1

1.1 Load in data

```python
data = np.genfromtxt("asteroids5000.csv", delimiter=",")
```

1.2. count number of NaNs and number of complete rows

```python
x=np.isnan(data)
print("number of values missing in this data: ",x.sum())
```

```python
print("number of complete rows: ",(np.sum(x,1)==0).sum())
```

1.3. create new array with only complete rows

```python
y=data[np.where(np.sum(x,1)==0)]
```

1.4 calculate max, min, mean, and std of each column

```python
print("maximum of each column: ",np.max(y,0))
print("minimum of each column: ",np.min(y,0))
print("mean of each column: ",np.mean(y,0))
print("standard deviation of each column: ",np.std(y,0))
```

2.3 use plt.scatter to make scatterplot of semi-major axis vs sine of inclination angle (which I assume are a and i)

```python
plt.scatter(y[:,0],np.sin(y[:,2]*np.pi/180))
plt.xlabel('a')
plt.ylabel('sin(i)')
```

The distribution is clustered somewhat regularly at certain values of semi-major axis.  The preferred semi-major axes could be explained by resonance of the orbits with the gravitational attraction of the other planets, especially Jupiter.  

2.4 Create H vs BV plot and overplot a line that divides the two families

```python
plt.scatter(y[:,3],y[:,5])
famline=np.mean(y[:,5])
plt.plot(y[:,3],famline*np.ones(np.shape(y[:,3])),linewidth=2,color='k')
plt.xlabel('H')
plt.ylabel('BV')
```

2.5 repeat orbit plot but plot with different colors for the two families 

```python
plt.scatter(y[y[:,5]>famline,0],np.sin(y[y[:,5]>famline,2]*np.pi/180),color='c')
plt.scatter(y[y[:,5]<famline,0],np.sin(y[y[:,5]<famline,2]*np.pi/180),color='m')
plt.xlabel('a')
plt.ylabel('sin(i)')
```

The asteroid family that has higher BV tend to have orbits closer to the sun than the asteroid family with the lower BV value.  In particular, the orbit near 3.5 is almost exclusively lower B-V values asteroids.

2.6 Compare to plots Parker et al 2008:

In Figure 3 of Parker et al (2008), they plot a vs sin(i) and also have gaps present at certain orbits.  For the spectral families that they define, they also find that certain spectral families are more likely to be found at certain orbits.  For example, their "C" family is more likely to be find at outer orbits and less likely to be found at inner and medium orbits.  The envelope of their a vs sin(i) curve is similar to ours here for inner orbits, but Parker et al's dataset has more asteroids with high sin(i) at the outer orbits.

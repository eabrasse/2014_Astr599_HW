## Part 0

```python
def mandel(x, y, max_iter):
    """
    Given z = x + iy and max_iter, determine whether the candidate
    is in the mandelbrot set for the given number of iterations
    """
    c = complex(x, y)
    z = 0.0j
    for i in range(max_iter):
        z = z*z + c
        if (z.real*z.real + z.imag*z.imag) >= 4:
            return i

    return max_iter

def create_fractal(Nx, Ny, xmin, xmax, ymin, ymax, max_iter):
    """Create and return a fractal image"""
    image = np.zeros((Ny, Nx), dtype=float)
    dx = (xmax - xmin) * 1. / Nx
    dy = (ymax - ymin) * 1. / Ny
    
    for x in range(Nx):
        rpart = xmin + x * dx
        for y in range(Ny):
            ipart = ymin + y * dy
            color = mandel(rpart, ipart, max_iter)
            image[y, x] = color
    return image
```

## Part 2

Use Numba

```python
from numba import autojit
from pylab import imshow, show
import matplotlib.pyplot as plt   
import numpy as np
from timeit import default_timer as timer
@autojit
def mandel_n(x, y, max_iters):
  c = complex(x, y)
  z = 0.0j
  for i in range(max_iters):
    z = z*z + c
    if (z.real*z.real + z.imag*z.imag) >= 4:
      return i

  return max_iters

@autojit
def create_fractal_n(min_x, max_x, min_y, max_y, image, iters):
  height = image.shape[0]
  width = image.shape[1]

  dx = (max_x - min_x) / width
  dy = (max_y - min_y) / height
    
  for x in range(width):
    real = min_x + x * dx
    for y in range(height):
      imag = min_y + y * dy
      color = mandel_n(real, imag, iters)
      image[y, x] = color
```

Next we create a 1024x1024 pixel image as a numpy array of bytes. We then call create_fractal with appropriate coordinates to fit the whole mandelbrot set.

```python
image_n = np.zeros((1024, 1536), dtype = np.uint8)
###another way for timer
start = timer()
create_fractal_n(-2.0, 1.0, -1.0, 1.0, image_n, 20) 
dt = timer() - start

print "Mandelbrot created in %f s" % dt
plt.imshow(image_n)
#show()
```

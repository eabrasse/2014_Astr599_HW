# Distributed computation of the Mandelbrot Set using IPython Parallel

###### Author: [Cosmo Harrigan](https://github.com/cosmoharrigan), based on Mandelbrot Set code by [Jake Vanderplas](https://github.com/jakevdp)

This notebook illustrates the use of [IPython Parallel](http://ipython.org/ipython-doc/dev/parallel/) for distributed computing over a cluster of nodes. The [Mandelbrot Set](http://en.wikipedia.org/wiki/Mandelbrot_set) fractal is computed according to the chosen parameters, and the time required to perform the computation is calculated on the local machine and on the cluster for comparison purposes.

Before executing this notebook, make sure that you have started the cluster, on the *Clusters* tab of IPython Notebook.

### Import the required libraries

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import fractal
```

### Setup the parallel computing clients

```python
from IPython import parallel
clients = parallel.Client()
clients = parallel.Client(profile='default')
print("Number of clients: {0}".format(len(clients)))
%px import fractal
```

```python
lview = clients.load_balanced_view()
lview.block = True
dview = clients.direct_view()
```

```python
num_chunks = len(clients)
```

### Configure the desired parameter settings for the Mandelbrot Set fractal

```python
Nx = 1500
Ny = 1000
xmin = -2
xmax = 1
ymin = -1
ymax = 1
max_iter = 100
```

Send the parameters to the clients:

```python
params = dict(Nx=Nx, Ny=Ny, xmin=xmin, xmax=xmax, ymin=ymin, ymax=ymax, max_iter=max_iter, num_chunks=num_chunks)
dview.push(params)
```

Only the chunk number will be an iterable, while the fractal parameters will remain the same, so in order to use map with distributed computing, we will need to wrap the compute_region function in a one-parameter lambda function.

```python
compute_region_lambda = lambda chunk: fractal.compute_region(chunk, num_chunks, Nx, Ny, xmin, xmax, ymin, ymax, max_iter)
```

### Comparison of results

By dividing the computation between multiple cores in the cluster, the total time required to compute the set is reduced. This can be confirmed by comparing the results below.

#### Test using one core, with the local **map** function

Check the execution time:

```python
%timeit map(compute_region_lambda, range(num_chunks))
```

Display the result:

```python
regions_local = map(compute_region_lambda, range(num_chunks))
```

```python
plt.imshow(np.hstack(regions_local), cmap=plt.cm.jet)
```

#### Test using multiple cores, with the IPython Parallel **map** function:

Check the execution time:

```python
%timeit lview.map(compute_region_lambda, range(num_chunks))
```

Display (and save) the result:

```python
regions_distributed = lview.map(compute_region_lambda, range(num_chunks))
```

```python
plt.axis('off')
plt.imshow(np.hstack(regions_distributed), cmap=plt.cm.jet)
plt.savefig("fractal.png", dpi=400)
```

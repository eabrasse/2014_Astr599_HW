# Homework 1: asteroid data

## Part 1: Loading and exploring the data

Load the data:

```python
import numpy as np
```

```python
data_raw = np.genfromtxt('asteroids5000.csv', delimiter=',')
```

Find out how many values are missing from the data:

```python
np.isnan(data_raw).sum()
```

Find out how many **rows** are complete, with no missing values:

```python
data_raw[~np.isnan(data_raw).any(axis=1)].size
```

Create a new array containing only the rows with no missing values:

```python
data = data_raw[~np.isnan(data_raw).any(axis=1)]
```

Compute the maximum, minimum, mean, and standard deviation of the values in each column:

**Maximum**

```python
np.max(data, axis=0)
```

**Minimum**

```python
np.min(data, axis=0)
```

**Mean**

```python
np.mean(data, axis=0)
```

**Standard Deviation**

```python
np.std(data, axis=0)
```

## Part 2: Plotting the data

Display the column names for the dataset:

```python
!head -n1 asteroids5000.csv
```

Here is what each column header represents, based on descriptions from the [JPL site](http://ssd.jpl.nasa.gov/sbdb_query.cgi) which is the source of the data:
* **a** semi-major axis (AU)
* **e** eccentricity
* **i** inclination (deg)
* **H** absolute magnitude parameter (mag)
* **diameter** object diameter (km)
* **BV**
* **UB**

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

Plot the semi-major axis versus the sine of the inclination angle in radians.

```python
plt.scatter(np.sin(np.deg2rad(data[:,2])), data[:,0]);
plt.xlabel('sine of inclination angle');
plt.ylabel('semi-major axis (AU)');
```

Based on this plot, it can be observed that there is a fairly dense cluster of similar observations, and then a smaller number of outliers spread around the central cluster.

Plot a color-magnitude diagram of the asteroids (H vs B-V). You should see two distinct "families" of asteroids indicated in this plot. Over-plot a line that divides these.

```python
plt.scatter(data[:,5], data[:,3], c=data[:,3]);
plt.plot([0.78, 0.78], [3, 15], 'k-', linewidth=3)
plt.xlabel('BV');
plt.ylabel('magnitude');
```

Repeat the orbital parameter plot from above, but plot the two "families" in different colors.

```python
plt.scatter(np.sin(np.deg2rad(data[:,2])), data[:,0], c=data[:,5]>0.78);
plt.xlabel('sine of inclination angle');
plt.ylabel('semi-major axis (AU)');
```

Based on this plot, we do not observe any clear correlation between color and orbit. This naturally raises the question of whether that means there is no such correlation, or if the wrong combination of parameters have been chosen in this plot configuration.

```python
import numpy as np
data = np.genfromtxt("data/asteroids5000.csv",delimiter = ',')
print data
```

```python
missingDataLen = len(data[np.isnan(data)])
print missingDataLen

mask = ~(np.isnan(data).any(1))
print len(data[mask,:])
```

```python
newData = data[mask]
newData
```

```python
print np.max(newData, axis = 0)
print np.min(newData, axis = 0)
print np.mean(newData, axis = 0)
print np.std(newData, axis = 0)
```

```python
plt.scatter(data[:,3],data[:,5])
plt.xlabel('H')
plt.ylabel('B-V')
plt.ylim(0.5,1.1)
plt.xlim(5,15)
plt.axhline(y = 0.79, linewidth=4, color='r')
```

```python
mask1 = np.where(data[:,5] >.79);
mask2 = np.where(data[:,5] <= .79);
plt.scatter(data[mask1,0],np.sin(data[mask1,2]/180 * np.pi), color = 'b')
plt.scatter(data[mask2,0],np.sin(data[mask2,2]/180 * np.pi), color = 'r', alpha = 0.3)
plt.xlabel('Semi-major axis')
plt.ylabel('Sine of inclination angle')
```

It seems that there are some correlations between semi-major axis and colors, but color itself

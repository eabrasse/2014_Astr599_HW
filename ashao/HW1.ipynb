```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
# Load data from given file
datafile = '/home/ashao/uw-apl/classes/ASTR599/2014_fall_ASTR599/notebooks/data/asteroids5000.csv'
ast_dat = np.genfromtxt(datafile,delimiter=',',skip_header=1)
```

```python
ast_shape = np.shape(ast_dat)
num_nan_total = np.sum(np.isnan(ast_dat));
num_nan_rows = np.sum(np.isnan(ast_dat),1) # Sum along isnan in a row
num_rows_missing_data = sum(num_nan_rows != 0)
num_rows_good = sum(num_nan_rows == 0)
print('Datafile contains\n\tNaNs: %d\n\tRows with missing data: %d\n\tRows with no missing data: %d'% (num_nan,num_rows_missing_data,num_rows_good))
```

```python
ast_dat_good = ast_dat[num_nan_rows==0,:]
ast_dat_good.max(0)
np.set_printoptions(precision=2,suppress=True)
print('DATA STATISTICS')
print('Maximum:')
print(ast_dat_good.max(0))
print('Minimum:')
print(ast_dat_good.min(0))
print('Mean:')
print(ast_dat_good.mean(0))
print('St. Dev:')
print(ast_dat_good.std(0))
```

```python
!head -n 1 $datafile
plt.scatter(ast_dat_good[:,0],np.sin(np.deg2rad(ast_dat_good[:,2])))
plt.xlabel('Semi-Major Axis')
plt.ylabel('Inclination')
```

The above plot shows distinct groups of asteroids with the similar semi-major axes but differing levels of inclination. Perhaps the asteroid belt is comprised of well-organized orbits (due to the effect of the interaction of the sun and planetary gravity wells) with objects primarily above the solar system's plane.

```python
plt.scatter(ast_dat_good[:,3],ast_dat_good[:,5])
critvalue = 0.79;
plt.plot([2,14],[critvalue, critvalue])
plt.xlabel('H')
plt.ylabel('BV')
```

```python
fam1 = ast_dat_good[:,5]<critvalue;
fam2 = ast_dat_good[:,5]>=critvalue;
plt.scatter(ast_dat_good[fam1,0],np.sin(np.deg2rad(ast_dat_good[fam1,2])),color='blue')
plt.scatter(ast_dat_good[fam2,0],np.sin(np.deg2rad(ast_dat_good[fam2,2])),color='red')
plt.xlabel('Semi-Major Axis [AU]')
plt.ylabel('Inclination [sin(d)]')
```

This seems similar to the Figure 5 in the paper which suggests that closer objects tend to have higher BV levels than thosew further away.

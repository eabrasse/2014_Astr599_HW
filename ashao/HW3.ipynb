```python
%matplotlib inline

import numpy as np
from scipy import optimize
import matplotlib.pyplot as plt
import astroML.datasets
import random
from astroML.time_series import lomb_scargle

class Fourier(object):
    def __init__(self, phase, flux, dflux, nterms):
        self.phase = phase
        self.flux = flux
        self.dflux = dflux
        self.nterms = nterms
        assert(self.nterms > 0)
        
    def evaluate(self, phase, terms):
        assert(len(terms) == 2 * self.nterms - 1)
        model = terms[0] * np.ones(len(phase))
        for i in range(self.nterms-1):
            model += terms[2*i + 1] * np.cos(2 * np.pi * (i+1) * phase + terms[2*i + 2])
        return model
    
    def chi2(self, args):
        model = self.evaluate(self.phase, args)
        chi   = (model - self.flux) / self.dflux
        return np.sum(chi**2)
    
    def makeGuess(self):
        guess = []
        guess.append(np.mean(self.flux))
        if self.nterms > 1:
            guess.append(np.std(self.flux))
            guess.append(2 * np.pi * (1 - self.phase[np.argsort(self.flux)[-1]]))
            for i in range(2, self.nterms):
                guess.append(0.0)
                guess.append(2 * np.pi * (1 - self.phase[np.argsort(self.flux)[-1]]))
        return guess


        
```

```python
# Load sampling of LINEAR data and choose a random star to analyze
lcs = astroML.datasets.fetch_LINEAR_sample()
randid = lcs.ids[random.randint(0,np.size(lcs.ids)-1)]
```

```python
# Extract some of the data into appropriate arrays and plot
time, flux, dflux = lcs[randid].T
plt.errorbar(time, flux, yerr=dflux, fmt="ro")
plt.gca().invert_yaxis()
plt.xlabel("Mean Julian Day")
plt.ylabel("Mag")
plt.show()
```

```python
# Calculate Lomb-Scargle to find periodagram
periods = np.logspace(-1, 0, 10000, base=10)
periodogram = lomb_scargle(time, flux, dflux, omega=2 * np.pi / periods, generalized=True)
plt.plot(periods, periodogram)
plt.semilogx()
plt.xlabel("Period (days)")
plt.ylabel("Power")
plt.show()
```

```python
# Fold around the freqency with maximal spectral power
idx = np.argmax(periodogram)
period = periods[idx]
phase = time / period - time // period 
plt.errorbar(phase, flux, yerr=dflux, fmt="ro")
plt.gca().invert_yaxis()
plt.xlabel("Phase")
plt.ylabel("Mag")
plt.show()
```

```python
# Now find the number of fourier fits that best explain the star
maxterm = 30;
nterms     = np.arange(1, maxterm, 2)
chi2       = []
chi2_BIC = []
chi2_AIC = []

for nterm in nterms:
    fourier = Fourier(phase, flux, dflux, nterm)
    guess   = fourier.makeGuess()
    chi2_term = optimize.fmin_bfgs(fourier.chi2, x0=[guess], disp=0, full_output=1)[1]
    chi2.append(chi2_term)
    chi2_BIC.append(chi2_term + len(guess)*np.log(len(fourier.phase)))
    chi2_AIC.append(chi2_term + len(guess)*2)
```

```python
chi2 = np.array(chi2)
chi2_BIC = np.array(chi2_BIC)
chi2_AIC = np.array(chi2_AIC)
plt.plot(nterms, chi2, label="chi2")
plt.plot(nterms, chi2_BIC, label="BIC")
plt.plot(nterms, chi2_AIC, label="AIC")
plt.xlabel("N Terms")
plt.ylabel("Chi2")
plt.semilogy()
plt.legend()
plt.show()
optimal_terms = nterms[np.argmin(chi2_BIC)],nterms[np.argmin(chi2_AIC)]
print optimal_terms
```

```python
print('Optimal number of terms for object %d\nBIC:%d\nAIC:%d' % (randid,optimal_terms[0],optimal_terms[1]))
if optimal_terms[0] == nterms[-1]:
    print('WARNING: BIC has not yet converged. Increase maxterms and try again')
if optimal_terms[1] == nterms[-1]:
    print('WARNING: AIC has not yet converged. Increase maxterms and try again')
```

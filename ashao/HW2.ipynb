```python
import numpy as np
import matplotlib.pyplot as plot
```

```python
def sine(x):
    return np.sin(x)

class Integrator(object):
    def __init__(self, a, b, n):
        self.a, self.b, self.n = a, b, n
        self.points, self.weights = self.method()
        
    def method(self):
        raise NotImplementedError('no rule in class %s' %self.__class__.__name__)
        
    def integrate(self,func):
        return np.sum(self.weights*func(self.points))

class Trapezoidal(Integrator):
    
    def __init__(self, a, b, n):
        super(Trapezoidal,self).__init__(a, b, n)    
        
    def method(self):
        x = np.linspace(self.a,self.b,self.n)        
        w = np.ones(np.size(x))*2
        w[0] = 1
        w[-1]= 1
        w = (self.b-self.a)/(self.n)        
        return x, w
        
class Midpoint(Integrator):
    
    def __init__(self, a, b, n):     
        super(Midpoint,self).__init__(a, b, n)
        
    def method(self):
        x = np.linspace(self.a,self.b,self.n)
        xmid = 0.5*(x[1:]+x[0:-1])        
        w = (x[1]-x[0])
        return xmid, w
    
def plot_quad_methods(x,func,n=5):
    a = 0    
    midx, midw = Midpoint(a,x,n).method()
    trapx, trapw = Trapezoidal(a,x,n).method()
    
    plot.subplot(1,2,1)
    plot.plot(midx,np.cumsum(func(midx)*midw),label='Midpoint')
    plot.plot(trapx,np.cumsum(func(trapx)*trapw),label='Trapezoid')
    plot.plot(trapx,1-np.cos(trapx),label='Analytic Solution')
    plot.title('Quadrature Estimate')
    plot.legend(loc=4)
    
    plot.subplot(1,2,2)
    plot.plot(midx,(1-np.cos(trapx[1:]))-np.cumsum(func(midx)*midw),label='Midpoint')
    plot.plot(trapx,(1-np.cos(trapx))-np.cumsum(func(trapx)*trapw),label='Trapezoid')
    plot.title('Difference from analytic solution')
    plot.legend(loc=4)
    #Midpoint(a,x,n)
    

```

```python
%matplotlib inline
from IPython.html.widgets import interact
def interact_driver(n):
    plot_quad_methods(2*np.pi,sine,n)

interact(interact_driver,n=(1,100))
```

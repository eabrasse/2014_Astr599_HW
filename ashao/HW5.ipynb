```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
from pylab import *
import scipy as sp
from mpl_toolkits.mplot3d import Axes3D
from scipy.special import jn,jn_zeros
```

```python
def drum_height(n,k,r,theta,t):
    lambda_sol = jn_zeros(n,k)[k-1]    
    return np.cos(t)*np.cos(n*theta)*jn(n,lambda_sol*r)
```

```python
def plot_Bessel(t):

    # Define the figure properties
    fig = plt.figure(figsize=(16,6));
    gs = plt.GridSpec(3,3,wspace=0.2,hspace=0.2)
    
    # Setup plotting grid both in polar and Cartesian Coordinates
    r = np.linspace(0,1,100)
    theta = np.linspace(0,2*np.pi,100)
    rgrid, thetagrid = np.meshgrid(r,theta)
    xgrid = rgrid*np.cos(thetagrid)
    ygrid = rgrid*np.sin(thetagrid)
    
    # Plot the results
    specs = {'cmap' : 'seismic', 'vmin' : -1, 'vmax' : 1,'rstride' : 1, 'cstride' : 2, 'linewidth' : 0}
    
    # Case: n=2, k=3
    ax1 = fig.add_subplot(gs[:3, :2],projection='3d')
    ax1.plot_surface(xgrid,ygrid,drum_height(2,3,rgrid,thetagrid,t),**specs)
    ax1.set_zlim(-1,1)
    
    #Case n=3,k=2
    ax2 = fig.add_subplot(gs[0, 2],projection='3d')
    ax2.plot_surface(xgrid,ygrid,drum_height(3,2,rgrid,thetagrid,t),**specs)
    ax2.set_zlim(-1,1)
    
    #Case n=4,k=1
    ax3 = fig.add_subplot(gs[2, 2],projection='3d')
    ax3.plot_surface(xgrid,ygrid,drum_height(4,1,rgrid,thetagrid,t),**specs)
    ax3.set_zlim(-1,1)
    return fig
    
    
```

```python
# Make the figures for animation
nframes = 5
for t in xrange(0,nframes,1):
    fig = plot_Bessel(t);
    fig.savefig('hw5_figs/frame{0:07d}.png'.format(t+1))    
    print t
    

```

```python
from JSAnimation import IPython_display, HTMLWriter
from matplotlib import animation
import Image
import glob

#Choose location of files to use
filenames=sorted(glob.glob('hw5_figs/frame*.png'))

#Create figure
fig = plt.figure(figsize=(10,5))
ax = fig.add_axes([0, 0, 1, 1])
ax.axis('off')
im = plt.imshow(Image.open(filenames[0]))

# Create initializatio function
def init():
    im.set_data(Image.open(filenames[0]))
    return im,

# Create animation function
def animate(i):
    image=Image.open(filenames[i])
    im.set_data(image)
    return im,

anim = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=nframes, interval=20, blit=True)

#set embed_frames=True to embed base64-encoded frames directly in the HTML
anim.save('hw5_figs/animation.html', writer=HTMLWriter(embed_frames=True))

# Show animation in ipython notebook
anim
```

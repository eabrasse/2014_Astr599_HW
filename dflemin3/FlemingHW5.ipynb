```python
#Imports
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy import *
from scipy.special import jn, jn_zeros

from mpl_toolkits.mplot3d import axes3d
from matplotlib import cm
```

1) Drumhead Height function

```python
def drumhead_height(n=0,k=1,r=0,theta=0,t=0):
    u = np.cos(t)*np.cos(n*theta)
    
    #Compute kth zero of the nth bessel function
    la = jn_zeros(n,k)[k-1]
    
    #Compute J(z*r) of order n
    return (u * jn(n,la*r))
```

2,3,4,5) Create Plot bessel function according to directions in HW.  See comments in code that indicate what each section does.

```python
def plot_Bessel(t = 0):
    #Create two arrays
    
    #Length of an array dimension
    length = 100
    
    #r [0,1], theta [0,2pi]
    rArr = np.linspace(0,1,length)
    thetaArr = np.linspace(0,2*np.pi,length)
    
    #Use meshgrid to create polar coords grid
    rm, thetam = meshgrid(rArr, thetaArr)
    
    #Create new x,y mesh-sized arrays
    xArr = np.empty_like(rm)
    yArr = np.empty_like(thetam)
    
    #Convert to Cartesian
    xArr = np.multiply(rm,np.cos(thetam))
    yArr = np.multiply(rm,np.sin(thetam))
      
    #Transform it into a cartesian grid
    
    # Create a plotting grid with gridspec
    # Create a 3 x 3 plotting grid object with horizontal and vertical spacing wspace and hspace, 4 subplots
    gs = plt.GridSpec(3, 3, wspace=-.4, hspace=0.5) 
    # Create a figure
    fig = plt.figure(figsize=(16, 10))

    #Create subfigures
    
    # SUBFIGURE 1
    #Takes up 3 rows, 1st 2 cols
    ax1 = fig.add_subplot(gs[:, :1])
    
    #Compute z with given parameters
    z = drumhead_height(2,3,rm,thetam,t)
    
    #Subplot 1 Plotting
    # Plot using pcolor and arrange colorbar
    colormap = 'RdBu'
    s1 = ax1.pcolor(rm,thetam,z, cmap = colormap, vmin=-0.3, vmax=.5) # vmin and vmax are min and max values of colormap
    cbar = fig.colorbar(s1) # Show colorbar
    cbar_ticks = np.linspace(-0.2,0.7,5) 
    cbar.set_ticks(cbar_ticks) # adjust colorbar ticks
    
    ax1.set_xlabel('r', fontsize = 20)
    ax1.set_ylabel('theta', fontsize = 20)
    ax1.set_ylim([0,2*np.pi])
    ax1.set_title('Bessel n = 2, k = 3', fontsize = 20)
    
    # SUBFIGURE 2
    # Create subfigure 2 1st row, last col
    #Compute z with given parameters
    z = drumhead_height(1,1,rm,thetam,t)
    
    ax2 = fig.add_subplot(gs[0, -1])
    s2 = ax2.pcolor(rm,thetam,z, cmap = colormap, vmin=-0.3, vmax=.5) # vmin and vmax are min and max values of colormap
    cbar = fig.colorbar(s2) # Show colorbar
    cbar_ticks = np.linspace(-0.2,0.7,5) 
    cbar.set_ticks(cbar_ticks) # adjust colorbar ticks
    
    ax2.set_xlabel('r', fontsize = 20)
    ax2.set_ylabel('theta', fontsize = 20)
    ax2.set_ylim([0,2*np.pi])
    ax2.set_title('Bessel n = 1, k = 1', fontsize = 20)
    
    
    # SUBFIGURE 3
    # Create subfigure 3 2nd row, last col
    z = drumhead_height(1,2,rm,thetam,t)
    ax3 = fig.add_subplot(gs[1, -1])
    s3 = ax3.pcolor(rm,thetam,z, cmap = colormap, vmin=-0.3, vmax=.5) # vmin and vmax are min and max values of colormap
    cbar = fig.colorbar(s3) # Show colorbar
    cbar_ticks = np.linspace(-0.2,0.7,5) 
    cbar.set_ticks(cbar_ticks) # adjust colorbar ticks
    
    ax3.set_xlabel('r', fontsize = 20)
    ax3.set_ylabel('theta', fontsize = 20)
    ax3.set_ylim([0,2*np.pi])
    ax3.set_title('Bessel n = 1, k = 2', fontsize = 20)
    
    
    # SUBFIGURE 4
    # Create subfigure 4 3rd row, last col
    z = drumhead_height(5,5,rm,thetam,t)
    ax4 = fig.add_subplot(gs[2, -1])
    s4 = ax4.pcolor(rm,thetam,z, cmap = colormap, vmin=-0.3, vmax=.5) # vmin and vmax are min and max values of colormap
    cbar = fig.colorbar(s4) # Show colorbar
    cbar_ticks = np.linspace(-0.2,0.7,5) 
    cbar.set_ticks(cbar_ticks) # adjust colorbar ticks
    
    ax4.set_xlabel('r', fontsize = 20)
    ax4.set_ylabel('theta', fontsize = 20)
    ax4.set_ylim([0,2*np.pi])
    ax4.set_title('Bessel n = 5, k = 5', fontsize = 20)

    return
```

Define function for plotting 3d Bessel plots.

```python
def plot_Bessel3d(t = 0):
    #Create two arrays
    
    #Length of an array dimension
    length = 100
    
    #r [0,1], theta [0,2pi]
    rArr = np.linspace(0,1,length)
    thetaArr = np.linspace(0,2*np.pi,length)
    
    #Use meshgrid to create polar coords grid
    rm, thetam = meshgrid(rArr, thetaArr)
    
    #Create new x,y mesh-sized arrays
    xArr = np.empty_like(rm)
    yArr = np.empty_like(thetam)
    
    #Convert to Cartesian
    xArr = np.multiply(rm,np.cos(thetam))
    yArr = np.multiply(rm,np.sin(thetam))
      
    #Transform it into a cartesian grid
    
    # Create a plotting grid with gridspec
    # Create a figure
    fig = plt.figure(figsize=(16, 10))

    #Create subfigures
    
    # SUBFIGURE 1
    ax1 = fig.gca(projection='3d')
    
    #Compute z with given parameters
    z = drumhead_height(5,5,rm,thetam,t)
    
    #Subplot 1 Plotting
    # Plot using pcolor and arrange colorbar
    colormap = 'RdBu'
    specs = {'cmap' : 'seismic', 'vmin' : -0.6, 'vmax' : 1.0}
    
    # Plot surface and contours
    ax1.plot_surface(xArr, yArr, z, rstride=8, cstride=2, alpha=0.3,**specs)
    cset = ax1.contourf(xArr, yArr, z, zdir='z', offset=-3, **specs)
    cset = ax1.contour(xArr, yArr, z, zdir='x', offset=0, cmap=cm.coolwarm)
    cset = ax1.contour(xArr, yArr, z, zdir='y', offset=0, cmap=cm.coolwarm)
    
    ax1.set_title('Bessel n = 5, k = 5', fontsize = 20)
    ax1.set_zlim(-2.01, 2.01)
    ax1.set_xlim(-1,1)
    ax1.set_ylim(-1,1)
    ax1.set_xlabel('x', fontsize = 20)
    ax1.set_ylabel('y', fontsize = 20)
    
    return
```

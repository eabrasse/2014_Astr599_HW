Declare initial mandel function with no speed-ups.

```python
def mandel(x, y, max_iter):
    """
    Given z = x + iy and max_iter, determine whether the candidate
    is in the mandelbrot set for the given number of iterations
    """
    c = complex(x, y)
    z = 0.0j
    for i in range(max_iter):
        z = z*z + c
        if (z.real*z.real + z.imag*z.imag) >= 4:
            return i

    return max_iter

def create_fractal(Nx, Ny, xmin, xmax, ymin, ymax, max_iter):
    """Create and return a fractal image"""
    image = np.zeros((Ny, Nx), dtype=float)
    dx = (xmax - xmin) * 1. / Nx
    dy = (ymax - ymin) * 1. / Ny
    
    for x in range(Nx):
        rpart = xmin + x * dx
        for y in range(Ny):
            ipart = ymin + y * dy
            color = mandel(rpart, ipart, max_iter)
            image[y, x] = color
    return image
```

```python
image = create_fractal(300, 200, -2, 1, -1, 1, 20)
plt.imshow(image, cmap=plt.cm.jet)
```

3) Speed up with numba now (function will be mandel3).  Note, change array to float64 data type.

```python
#Includes
from numba import autojit
```

```python
@autojit
def mandel3(x, y, max_iter):
    """
    Given z = x + iy and max_iter, determine whether the candidate
    is in the mandelbrot set for the given number of iterations
    """
    c = complex(x, y)
    z = 0.0j
    for i in range(max_iter):
        z = z*z + c
        if (z.real*z.real + z.imag*z.imag) >= 4:
            return i

    return max_iter

@autojit
def create_fractal3(Nx, Ny, xmin, xmax, ymin, ymax, max_iter):
    """Create and return a fractal image"""
    image = np.zeros((Ny, Nx), dtype=np.float64)
    dx = (xmax - xmin) * 1. / Nx
    dy = (ymax - ymin) * 1. / Ny
    
    for x in range(Nx):
        rpart = xmin + x * dx
        for y in range(Ny):
            ipart = ymin + y * dy
            color = mandel3(rpart, ipart, max_iter)
            image[y, x] = color
    return image
```

```python
#Proof it works
image = create_fractal3(300, 200, -2, 1, -1, 1, 20)
plt.imshow(image, cmap=plt.cm.jet)
```

#Numba
Numba @autojit functions (create_fractal3 and mandel3) are significantly faster than the original non-speed up python code (mandel).  Roughly, it's over 100 times faster!

#Cython
Simply including cython magic (create_fractal2, mandel2) speed up the function (relative to vanilla python) by about a factor of 1.5.  After defining the types of all variables (especially those used in the numpy array for indexing/writing) and expanding the mandel function into more c-like syntax, the cython code ran two orders of magnitude faster than the vanilla python code and about 0.5 ms faster than the numba code.  Take that, numba!

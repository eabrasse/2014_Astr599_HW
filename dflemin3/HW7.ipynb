Import data sets, set up test/training data

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from astroML.datasets import fetch_rrlyrae_combined
from sklearn.cross_validation import train_test_split


X, y = fetch_rrlyrae_combined()

# For now, we'll only fit the first two colors
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

Plot Data

```python
N_plot = 5000
plt.scatter(X[-N_plot:, 0], X[-N_plot:, 1], c=y[-N_plot:],
            edgecolors='none')
plt.xlabel('u-g color')
plt.ylabel('g-r color');
```

K-Neighbors classification, train on part of data, test on rest.  Print results.

```python
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier(n_neighbors=5)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

print(np.sum(y_pred == y_test))
print(np.sum(y_pred != y_test))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred,
                            target_names=['MS star', 'RR Lyrae']))
```

Notes:  Use same training/testing data as above for simplicity.

Gaussian Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(X_train, y_train)
y_pred = gnb.predict(X_test)

print(np.sum(y_pred == y_test))
print(np.sum(y_pred != y_test))

print(classification_report(y_test, y_pred,
                            target_names=['MS star', 'RR Lyrae']))
```

Very fast, but quite inaccurate given the very low precision and f1-scores for RR Lyraes.

LDA Method

```python
from sklearn.lda import LDA
clf = LDA()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print(np.sum(y_pred == y_test))
print(np.sum(y_pred != y_test))

print(classification_report(y_test, y_pred,
                            target_names=['MS star', 'RR Lyrae']))
```

As expected, fast but pretty inaccurate given very low precision and f1-scores for RR Lyrae.

SVC Method

```python
from sklearn.svm import SVC
svc = SVC(kernel='poly', degree = 5)

svc.fit(X_train, y_train)
y_pred = svc.predict(X_test)
print(np.sum(y_pred == y_test))
print(np.sum(y_pred != y_test))

print(classification_report(y_test, y_pred,
                            target_names=['MS star', 'RR Lyrae']))
```

sklearn.svm.SVC with kernel='rbf' Method (Default)

```python
svck = SVC(kernel='rbf')
svck.fit(X_train, y_train)
y_pred = svck.predict(X_test)
print(np.sum(y_pred == y_test))
print(np.sum(y_pred != y_test))

print(classification_report(y_test, y_pred,
                            target_names=['MS star', 'RR Lyrae']))
```

sklearn.ensemble.RandomForestClassifier Method

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=15,criterion='entropy',min_samples_split=3)

rfc.fit(X_train, y_train)
y_pred = rfc.predict(X_test)
print(np.sum(y_pred == y_test))
print(np.sum(y_pred != y_test))

print(classification_report(y_test, y_pred,
                            target_names=['MS star', 'RR Lyrae']))
```

RandomForestClassifier seems to be the best and improves in every catergory but recall.

#Summary
The random forest classifiers seems to give me the best combination of precision, recall and f1-score when compared to the other methods after tuning its parameters.  Although not perfect, the random forest method still does a rather good job at classifying RR Lyrae stars.  The SVC method seemed to have nan/degeneracy errors for me so I wasn't able to glean any useful data from it.  Second best was the Naive-Bayes method, but it still paled in comparison to the random forest method's predictive power.

#2) Regression

```python
#Copy/Paste of Jake's code
from astroML.datasets import fetch_sdss_specgals
```

```python
data = fetch_sdss_specgals()

# put magnitudes in a matrix
X = np.vstack([data['modelMag_%s' % f] for f in 'ugriz']).T
y = data['z']

# Split into training and testing data
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
from sklearn.linear_model import LinearRegression
est = LinearRegression()
est.fit(X_train, y_train)
y_pred = est.predict(X_test)
```

```python
plt.plot(y_test, y_pred, ',k')
plt.plot([0, 1], [0, 1], ':k')
plt.xlim(0, 0.6)
plt.ylim(0, 0.6)
```

```python
rms = np.sqrt(np.mean((y_test - y_pred) ** 2))
print(rms)
```

Try the following methods in order to better fit the data:

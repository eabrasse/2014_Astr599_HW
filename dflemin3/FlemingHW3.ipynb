```python
#Imports
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize
import astroML.datasets
```

Import data, select appropriate data set, plot raw lightcurve

```python
lcs = astroML.datasets.fetch_LINEAR_sample()
time, flux, dflux = lcs[lcs.ids[10]].T
plt.errorbar(time, flux, yerr=dflux, fmt="ro")
plt.gca().invert_yaxis()
plt.xlabel("Mean Julian Day")
plt.ylabel("Mag")
plt.show()
```

Use Lombe-Scargle to produce periodogram and display it.

```python
from astroML.time_series import lomb_scargle
periods = np.logspace(-1, 0, 10000, base=10)
periodogram = lomb_scargle(time, flux, dflux, omega=2 * np.pi / periods, generalized=True)
plt.plot(periods, periodogram)
plt.semilogx()
plt.xlabel("Period (days)")
plt.ylabel("Power")
plt.show()
```

Find the peak of the periodogram.

```python
i = np.argmax(periodogram)

print periods[i], periodogram[i]
```

Fold lightcurve given the found period, print min/max phases.  Plot resultant scaled lightcurve.

```python
period = periods[i]
phase = time / period - time // period 
print min(phase), max(phase)
plt.errorbar(phase, flux, yerr=dflux, fmt="ro")
plt.gca().invert_yaxis()
plt.xlabel("Phase")
plt.ylabel("Mag")
plt.show()
```

Class implimentation from 14_Optimization class ipython notebook.  

```python
class Fourier(object):
    def __init__(self, phase, flux, dflux, nterms):
        self.phase = phase
        self.flux = flux
        self.dflux = dflux
        self.nterms = nterms
        assert(self.nterms > 0)
        
    def evaluate(self, phase, terms):
        assert(len(terms) == 2 * self.nterms - 1)
        model = terms[0] * np.ones(len(phase))
        for i in range(self.nterms-1):
            model += terms[2*i + 1] * np.cos(2 * np.pi * (i+1) * phase + terms[2*i + 2])
        return model
    
    def chi2(self, args):
        model = self.evaluate(self.phase, args)
        chi   = (model - self.flux) / self.dflux
        return np.sum(chi**2)
```

Begin attempting to fit actual data using optimize's fmin.

```python
fourier = Fourier(phase, flux, dflux, 2)
guess   = []  #Mean brightness, the amplitude, and minimum used for guess fitting parameters
guess.append(np.mean(fourier.flux))
guess.append(np.std(fourier.flux))
guess.append(2 * np.pi * (1 - fourier.phase[np.argmax(fourier.flux)]))
#optimize
result = optimize.fmin(fourier.chi2, x0=[guess], disp=0, full_output=1)
best_fit = result[0]
```

Plot basic initial result.

```python
mphase = np.arange(0, 1, 0.01)
plt.errorbar(fourier.phase, fourier.flux, yerr=fourier.dflux, fmt="ro")
plt.plot(mphase, fourier.evaluate(mphase, guess), "b--", label="Initial guess")
plt.plot(mphase, fourier.evaluate(mphase, best_fit), "g-", label="Best fit")
plt.gca().invert_yaxis()
plt.xlabel("Phase")
plt.ylabel("Mag")
plt.legend()
plt.show()
```

Interestingly enough, the 1st order best fit looks pretty good.

Now add more terms, see what happens.

```python
fourier2 = Fourier(phase, flux, dflux, 3)
guess2   = [] 
guess2.append(np.mean(fourier2.flux))
guess2.append(np.std(fourier2.flux))
guess2.append(2 * np.pi * (1 - fourier2.phase[np.argsort(fourier2.flux)[-1]]))
guess2.append(0.0)
guess2.append(2 * np.pi * (1 - fourier2.phase[np.argsort(fourier2.flux)[-1]]))
result2 = optimize.fmin_bfgs(fourier2.chi2, x0=[guess2], disp=0, full_output=1)
best_fit2 = result2[0]
#Now plot
plt.errorbar(fourier2.phase, fourier2.flux, yerr=fourier2.dflux, fmt="ro")
plt.plot(mphase, fourier2.evaluate(mphase, guess2), "b--", label="Initial guess")
plt.plot(mphase, fourier2.evaluate(mphase, best_fit2), "g-", label="Best fit")
plt.gca().invert_yaxis()
plt.xlabel("Phase")
plt.ylabel("Mag")
plt.show()
```

As expected, more terms yield better fit, but how many terms is enough?

Below is the inherited class implimentation to make guesses given in the class ipython notebook.

```python
class Fourier2(Fourier):
    def __init__(self, phase, flux, dflux, nterms):
        Fourier.__init__(self, phase, flux, dflux, nterms)
                
    def makeGuess(self):
        guess = []
        guess.append(np.mean(self.flux))
        if self.nterms > 1:
            guess.append(np.std(self.flux))
            guess.append(2 * np.pi * (1 - self.phase[np.argsort(self.flux)[-1]]))
            for i in range(2, self.nterms):
                guess.append(0.0)
                guess.append(2 * np.pi * (1 - self.phase[np.argsort(self.flux)[-1]]))
        return guess
```

Check chi^2 evolution and compare it to delta chi.

```python
nterms = np.arange(1, 10)
chi2   = []
for nterm in nterms:
    fourier = Fourier2(phase, flux, dflux, nterm)
    guess = fourier.makeGuess()
    #print(guess)
    chi2.append(optimize.fmin_bfgs(fourier.chi2, x0=[guess], disp=0, full_output=1)[1])
chi2 = np.array(chi2)
plt.plot(nterms, chi2)
plt.xlabel("N Terms")
plt.ylabel("Chi2")
plt.show()
```

Looks like 4 terms is sufficient to model the data as change in chi^2 pretty much stops after 4 maybe 5 terms.  Now need to quantify this.  Look at Delta_chi2 then BIC/AIC.

```python
delta_chi2 = chi2[:-1] - chi2[1:]
plt.plot(nterms[1:], delta_chi2)
plt.xlabel("N terms")
plt.ylabel("Delta chi2")
plt.show()

#Find maximum
print delta_chi2
```

Delta chi2 refers to the difference in chi2 between two nested models as a test statistics that follows its own chi2 distribution with degrees of freedom being the difference in parameter number between the two models, or in this case, 2 for the ith A and ith phase.

Therefore, the delta chi2 values will follow a chi2 distribution with degrees of freedom df = 2 since the difference in total model parameters between the Nth order model and N+1th order model is 2 (A, phase).

If the delta chi2 between 2 models is significant, the higher order model fits well enough to justify the additional parameters, otherwise, go with the smaller ordered model (source: http://www.psychologie.uzh.ch/fachrichtungen/methoden/team/christinawerner/sem/chisquare_diff_en.pdf)

To find the correct model order, one needs to find the point in the delta_chi2 array where the delta chi2 returns a significance of >0.05.  This corresponds to the chi2 value above which the area under the distribution is <=0.05 so any value below this point will yiled >0.05.  For a df = 2 chi2 distribution, a chi2 value of 5.99 corresponds to a significance of 0.05.

By inspection, we can see that delta_chi2 is less than 5.99 for the 3rd (0,1,2,3=3rd) index with a value of ~3.69 indicating that it does not reach the significance threshold of 0.05.  From this, we can see that the 2nd index corresponding from a model change from N terms = 3->4 is the last important significant delta chi2 so the Nterms index = 4.  This is used below to properly fit the data.

```python
nterms     = np.arange(1, 10)
chi2       = []
BICpenalty = []
AICpenalty = []
for nterm in nterms:
    fourier = Fourier2(phase, flux, dflux, nterm)
    guess   = fourier.makeGuess()
    BICpenalty.append(len(guess) * np.log(len(fourier.phase)))
    AICpenalty.append(len(guess) * 2.0)
    chi2.append(optimize.fmin_bfgs(fourier.chi2, x0=[guess], disp=0, full_output=1)[1])
chi2 = np.array(chi2)
BICpenalty = np.array(BICpenalty)
AICpenalty = np.array(AICpenalty)
plt.plot(nterms, chi2, label="chi2")
plt.plot(nterms, chi2 + BICpenalty, label="BIC")
plt.plot(nterms, chi2 + AICpenalty, label="AIC")
plt.xlabel("N Terms")
plt.ylabel("Chi2")
plt.semilogy()
plt.legend()
plt.show()
```

Can see that BIC/AIC/Chi^2 terms flatten out around 3, 4 indicating that the benefit of adding additional terms to the model becomes marginal.  Output fits that minimize BIC/AIC and the corresponding model order.

```python
print np.argmin(chi2+BICpenalty), np.argmin(chi2+AICpenalty)
print nterms[3], nterms[6]
```

BIC indeicates that a 4th order model provides enough terms to model the data without becoming to variable while AIC penalties indicate a 7th order model suffices.  Now print the final results and produce the best fit graphs. BIC 4th order results first 

```python
fourier5 = Fourier(phase, flux, dflux, 4)
guess5   = [] 
guess5.append(np.mean(fourier5.flux))
guess5.append(np.std(fourier5.flux))
guess5.append(2 * np.pi * (1 - fourier5.phase[np.argsort(fourier5.flux)[-1]]))
for i in range(2, 4):
    guess5.append(0.0)
    guess5.append(2 * np.pi * (1 - fourier5.phase[np.argsort(fourier5.flux)[-1]]))
result5 = optimize.fmin_bfgs(fourier5.chi2, x0=[guess5], disp=0, full_output=1)
best_fit5 = result5[0]
#plot
plt.errorbar(fourier5.phase, fourier5.flux, yerr=fourier5.dflux, fmt="ro")
plt.plot(mphase, fourier5.evaluate(mphase, guess5), "b--", label="Initial guess")
plt.plot(mphase, fourier5.evaluate(mphase, best_fit5), "g-", label="Best fit")
plt.gca().invert_yaxis()
plt.xlabel("Phase")
plt.ylabel("Mag")
plt.show()
```

AIC influenced model with 7 terms.

```python
fourierTest = Fourier2(phase, flux, dflux, 7)
guess = fourierTest.makeGuess()

fourier7 = Fourier(phase, flux, dflux, 7)
result = optimize.fmin_bfgs(fourier7.chi2, x0=[guess], disp=0, full_output=1)
best_fit = result[0]
#plot
plt.errorbar(fourier7.phase, fourier7.flux, yerr=fourier7.dflux, fmt="ro")
plt.plot(mphase, fourier7.evaluate(mphase, guess), "b--", label="Initial guess")
plt.plot(mphase, fourier7.evaluate(mphase, best_fit), "g-", label="Best fit")
plt.gca().invert_yaxis()
plt.xlabel("Phase")
plt.ylabel("Mag")
plt.show()
```

Fit looks good, as expected.

In summary, the BIC/AIC and Delta_chi2 analyses predicted models of order 4, 7 and 5 respectively.  The BIC and Delta_chi2 analysis seem to agree that 4,5 terms sufficiently model the data while AIC suggests 7 terms (and even more parameters) are neccessary.  Since 2/3 of the analysis suggests having 4/5 terms, it is likely best to go this route when modeling this particular light curve as any additional terms that are not neccessary adds additional parameters that could potentially skew the analysis.

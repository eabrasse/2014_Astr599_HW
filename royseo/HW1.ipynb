```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_hw1 = np.genfromtxt('../data/asteroids5000.csv',delimiter =',') 
nan_row_count=0
new_array=[]
# iterate through array by row
for row in data_hw1:
    # if the row has missing value, increment nan_row_count (1.2)
    if (np.isnan(row).sum())!=0:
        nan_row_count+=1
    # else, add to new array (1.3)
    else:
        new_array.append(row)
        
print("The number of nan is " + str(np.isnan(data_hw1).sum()))
print("The number of complete row is " + str((len(data_hw1)-nan_row_count)))

new_array = np.array(new_array)

#    maximum, minimum, mean, and standard deviation (1.4)
for i in range(0,new_array.shape[1]):
    clm = new_array[:,i]
    print '%d column avg: %f, std: %f, max:%f, min:%f'%(i+1,clm.mean(),clm.std(),clm.max(),clm.min())
```

```python
import math
# Question2
#0,1,2,3,4,5,6
#a,e,i,H,diameter,BV,UB
# semi-major axis versus the sine of the inclination 
a=new_array[:,0]
i=new_array[:,2]
rad_i = map(lambda x: math.sin(math.radians(x)),i)
c = plt.scatter(a,rad_i)
plt.ylabel('sin(i)')
plt.xlabel('semi-axis')
plt.show()
# a color-magnitude diagram of the asteroids (H vs B-V)
H=new_array[:,3]
BV=new_array[:,5]
c = plt.scatter(H,BV)
plt.plot([2,14],[0.8,0.8])
plt.ylabel('BV')
plt.xlabel('H')
plt.show()

```

```python
# classify asteroids into two families by H values
familyA = np.array(filter(lambda row: row[5]>0.8,new_array)) # asteroids with larger than 0.8 BV values
familyB = np.array(filter(lambda row: row[5]<=0.8,new_array))# asteroids with smaller than or equal to 0.8 BV values
print 'family a size: ',len(familyA), '\nfamily b size: ',len(familyB)
# semi-major axis versus the sine of the inclination 
# draw plots for family A
a_A=familyA[:,0]
i_A=familyA[:,2]
rad_i_A=map(lambda x: math.sin(math.radians(x)),familyA[:,2])
c = plt.scatter(a_A,rad_i_A,c='blue')
#draw plots for family B
a_B=familyB[:,0]
i_B=familyB[:,2]

rad_i_B=map(lambda x: math.sin(math.radians(x)),familyB[:,2])
c = plt.scatter(a_B,rad_i_B,c='red')
plt.ylabel('sin(i)')
plt.xlabel('semi-axis')
plt.show()
H=familyA[:,3]
BV=familyA[:,5]
c = plt.scatter(H,BV,c='blue')
H=familyB[:,3]
BV=familyB[:,5]
c = plt.scatter(H,BV,c='red')

```

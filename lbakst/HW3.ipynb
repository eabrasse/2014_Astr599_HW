```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize
import astroML.datasets
from astroML.time_series import lomb_scargle
lcs = astroML.datasets.fetch_LINEAR_sample()
```

```python
#Import data for a star
time, flux, dflux = lcs[50402].T
```

```python
#Plot raw lightcurve
plt.errorbar(time, flux, yerr=dflux, fmt="ro")
plt.gca().invert_yaxis()
plt.xlabel("Mean Julian Day")
plt.ylabel("Mag")
plt.show()
```

The star looks periodic so we want to plot a periodogram.

```python
#plot periodogram
periods = np.logspace(-1, 0, 10000, base=10)
periodogram = lomb_scargle(time, flux, dflux, omega=2 * np.pi / periods, generalized=True)

plt.plot(periods, periodogram)
plt.semilogx()
plt.xlabel("Period (days)")
plt.ylabel("Power")
plt.show()
```

```python
#find max in periodogram & "fold" data at this best period
idx = np.argmax(periodogram)
period = periods[idx]
phase = time / period - time // period 
```

```python
#plot this resultant overlapping data
plt.errorbar(phase, flux, yerr=dflux, fmt="ro")
plt.gca().invert_yaxis()
plt.xlabel("Phase")
plt.ylabel("Mag")
plt.show()
```

We want to assess the model fit, so we create a class that will include the data as member variables and return the chi square fit.

```python
class Fourier(object):
    def __init__(self, phase, flux, dflux, nterms):
        self.phase = phase
        self.flux = flux
        self.dflux = dflux
        self.nterms = nterms
        assert(self.nterms > 0)
        
    def evaluate(self, phase, terms):
        assert(len(terms) == 2 * self.nterms - 1)
        model = terms[0] * np.ones(len(phase))
        for i in range(self.nterms-1):
            model += terms[2*i + 1] * np.cos(2 * np.pi * (i+1) * phase + terms[2*i + 2])
        return model
    
    def chi2(self, args):
        model = self.evaluate(self.phase, args)
        chi   = (model - self.flux) / self.dflux
        return np.sum(chi**2)
```

Create child class of Fourier that initializes the guess for us.

```python
class Fourier2(Fourier):
    def __init__(self, phase, flux, dflux, nterms):
        Fourier.__init__(self, phase, flux, dflux, nterms)
                
    def makeGuess(self):
        guess = []
        guess.append(np.mean(self.flux))
        if self.nterms > 1:
            guess.append(np.std(self.flux))
            guess.append(2 * np.pi * (1 - self.phase[np.argsort(self.flux)[-1]]))
            for i in range(2, self.nterms):
                guess.append(0.0)
                guess.append(2 * np.pi * (1 - self.phase[np.argsort(self.flux)[-1]]))
        return guess
```

```python
#Plot the chi2 for models with a range terms, include pentalties as imposed by AIC/BIC
nterms     = np.arange(1, 30, 2)
chi2       = []
BICpenalty = []
AICpenalty = []
for nterm in nterms:
    fourier = Fourier2(phase, flux, dflux, nterm)
    guess   = fourier.makeGuess()
    BICpenalty.append(len(guess) * np.log(len(fourier.phase)))
    AICpenalty.append(len(guess) * 2.0)
    chi2.append(optimize.fmin_bfgs(fourier.chi2, x0=[guess], disp=0, full_output=1)[1])
chi2 = np.array(chi2)
BICpenalty = np.array(BICpenalty)
AICpenalty = np.array(AICpenalty)
```

```python
#Plot the results
plt.plot(nterms, chi2, label="chi2")
plt.plot(nterms, chi2 + BICpenalty, label="BIC")
plt.plot(nterms, chi2 + AICpenalty, label="AIC")
plt.xlabel("N Terms")
plt.ylabel("Chi2")
plt.semilogy()
plt.legend()
plt.show()
```

The model fit (as measured by chi2) improves with the addition of each term. However if we look at the change in chi2 with each additional term (delta chi2, plotted below), it becomes clear that the improvement in fit imparted to the model with each term decreases.

```python
delta_chi2 = chi2[:-1] - chi2[1:]
plt.plot(nterms[1:], delta_chi2)
plt.xlabel("N terms")
plt.ylabel("Delta chi2")
plt.show()
```

As you add terms to the model, delta chi2 decreases. This means that each term is no longer improving the model fit as much as the previous. I decided to incorporate this as a penalty by subracting the delta chi2 from the chi2 and looking for the minimum. This way, a term that has imparted a large benefit in model fit will be 'weighted' as such.

```python
#find fits with the minimul AIC & BIC
print np.argmin(chi2+BICpenalty), np.argmin(chi2+AICpenalty), np.argmin(chi2[1:]-delta_chi2)
```

The minimum of the chi2 penalized by the BIC (BIC added to the chi2) is at index 3. The minimum of chi2 penalized by AIC is at index 4, and the minimum of chi2 while taking into account change in chi2 (subtracting the change in chi2) gives us index 1.

```python
print nterms[3], nterms[4], nterms[1]
```

This means that using the BIC penalty we would want to use a model with 7 terms. Using AIC we would want 9 terms, and delta chi2 would suggest a model with 3 terms.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
from astroML.datasets import fetch_rrlyrae_combined
from sklearn.cross_validation import train_test_split

X, y = fetch_rrlyrae_combined()

X_train, X_test, y_train, y_test = train_test_split(X,y)
```

```python
N_plot = 5000
plt.scatter(X[-N_plot:, 0], X[-N_plot:, 1], c=y[-N_plot:], edgecolors='none')
plt.xlabel('u-g color')
plt.ylabel('g-r color');
```

```python
from sklearn.neighbors import KNeighborsClassifier

clf = KNeighborsClassifier(n_neighbors=5)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
```

```python
print(np.sum(y_pred == y_test))
print(np.sum(y_pred != y_test))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred, target_names=['MS star', 'RR Lyrae']))
```

Exercise 1: Use another classifier to try to improve the precision/recall score

```python
from sklearn.ensemble import RandomForestClassifier
clf2 = RandomForestClassifier()
clf2.fit(X_train, y_train)
y_pred2 = clf2.predict(X_test)
```

```python
print classification_report(y_test, y_pred2, target_names=['MS star','RR Lyrae'])
```

```python
from sklearn.lda import LDA
clf3 = LDA()
clf3.fit(X_train, y_train)
y_pred3 = clf3.predict(X_test)
```

```python
print classification_report(y_test, y_pred3, target_names=['MS star','RR Lyrae'])
```

Compared to KNeighbors classifier, the LDA classifier was worse on nearly every score. The Random Forest classifier performed well, though. For the main sequence stars both classifiers performed equally. The Random Forest showed a clear increase in precision for RR-Lyrae, a slight decrease in recall and a tick up of 0.01 in f1.

```python
from astroML.datasets import fetch_sdss_specgals
```

```python
data =  fetch_sdss_specgals()

X = np.vstack([data['modelMag_%s' % f] for f in 'ugriz']).T
y = data['z']

X_train, X_test, y_train, y_test = train_test_split(X,y)
```

```python
from sklearn.linear_model import LinearRegression
est = LinearRegression()
est.fit(X_train,y_train)
y_pred = est.predict(X_test)
```

```python
plt.plot(y_test,y_pred,',k')
plt.plot([0, 1], [0, 1], ':k')
plt.xlim(0, 0.6)
plt.ylim(0, 0.6)
```

```python
rms = np.sqrt(np.mean((y_test - y_pred) ** 2))
print rms
```

Exercise 2: Improve RMS using another algorithm

```python
from sklearn.ensemble import RandomForestRegressor
est2 = RandomForestRegressor()
est2.fit(X_train,y_train)
y_pred2 = est2.predict(X_test)
```

```python
rms = np.sqrt(np.mean((y_test - y_pred2) ** 2))
print rms
```

This Random Forest regressor improved on the RMS, decreasing it to ~0.0231 from ~0.0324 using the linear regression.

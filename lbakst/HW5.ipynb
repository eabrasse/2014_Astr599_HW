```python
#Q1. Drumhead Height
#Import
from mpl_toolkits.mplot3d import Axes3D
from scipy import *
from scipy.special import jn, jn_zeros
%pylab inline
from pylab import *

def drumhead_height(n,k,r,theta,t):
    nth_zero = jn_zeros(n, k)
    #print jn(n,r*nth_zero)
    return np.cos(t)*np.cos(n*theta)*jn(n, r*nth_zero[k-1])
```

```python
#Q2. Plot Bessel

def plot_Bessel(t):
    #create arrays
    r_lin = linspace(0,1,100)
    theta_lin = linspace(0,2*np.pi,100)
    
    #polar coordinates grid
    rm, thetam = meshgrid(r_lin,theta_lin)
    
    #convert polar to cartesian coordinates
    x = r_lin*cos(theta_lin)
    y = r_lin*sin(theta_lin)   
    xm_cart, ym_cart = meshgrid(x,y)
    
    gs = plt.GridSpec(3, 3, wspace=0.2, hspace=0.2)
    
    fig = plt.figure(figsize=(16, 6))
    
    #subplot 1: 3x2
    ax1 = fig.add_subplot(gs[:3, :2],projection='3d')
    z = drumhead_height(2, 3, rm, thetam, t)
    specs = {'cmap' : 'seismic', 'vmin' : -0.6, 'vmax' : 1.0}
    surf = ax1.plot_surface(xm_cart, ym_cart, z, rstride = 1, cstride = 2, linewidth=0, **specs)
    fig.colorbar(surf, shrink=0.5, aspect=5)
    
    #subplot 2: 1x1
    ax2 = fig.add_subplot(gs[0,2],projection='3d')
    z = drumhead_height(3, 2, rm, thetam, t)
    specs = {'cmap' : 'seismic', 'vmin' : -0.6, 'vmax' : 1.0}
    surf = ax2.plot_surface(xm_cart, ym_cart, z, rstride = 1, cstride = 2, linewidth=0, **specs)
    
    #subplot 3: 1x1
    ax3 = fig.add_subplot(gs[1,2],projection='3d')
    z = drumhead_height(2, 4, rm, thetam, t)
    specs = {'cmap' : 'seismic', 'vmin' : -0.6, 'vmax' : 1.0}
    surf = ax3.plot_surface(xm_cart, ym_cart, z, rstride = 1, cstride = 2, linewidth=0, **specs)
    
    #subplot 4: 1x1
    ax4 = fig.add_subplot(gs[2,2],projection='3d')
    z = drumhead_height(2, 10, rm, thetam, t)
    specs = {'cmap' : 'seismic', 'vmin' : -0.6, 'vmax' : 1.0}
    surf = ax4.plot_surface(xm_cart, ym_cart, z, rstride = 1, cstride = 2, linewidth=0, **specs)
    
```

```python
#Q6.

from JSAnimation import IPython_display, HTMLWriter
from matplotlib import animation
import Image
import glob

n_frames = 5
for i in xrange(0,n_frames,1):
    plot_Bessel(i)
    plt.savefig('animation_frames/frame{0:07d}.png'.format(i+1));
    plt.close()

# Choose location of files to use
filenames=sorted(glob.glob('animation_frames/frame*.png'))

#Create figure
fig = plt.figure(figsize=(10,5))
ax = fig.add_axes([0, 0, 1, 1])
ax.axis('off')
im = plt.imshow(Image.open(filenames[0]))

# Create initializatio function
def init():
    im.set_data(Image.open(filenames[0]))
    return im,

# Create animation function
def animate(i):
    image=Image.open(filenames[i])
    im.set_data(image)
    return im,

anim = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=100, interval=20, blit=True)

#set embed_frames=True to embed base64-encoded frames directly in the HTML
anim.save('animation_frames/animation.html', writer=HTMLWriter(embed_frames=True))

# Show animation in ipython notebook
anim
```

```python
import numpy as np
```

```python
#Import data
asteroids = np.genfromtxt('data/asteroids5000.csv',delimiter=',')
```

```python
#number of missing values
n_missing = np.sum(np.isnan(asteroids))
print n_missing

#There are 11241 values missing from the data.
```

```python
#Number of rows without any missing values
avg_rows = np.mean(asteroids,1) #this will throw a nan in the index of whichever rows have missing data
n_fullrows = np.sum(~np.isnan(avg_rows))
print n_fullrows

#There are 767 full rows in the data set.

#indices of complete rows
rows_idx = np.where(~np.isnan(avg_rows))
```

```python
#Data with only complete rows
full_data = asteroids[rows_idx,:]
full_data = np.reshape(full_data,(767,7))
```

```python
#Descriptives for complete rows: mean
avgs = np.mean(full_data,0)
print avgs
```

```python
#maximum
maxs = np.amax(full_data,0)
print maxs
```

```python
#minimum
mins = np.amin(full_data,0)
print mins
```

```python
#standard deviation
stds = np.std(full_data,0)
print stds
```

```python
#import matplotlib & make figures appear inline
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
#Semi-major axis vs. Inclination angle
#Convert i to radians
i_rad = np.radians(full_data[:,2])
```

```python
#Figure 1. A vs. Sin(I)
plt.scatter(full_data[:,0],np.sin(i_rad))
plt.xlabel('Semi-Major Axis')
plt.ylabel('Sine of Inclination Angle')
plt.title('Semi-Major Axis vs. Sine of the Inclination Angle')
```

```python
#Color-magnitude diagram: BV vs. H
plt.scatter(full_data[:,5],full_data[:,3])
plt.xlabel('Color (B-V)')
plt.ylabel('Absolute Magnitude (H)')
plt.title('Color-Magnitude Diagram')

#Dividing line: I'm estimating that the mean of the B-V variable (+ a little bit) will separate the two
x = [0, 0]
y = [0, 0]
x[0] = avgs[5]+.02
x[1] = avgs[5]+.02
y[0] = maxs[3]
y[1] = mins[3]
plt.plot(x,y)
```

```python
#Plot semi-major axis and inclination angle again but with the two groups from above in different colors
#threshold
thresh = x[0]
idx = full_data[:,5]>=thresh

#plot
group1 = plt.scatter(full_data[~idx,0],np.sin(i_rad[~idx]),color='r')
group2 = plt.scatter(full_data[idx,0],np.sin(i_rad[idx]))
plt.legend((group1, group2),
           ('Left Cluster','Right Cluster'))
plt.xlabel('Semi-Major Axis')
plt.ylabel('Sine Inclination Angle')
plt.title('Semi-Major Axis vs. Sine of Inclination Angle for Two Groups of Asteroids')
```

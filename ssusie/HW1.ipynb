Load  the text file with missing data:

```python
import numpy as np
text=np.genfromtxt('asteroids5000.csv',  delimiter=",",filling_values=np.nan)
(np.isnan(text)).sum() #number of elements that are nan-s
```

The last line indicates how we can find  how many times the disired value appears in the data. So there are 11241 missing elements.

```python
ind=np.where(np.isnan(text))
```

This command gives the indices of elements that are nan. It returns two arrays. 
First of them indicates the index of the row, the second array indicates the column.

```python
ind
```

Now, delete all rows that contain nan. Clearly, these rows are indicated by variable ind. 

```python
text_new=x = np.delete(text, (ind[0]), axis=0)
```

```python
print(text.shape)
print(text_new.shape)
```

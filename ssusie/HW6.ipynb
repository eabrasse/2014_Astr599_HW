```python
%load_ext cythonmagic
%matplotlib inline
import matplotlib.pyplot as plt
from timeit import default_timer as timer
```

```python
%%cython -a
cimport cython
cimport numpy as np
import numpy as np


cdef int mandel_c(float x, float y, int max_iter):
    """
    Given z = x + iy and max_iter, determine whether the candidate
    is in the mandelbrot set for the given number of iterations
    """

    cdef int i
    cdef complex c,z
    c = complex(x, y)
    z = complex(0,0)
    cdef float zr, zi,ab 
    
    for i in range(max_iter):
        z = z*z + c
        zr=z.real
        zi=z.imag
        ab=zr*zr + zi*zi
        if (ab)>= 4:
            return i

    return max_iter

@cython.boundscheck(False) 
def create_fractal(int Nx,int Ny,float xmin,float xmax,float ymin,float ymax,int max_iter):
    """Create and return a fractal image"""
    cdef  np.ndarray[np.float64_t, ndim=2] image
    image = np.zeros((Ny, Nx), dtype=float)
    cdef float dx = (xmax - xmin) * 1. / Nx
    cdef float dy = (ymax - ymin) * 1. / Ny
    
    cdef int x, y
    cdef float rpart, ipart
    
    for x in range(Nx):
        rpart = xmin + x * dx
        for y in range(Ny):
            ipart = ymin + y * dy
            image[y, x] = mandel_c(rpart, ipart, max_iter)
          
    return image
```

```python
start = timer()
image = create_fractal(300, 200, -2.0, 1.0, -1.0, 1.0, 20)
dt=dt = timer() - start
```

```python
from numba import autojit
```

```python
@autojit
def mandel(x, y, max_iter):
    """
    Given z = x + iy and max_iter, determine whether the candidate
    is in the mandelbrot set for the given number of iterations
    """
    c = complex(x, y)
    z = 0.0j
    for i in range(max_iter):
        z = z*z + c
        if (z.real*z.real + z.imag*z.imag) >= 4:
            return i

    return max_iter

@autojit
def create_fractal(Nx, Ny, xmin, xmax, ymin, ymax, max_iter):
    """Create and return a fractal image"""
    image = np.zeros((Ny, Nx), dtype=np.float64)
    dx = (xmax - xmin) * 1. / Nx
    dy = (ymax - ymin) * 1. / Ny
    
    for x in range(Nx):
        rpart = xmin + x * dx
        for y in range(Ny):
            ipart = ymin + y * dy
            color = mandel(rpart, ipart, max_iter)
            image[y, x] = color
    return image


```

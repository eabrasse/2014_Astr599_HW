```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
from scipy import special
from scipy.special import jn, jn_zeros
```

```python
def drumhead_height(n,k,r,theta,t):
    L_nk=jn_zeros(n,k)[k-1]
    L=np.cos(t)*np.cos(n*theta)*jn(n,L_nk*r)
    
    return L
    
```

```python
def plot_Bessel(t):
    r=np.linspace(0,1,100)
    theta=np.linspace(0, 2*np.pi, 100)
    rm, thetam=np.meshgrid(r,theta)
    xm=rm*np.cos(thetam)
    ym=rm*np.sin(thetam)
    
    
    gs = plt.GridSpec(3, 3, wspace=0.2, hspace=0.2) 
    fig = plt.figure(figsize=(16, 6))
    
    ax1 = fig.add_subplot(gs[:3, :2])
    ax1.plot(drumhead_height(2,3,rm,thetam, t))
    
    ax2 = fig.add_subplot(gs[0, 2])
    ax2.plot(drumhead_height(1,1,rm,thetam, t))
    
    ax3 = fig.add_subplot(gs[1, 2])
    ax3.plot(drumhead_height(1,5,rm,thetam, t))
    
    ax4 = fig.add_subplot(gs[2, 2])
    ax4.plot(drumhead_height(6,6,rm,thetam, t))

    
```

```python
 import mpl_toolkits.mplot3d 


def plot_Bessel(t):
    r=np.linspace(0,1,100)
    theta=np.linspace(0, 2*np.pi, 100)
    rm, thetam=np.meshgrid(r,theta)
    xm=rm*np.cos(thetam)
    ym=rm*np.sin(thetam)
    
    
    gs = plt.GridSpec(3, 3, wspace=0.2, hspace=0.2) 
  
    fig = plt.figure(figsize=(16, 6))
    
    ax1 = fig.add_subplot(gs[:3, :2],projection='3d') 
    ax1.plot_surface(xm, ym,drumhead_height(2,3,rm,thetam, t))
    
    ax2 = fig.add_subplot(gs[0, 2],projection='3d')
    ax2.plot_surface(xm, ym, drumhead_height(1,1,rm,thetam, t))
    
    ax3 = fig.add_subplot(gs[1, 2],projection='3d')
    ax3.plot_surface(xm, ym, drumhead_height(1,5,rm,thetam, t))
    
    ax4 = fig.add_subplot(gs[2, 2],projection='3d')
    ax4.plot_surface(xm, ym, drumhead_height(6,6,rm,thetam, t))

    

```

```python
from JSAnimation import IPython_display, HTMLWriter
from matplotlib import animation
import Image
import glob
```

```python
filenames=sorted(glob.glob('pictures/frame*.png'))
nframes=5
fig = plt.figure(figsize=(10,5))
ax = fig.add_axes([0, 0, 1, 1])
ax.axis('off')
im = plt.imshow(Image.open(filenames[0]))

# Create initializatio function
def init():
    im.set_data(Image.open(filenames[0]))
    return im,

# Create animation function
def animate(i):
    image=Image.open(filenames[i])
    im.set_data(image)
    return im,

anim = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=nframes, interval=20, blit=True)

#set embed_frames=True to embed base64-encoded frames directly in the HTML
anim.save('pictures/animation.html', writer=HTMLWriter(embed_frames=True))

# Show animation in ipython notebook
anim
```

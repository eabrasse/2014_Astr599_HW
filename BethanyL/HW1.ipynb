```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

# Part 1: loading and exploring the data

```python
data = np.genfromtxt('asteroids5000.csv',delimiter=',')
```

How many values are missing in this data?

```python
size(data[np.isnan(data)])
```

How many complete rows are there?

```python
rowsums = data.sum(1)
```

```python
size(rowsums[~np.isnan(rowsums)])
```

Create a new array containing only the rows with no missing values.

```python
CompleteRows = data[~np.isnan(rowsums),:]
```

Compute the maximum, minimum, mean, and standard deviation of the values in each column.

```python
np.max(CompleteRows,0)
```

```python
np.min(CompleteRows,0)
```

```python
np.mean(CompleteRows,0)
```

```python
np.std(CompleteRows,0)
```

plot the semi-major axis versus the sine of the inclination angle

```python
c = plt.scatter(data[:,0],np.sin(np.radians(data[:,2])))
plt.xlabel('semi-major axis')
plt.ylabel('sin(inclination angle)')
```

Plot a color-magnitude diagram of the asteroids (H vs B-V). Over-plot a line that divides the two families.

```python
c = plt.scatter(data[:,3],data[:,5])
x = np.linspace(np.min(data[:,3]), np.max(data[:,3]), 50)
y = .001*x + .77
plt.plot(x, y, linewidth=2);
```

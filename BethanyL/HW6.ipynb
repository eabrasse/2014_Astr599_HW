```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
def mandel(x, y, max_iter):
    """
    Given z = x + iy and max_iter, determine whether the candidate
    is in the mandelbrot set for the given number of iterations
    """
    c = complex(x, y)
    z = 0.0j
    for i in range(max_iter):
        z = z*z + c
        if (z.real*z.real + z.imag*z.imag) >= 4:
            return i

    return max_iter

def create_fractal(Nx, Ny, xmin, xmax, ymin, ymax, max_iter):
    """Create and return a fractal image"""
    image = np.zeros((Ny, Nx), dtype=float)
    dx = (xmax - xmin) * 1. / Nx
    dy = (ymax - ymin) * 1. / Ny
    
    for x in range(Nx):
        rpart = xmin + x * dx
        for y in range(Ny):
            ipart = ymin + y * dy
            color = mandel(rpart, ipart, max_iter)
            image[y, x] = color
    return image
```

```python
image = create_fractal(300, 200, -2, 1, -1, 1, 20)
plt.imshow(image, cmap=plt.cm.jet)
```

```python
%%cython
cimport cython
import numpy as np
cimport numpy as np
import matplotlib.pyplot as plt

cdef int mandel_cython(float x, float y, int max_iter):
    """
    Given z = x + iy and max_iter, determine whether the candidate
    is in the mandelbrot set for the given number of iterations
    """
    cdef complex c = complex(x, y)
    cdef complex z = complex(0, 0)
    cdef int i
    cdef float sizeZ, zReal, zImag
    
    for i in range(max_iter):
        z = z*z + c
        zReal = z.real
        zImag = z.imag
        sizeZ = zReal*zReal + zImag*zImag 
        if sizeZ >= 4:
            return i

    return max_iter

@cython.boundscheck(False) 
def create_fractal_cython(int Nx, int Ny, float xmin, float xmax, float ymin, float ymax, int max_iter):
    """Create and return a fractal image"""
    #cdef np.ndarray image = np.zeros((Ny, Nx), dtype=int)
    cdef  np.ndarray[np.int_t, ndim=2] image = np.zeros((Ny, Nx), dtype=np.int)
    cdef float dx = (xmax - xmin) * 1. / Nx
    cdef float dy = (ymax - ymin) * 1. / Ny
    
    cdef int x, y, color
    cdef float rpart, ipart
    for x in range(Nx):
        rpart = xmin + x * dx
        for y in range(Ny):
            ipart = ymin + y * dy
            color = mandel_cython(rpart, ipart, max_iter)
            image[y, x] = color
    return image
```

```python
image = create_fractal_cython(300, 200, -2, 1, -1, 1, 20)
plt.imshow(image, cmap=plt.cm.jet)
```

Part 2: Speed up this example using Numba. 

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from numba import autojit
```

```python
@autojit
def mandel_numba(x, y, max_iter):
    """
    Given z = x + iy and max_iter, determine whether the candidate
    is in the mandelbrot set for the given number of iterations
    """
    c = complex(x, y)
    z = 0.0j
    for i in range(max_iter):
        z = z*z + c
        if (z.real*z.real + z.imag*z.imag) >= 4:
            return i

    return max_iter

@autojit
def create_fractal_numba(Nx, Ny, xmin, xmax, ymin, ymax, max_iter):
    """Create and return a fractal image"""
    image = np.zeros((Ny, Nx), dtype=float)
    dx = (xmax - xmin) * 1. / Nx
    dy = (ymax - ymin) * 1. / Ny
    
    for x in range(Nx):
        rpart = xmin + x * dx
        for y in range(Ny):
            ipart = ymin + y * dy
            color = mandel_numba(rpart, ipart, max_iter)
            image[y, x] = color
    return image
```

```python
image = create_fractal_numba(300, 200, -2, 1, -1, 1, 20)
plt.imshow(image, cmap=plt.cm.jet)
```

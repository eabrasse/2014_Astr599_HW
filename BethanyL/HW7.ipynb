# Exercise 1: Classification of RR Lyrae

```python
from astroML.datasets import fetch_rrlyrae_combined
from sklearn.cross_validation import train_test_split

X, y = fetch_rrlyrae_combined()

# For now, we'll only fit the first two colors
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

N_plot = 5000
plt.scatter(X[-N_plot:, 0], X[-N_plot:, 1], c=y[-N_plot:],
            edgecolors='none')
plt.xlabel('u-g color')
plt.ylabel('g-r color');
```

### Original Test - K Nearest Neighbors

```python
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier(n_neighbors=5)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
```

```python
print(np.sum(y_pred == y_test))
print(np.sum(y_pred != y_test))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred,
                            target_names=['MS star', 'RR Lyrae']))
```

### Gaussian Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
clfGNB = GaussianNB()
clfGNB.fit(X_train, y_train)
y_predGNB = clfGNB.predict(X_test)
```

```python
print(np.sum(y_predGNB == y_test))
print(np.sum(y_predGNB != y_test))
```

```python
print(classification_report(y_test, y_predGNB,
                            target_names=['MS star', 'RR Lyrae']))
```

### Linear Discriminant Analysis (LDA)

```python
from sklearn.lda import LDA
clfLDA = LDA()
clfLDA.fit(X_train, y_train)
y_predLDA = clfLDA.predict(X_test)
```

```python
print(np.sum(y_predLDA == y_test))
print(np.sum(y_predLDA != y_test))
```

```python
print(classification_report(y_test, y_predLDA,
                            target_names=['MS star', 'RR Lyrae']))
```

### C-Support Vector Classification

```python
from sklearn.svm import SVC
clfSVC = SVC()
clfSVC.fit(X_train, y_train)
y_predSVC = clfSVC.predict(X_test)
```

```python
print(np.sum(y_predSVC == y_test))
print(np.sum(y_predSVC != y_test))
```

```python
print(classification_report(y_test, y_predSVC,
                            target_names=['MS star', 'RR Lyrae']))
```

### C-Support Vector Classification with RBF kernel

### Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
#n_estimators = number of trees (10 is default)
#max_features = number of features to consider (sqrt(#features) is default - 2 here)
clfRFC = RandomForestClassifier(n_estimators=20, max_features=4)
clfRFC.fit(X_train, y_train)
y_predRFC = clfRFC.predict(X_test)
```

```python
print(np.sum(y_predRFC == y_test))
print(np.sum(y_predRFC != y_test))
```

```python
print(classification_report(y_test, y_predRFC,
                            target_names=['MS star', 'RR Lyrae']))
```

```python
def PlotFeatureImportances(forest):
    #this function adapted from http://scikit-learn.org/stable/auto_examples/ensemble/plot_forest_importances.html
    importances = forest.feature_importances_
    std = np.std([tree.feature_importances_ for tree in forest.estimators_],
                 axis=0)
    indices = np.argsort(importances)[::-1]

    # Print the feature ranking
    print("Feature ranking:")
    n_samples, n_features = X.shape
    print(n_features)
    
    for f in range(n_features):
        print("%d. feature %d (%f)" % (f + 1, indices[f], importances[indices[f]]))

    # Plot the feature importances of the forest
    plt.figure()
    plt.title("Feature importances")
    plt.bar(range(n_features), importances[indices],
           color="r", yerr=std[indices], align="center")
    plt.xticks(range(n_features), indices)
    plt.xlim([-1, 10])
    plt.show()
```

```python
PlotFeatureImportances(clfRFC)
```

There is not a big drop-off in feature importance, so I'll tell the algorithm to consider all of them.

My best precision/recall for RR Lyrae is  0.77/0.54 with Random Forest 

# Exercise 2: Regression: Photometric Redshifts

```python
from astroML.datasets import fetch_sdss_specgals
```

```python
data = fetch_sdss_specgals()

# put magnitudes in a matrix
X = np.vstack([data['modelMag_%s' % f] for f in 'ugriz']).T
y = data['z']

# Split into training and testing data
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

### Original Test - Simple Linear Regression

```python
from sklearn.linear_model import LinearRegression
est = LinearRegression()
est.fit(X_train, y_train)
y_pred = est.predict(X_test)
```

```python
plt.plot(y_test, y_pred, ',k')
plt.plot([0, 1], [0, 1], ':k')
plt.xlim(0, 0.6)
plt.ylim(0, 0.6)
```

```python
rms = np.sqrt(np.mean((y_test - y_pred) ** 2))
print(rms)
```

### Regression based on k-nearest neighbors.

```python
from sklearn.neighbors import KNeighborsRegressor
estK = KNeighborsRegressor()
estK.fit(X_train, y_train)
y_predK = estK.predict(X_test)
```

```python
plt.plot(y_test, y_predK, ',k')
plt.plot([0, 1], [0, 1], ':k')
plt.xlim(0, 0.6)
plt.ylim(0, 0.6)
```

```python
rms = np.sqrt(np.mean((y_test - y_predK) ** 2))
print(rms)
```

### epsilon-Support Vector Regression.

```python
from sklearn.svm import SVR
estSVR = SVR()
estSVR.fit(X_train[::5], y_train[::5])
y_predSVR = estSVR.predict(X_test)
```

```python
plt.plot(y_test, y_predSVR, ',k')
plt.plot([0, 1], [0, 1], ':k')
plt.xlim(0, 0.6)
plt.ylim(0, 0.6)
```

```python
rms = np.sqrt(np.mean((y_test - y_predSVR) ** 2))
print(rms)
```

### Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
estRF = RandomForestRegressor()
estRF.fit(X_train, y_train)
y_predRF = estRF.predict(X_test)
```

```python
plt.plot(y_test, y_predRF, ',k')
plt.plot([0, 1], [0, 1], ':k')
plt.xlim(0, 0.6)
plt.ylim(0, 0.6)
```

```python
rms = np.sqrt(np.mean((y_test - y_predRF) ** 2))
print(rms)
```

### Gradient Boosting for regression

```python
from sklearn.ensemble import GradientBoostingRegressor
estGB = GradientBoostingRegressor()
estGB.fit(X_train[::3], y_train[::3])
y_predGB = estGB.predict(X_test)
```

```python
plt.plot(y_test, y_predGB, ',k')
plt.plot([0, 1], [0, 1], ':k')
plt.xlim(0, 0.6)
plt.ylim(0, 0.6)
```

```python
#import packages
%pylab inline 
import numpy as np
import matplotlib.pyplot as plt
from scipy import *
from scipy.special import jn, jn_zeros
```

```python
def drumhead_height(n,k,r,theta,t):
    zeros = jn_zeros(n,k)   
    t1,t2, t3= np.meshgrid( r[0],zeros,theta[0])
    test = np.array(t2*r)
    height = np.array(np.cos(t) * np.cos(n * t3[0]) *  jn(n,test[0]) )
    return height
```

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
def plot_Bessel(t):
   r = np.linspace(0,1)
   theta = np.linspace(0,2 * np.pi)
   rmesh,thetam = np.meshgrid(r,theta,indexing = 'ij')
   x = np.array(rmesh * np.cos(thetam))
   y = np.array(rmesh * np.sin(thetam))
   gs = plt.GridSpec(3 ,3, wspace = 0.3, hspace = 0.2)
   fig = plt.figure(figsize = (16,6))
   ax1 = fig.add_subplot(gs[:3,:2],projection = '3d')
   ax2 = fig.add_subplot(gs[0,2],projection = '3d')
   ax3 = fig.add_subplot(gs[1,2],projection = '3d') 
   ax4 = fig.add_subplot(gs[2,2],projection = '3d')  
   height1 = drumhead_height(2 ,3, rmesh, thetam, t)
   height2 = drumhead_height(5 ,10 , rmesh, thetam, t)
   height3 = drumhead_height(7 ,15 , rmesh, thetam, t)
   height4 = drumhead_height(4 ,4 , rmesh, thetam, t)
   ax1.plot_surface(x,y,height1, rstride=1, cstride=1, cmap=cm.jet) 
   ax2.plot_surface(x,y,height2, rstride=1, cstride=1, cmap=cm.jet) 
   ax3.plot_surface(x,y,height3, rstride=1, cstride=1, cmap=cm.jet) 
   ax4.plot_surface(x,y,height4, rstride=1, cstride=1, cmap=cm.jet) 
    
```

```python
plot_Bessel(1.0)
plot_Bessel(2.0)
plot_Bessel(3.0)
plot_Bessel(4.0)
plot_Bessel(5.0)


nframes = 5 

for i in xrange(0,nframes,1):
    time = 1.5 * i
    # Call plotting function and save figure into file
    plot_Bessel(time);
    plt.savefig('drumhead/aframe{0:07d}.png'.format(i+1));
    plt.close()
    
```

```python
# Import JSAnimation 
from JSAnimation import IPython_display, HTMLWriter
from matplotlib import animation
import Image
import glob

filenames=sorted(glob.glob(('drumhead/aframe*.png')))

fig =    fig = plt.figure(figsize = (16,6))

im = plt.imshow(Image.open(filenames[0]))
def init():
    im.set_data(Image.open(filenames[0]))
    return im,
def animate(i):
    image=Image.open(filenames[i])
    im.set_data(image)
    return im,    

anim = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=nframes, interval=20, blit=True)

anim.save('drumhead/animation.html', writer=HTMLWriter(embed_frames=True))


# Show animation in ipython notebook
anim
```

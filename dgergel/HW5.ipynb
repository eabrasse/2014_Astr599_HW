```python
## 1) 
# Create a function called drumhead_height that calculates un,k(r,θ,t). This function should take as argument n,k, 
# the radius r, the angle θ and the time t. In order to calculate the Bessel functions import:
def drumhead_height(n,k,r,theta,t):
    import numpy as np
    from scipy.special import jn,jn_zeros
    ## note: The function jn(n,x) gives the Bessel function Jn(x), and the function jn_zeros(n,k) gives you a list with 
    ## the k zeros of Jn function.
    k_zeros = jn_zeros(n,k)
    lambda_1 = k_zeros[len(k_zeros)-1]
    ht = np.cos(t)*np.cos(n*theta)*jn(n,lambda_1*r)
    return ht
```

```python
## 2) 
# Create a fuction called plot_Bessel that takes time t as an argument. Inside this function create a figure with the 
# GridSpec() function as in the example in the notebook; the grid should be 3x3 units. You are going to make four subplots. 
# The first subplot is going to occupy the three rows and the two first columns. The next subplots will each occupy one row 
# of the third column of your grid.
# At the beggining of the plot_Bessel function, create two arrays with linspace. One for the radius r from 0 to 1 and 
# another one for the angle θ from 0 to 2π. Then use the function meshgrid, to create the polar coordintes grid (r,θ). 
# Use the output just obtained and transform it to cartesian grid using,
def plot_Bessel(t):
    ## two arrays with linspace
    from math import pi
    from mpl_toolkits.mplot3d import axes3d
    import matplotlib.pyplot as plt
    import numpy as np
    %matplotlib inline
    rr = np.linspace(0,1,100)
    thetat = np.linspace(0,pi,100)
    r,theta = np.meshgrid(rr,thetat)
    x = r*np.cos(thetat)
    y = r*np.sin(thetat)
    # 3) 
    # Create a 3 x 3 plotting grid object with horizontal and vertical spacing wspace and hspace
    gs = plt.GridSpec(3, 3, wspace=0.2, hspace=0.2) 
    # Create a figure
    fig = plt.figure(figsize=(16, 6))
    # 4) 
    # For each of the subplots, calculate the value of the elevation of the drumhead using the function you created in 1). 
    # Remeber the drumhead_height function takes as input the grid in polar coordinates. For the first subplot use the 
    # values n=2 and k=3 to calculate your solution. For the other three subplots, you can choose your favorite values 
    # of n and k.
    # drumhead_height(n,k,r,theta,t)
    #### subfigure 1
    ax1 = fig.add_subplot(gs[:4, :2],projection='3d')
    ht = drumhead_height(2,3,r,theta,t)
    ax1.plot_surface(x,y,ht,rstride=8, cstride=8, alpha=0.3)
    #### subfigure 2
    ax2 = fig.add_subplot(gs[0,2],projection='3d')
    ht = drumhead_height(1,2,r,theta,t)
    ax2.plot_surface(x,y,ht,rstride=8, cstride=8, alpha=0.3)
    #### subfigure 3
    ax3 = fig.add_subplot(gs[1,2],projection='3d')
    ht = drumhead_height(4,1,r,theta,t)
    ax3.plot_surface(x,y,ht,rstride=8, cstride=8, alpha=0.3)
    #### subfigure 4
    ax4 = fig.add_subplot(gs[2,2],projection='3d')
    ht = drumhead_height(5,4,r,theta,t)
    ax4.plot_surface(x,y,ht,rstride=8, cstride=8, alpha=0.3)
    # 5) 
    # In each of the subplots, use the function plot_surface from the 3D plotting library mplot3d to plot the 
    # surface (just as in the class examples). Note you will require to input the cartesian grid in the plotting function 
    # (not the polar grid).
```

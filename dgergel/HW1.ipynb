Part 1: loading and exploring the data

```python
## load data
import numpy as np
asts = np.genfromtxt('/usr1/gergel/ASTR_599/asteroids5000.csv',dtype=float, delimiter=",")

## how many values are missing in the data?
mask = np.isnan(asts)
print ('there are %.2f missing values from the data' %np.sum(mask))

## how many complete rows are there? i.e. how many objects have no missing values
    ## compute sum along rows
    ## if row sum = 0, there are no missing values in the row (i.e. all mask values in the row are false, no nan's)
no_rows = np.sum(mask,axis=1)
print ('there are %.0f rows with no missing values' %np.sum(no_rows == 0))

## create new array containing only the rows with no missing values
asts_no_miss_vals = asts[no_rows== 0]

## compute maximum, minimum, mean, and standard deviation of values in each column
## maximum
max_cols = asts_no_miss_vals.max(axis=0)
print ('the max of each column is')
print max_cols
## minimum
min_cols = asts_no_miss_vals.min(axis=0)
print ('the min of each column is')
print min_cols
## mean
mean_cols = asts_no_miss_vals.mean(axis=0)
print ('the mean of each column is')
print mean_cols
## std
std_cols = asts_no_miss_vals.std(axis=0)
print ('the std of each column is')
print std_cols
```

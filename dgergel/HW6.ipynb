```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
def mandel(x, y, max_iter):
    """
    Given z = x + iy and max_iter, determine whether the candidate
    is in the mandelbrot set for the given number of iterations
    """
    c = complex(x, y)
    z = 0.0j
    for i in range(max_iter):
        z = z*z + c
        if (z.real*z.real + z.imag*z.imag) >= 4:
            return i

    return max_iter

def create_fractal(Nx, Ny, xmin, xmax, ymin, ymax, max_iter):
    """Create and return a fractal image"""
    image = np.zeros((Ny, Nx), dtype=float)
    dx = (xmax - xmin) * 1. / Nx
    dy = (ymax - ymin) * 1. / Ny
    
    for x in range(Nx):
        rpart = xmin + x * dx
        for y in range(Ny):
            ipart = ymin + y * dy
            color = mandel(rpart, ipart, max_iter)
            image[y, x] = color
    return image
```

```python
%%cython 
import numpy as np
cimport numpy as np
cimport cython

cdef inline int mandel_cy(double x, double y, int max_iter):
    """
    Given z = x + iy and max_iter, determine whether the candidate
    is in the mandelbrot set for the given number of iterations
    """
    cdef double complex c 
    c.real = x
    c.imag = y
    cdef double complex z = 0.0j
    
    for i in range(max_iter):
        z = z*z + c
        zr = z.real
        zi = z.imag
        if (zr*zr + zi*zi) >= 4:
            return i

    return max_iter

def create_fractal_cy(int Nx, int Ny, double xmin, double xmax, double ymin, double ymax, int max_iter):
    """Create and return a fractal image"""
    cdef np.ndarray[np.int32_t, ndim=2] image = np.zeros((Ny, Nx), dtype=np.int32)
    cdef double dx = (xmax - xmin) * 1. / Nx
    cdef double dy = (ymax - ymin) * 1. / Ny
    cdef int x,y
    cdef double rpart,ipart
    cdef int color
    
    for x in range(Nx):
        rpart = xmin + x * dx
        for y in range(Ny):
            ipart = ymin + y * dy
            color = mandel_cy(rpart, ipart, max_iter)
            image[y, x] = color
    return image
```

```python
from numba import autojit 
mandel_nb = autojit(mandel)
create_fractal_nb = autojit(create_fractal)
```

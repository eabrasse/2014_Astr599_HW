```python
import numpy as np
import matplotlib.pyplot as plt
```

```python


from astroML.datasets import fetch_rrlyrae_combined
from sklearn.cross_validation import train_test_split

X, y = fetch_rrlyrae_combined()

# For now, we'll only fit the first two colors
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python


N_plot = 5000
plt.scatter(X[-N_plot:, 0], X[-N_plot:, 1], c=y[-N_plot:],
            edgecolors='none')
plt.xlabel('u-g color')
plt.ylabel('g-r color');
```

```python
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier(n_neighbors=5)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

print(np.sum(y_pred == y_test))
print(np.sum(y_pred != y_test))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred, target_names=['MS star', 'RR Lyrae']))
```

Exercise 1
Now to try some of the different available scikitlearn schema

```python
from sklearn.lda import LDA
clf = LDA()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

print(np.sum(y_pred ==y_test))
print(np.sum(y_pred!=y_test))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred, target_names=['MS star', 'RR Lyrae']))
```

```python
from sklearn.svm import SVC
clf = SVC(kernel='rbf')
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

print(np.sum(y_pred ==y_test))
print(np.sum(y_pred!=y_test))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred, target_names=['MS star', 'RR Lyrae']))
```

So both of these do alright, but not as well as KNeighborsClassifier.

```python
from astroML.datasets import fetch_sdss_specgals
```

```python
data=fetch_sdss_specgals()

X=np.vstack([data['modelMag_%s' % f] for f in 'ugriz']).T
y = data['z']

#Split into training and testing data
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
from sklearn.linear_model import LinearRegression
est = LinearRegression()
est.fit(X_train, y_train)
y_pred=est.predict(X_test)
```

```python
plt.plot(y_test, y_pred, ',k')
plt.plot([0, 1], [0, 1], ':k')
plt.xlim(0,0.6)
plt.ylim(0,0.6)
```

```python
rms = np.sqrt(np.mean((y_test - y_pred) **2 ))
print(rms)
```

That's not a very good fit, so let's see what we can do with some of the other schema

```python
from sklearn.neighbors import KNeighborsRegressor
est = KNeighborsRegressor(n_neighbors=5)
est.fit(X_train, y_train)
y_pred=est.predict(X_test)
```

```python
plt.plot(y_test, y_pred, ',k')
plt.plot([0, 1], [0, 1], ':k')
plt.xlim(0,0.6)
plt.ylim(0,0.6)
```

```python
rms = np.sqrt(np.mean((y_test - y_pred) **2 ))
print(rms)
```

Even though it looks like a better fit, the rms is even worse than before. Let's try something else.

```python
from sklearn.ensemble import RandomForestRegressor
est= RandomForestRegressor(n_estimators=10)
est.fit(X_train,y_train)
y_pred=est.predict(X_test)
```

```python
plt.plot(y_test, y_pred, ',k')
plt.plot([0, 1], [1, 0], ':k')
plt.xlim(0,0.6)
plt.ylim(0,0.6)
```

```python
rms = np.sqrt(np.mean((y_test - y_pred) **2 ))
print(rms)
```

That didn't work at all.

I don't know if I'll get a better fit of the data with these regressors.

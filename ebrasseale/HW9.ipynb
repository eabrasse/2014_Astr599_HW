```python
from IPython import parallel
clients=parallel.Client()
clients.block=True
print(clients.ids)
dview=clients[:]
```

```python
%%px
from astroML.datasets import fetch_sdss_specgals
import numpy as np
data=fetch_sdss_specgals()

#put magnitudes in a matrix
X=np.vstack([data['modelMag_%s' % f] for f in 'ugriz']).T
y= data['z']

#down-sample the data for speed
X=X[::10]
y=y[::10]
```

```python

#%px from sklearn.ensemble import RandomForestRegressor
#%px from sklearn.cross_validation import cross_val_score, StratifiedKFold

dview.scatter('a',range(5,25))

dview.execute('''
score=[]
for num in a:
    # define the model with the given value of C
    model = RandomForestRegressor(n_estimators=num, max_depth=3)
    #compute the scores via cross-validation
    scores = cross_val_score(model, X, y, scoring= 'r2',cv=3)
    
    #print the mean of the cross-validation scores
    score += [np.mean(scores)]''')

score=dview.gather('score')
print(score)
```

```python
import matplotlib.pyplot as plt

plt.plot(range(5,25),score)
plt.xlabel('Number of Estimators')
plt.ylabel('Mean score')
plt.title('Change in mean score when # of estimators is varied')
```

No real correspondence between the number of estimators and improvement in cross validation score. Let's try fiddling with max depth

```python
dview.scatter('a',range(5,25))

dview.execute('''
score=[]
for num in a:
    # define the model with the given value of C
    model = RandomForestRegressor(n_estimators=10, max_depth=num)
    #compute the scores via cross-validation
    scores = cross_val_score(model, X, y, scoring= 'r2',cv=3)
    
    #print the mean of the cross-validation scores
    score += [np.mean(scores)]''')

score=dview.gather('score')

plt.plot(range(5,25),score)
plt.xlabel('max depth')
plt.ylabel('mean score')
plt.title('Change in mean score as max depth is varied')
```

That's much better! There seems to be exponential increase that sort of asymptotes off. I wonder if there's a maximum between 15 and 20?

```python
maxscore=a[find(score==max(score))]
print([max(score),maxscore])
```

So it seems the maximum score is 0.8, which occurs with a max_depth value of 16.
If we plug that max_depth value into the num_estimators variance, can we get a better score?

```python
dview.scatter('a',range(5,25))

dview.execute('''
score=[]
for num in a:
    # define the model with the given value of C
    model = RandomForestRegressor(n_estimators=num, max_depth=16)
    #compute the scores via cross-validation
    scores = cross_val_score(model, X, y, scoring= 'r2',cv=3)
    
    #print the mean of the cross-validation scores
    score += [np.mean(scores)]''')

score=dview.gather('score')

plt.plot(range(5,25),score)
plt.xlabel('Number of Estimators')
plt.ylabel('Mean score')
plt.title('Change in mean score when # of estimators is varied')
```

Wow! Now there appears to be a distinct trend in the # of estimators. Let's move the range up to see if I can find a maximum.

```python
dview.scatter('a',range(45,52))

dview.execute('''
score=[]
for num in a:
    # define the model with the given value of C
    model = RandomForestRegressor(n_estimators=num, max_depth=16)
    #compute the scores via cross-validation
    scores = cross_val_score(model, X, y, scoring= 'r2',cv=3)
    
    #print the mean of the cross-validation scores
    score += [np.mean(scores)]''')

score=dview.gather('score')

plt.plot(range(45,52),score)
plt.xlabel('Number of Estimators')
plt.ylabel('Mean score')
plt.title('Change in mean score when # of estimators is varied')
```

Well, it seems to continue increasing as the number of estimators goes up. I'll say the maximum value I can reach is around 0.81

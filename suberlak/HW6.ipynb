## Part 0 : Original Code 

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def mandel(x, y, max_iter):
    """
    Given z = x + iy and max_iter, determine whether the candidate
    is in the mandelbrot set for the given number of iterations
    """
    c = complex(x, y)
    z = 0.0j
    for i in range(max_iter):
        z = z*z + c
        if (z.real*z.real + z.imag*z.imag) >= 4:
            return i

    return max_iter

def create_fractal(Nx, Ny, xmin, xmax, ymin, ymax, max_iter):
    """Create and return a fractal image"""
    image = np.zeros((Ny, Nx), dtype=float)
    dx = (xmax - xmin) * 1. / Nx
    dy = (ymax - ymin) * 1. / Ny
    
    for x in range(Nx):
        rpart = xmin + x * dx
        for y in range(Ny):
            ipart = ymin + y * dy
            color = mandel(rpart, ipart, max_iter)
            image[y, x] = color
    return image

image = create_fractal(300, 200, -2, 1, -1, 1, 20)
plt.imshow(image, cmap=plt.cm.jet)
```

```python
%%file mandel.pyx
%%cython
#cimport cython
import numpy as np
cimport numpy as np

def c_mandel(float x, float y, int max_iter):
    """
    Given z = x + iy and max_iter, determine whether the candidate
    is in the mandelbrot set for the given number of iterations
    """
    cdef float complex c = complex(x,y) 
    cdef float complex z  = complex(0,0)
  
    cdef int i
    for i in range(max_iter):
        z = z*z + c
        if (z.real*z.real + z.imag*z.imag) >= 4:
            return i

    return max_iter

@cython.boundscheck(False)
def c_create_fractal(int Nx,int Ny,float xmin,float xmax,float ymin,float ymax,int max_iter):
    """Create and return a fractal image"""
    cdef np.ndarray[long,ndim=2] image = np.zeros((Ny, Nx), dtype=np.int)
    cdef float dx = (xmax - xmin)  / Nx
    cdef float dy = (ymax - ymin)  / Ny
    
    cdef int x,y
    cdef float rpart, ipart
    cdef int color=0
    
    for x in range(Nx):
        rpart = xmin + x * dx
        for y in range(Ny):
            ipart = ymin + y * dy
            color = c_mandel( rpart, ipart, max_iter)
            image[y, x] = color
    return image
```

## Part 2 : Speed up with Numba 


Speed up this example using Numba. Numba is much more of a black box, but you should be able to do it very quickly. You'll likely find that simply adding the decorator to the function doesn't work. I'll give you a hint: Numba doesn't like an array specified with dtype=float. Look back to our previous example for clues about how to fix this. (Also, you might find this notebook to be a useful resource: it is all about doing the mandelbrot set with Numba).



```python
from numba import autojit
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 

@autojit
def numba_mandel(x, y, max_iter):
    """
    Given z = x + iy and max_iter, determine whether the candidate
    is in the mandelbrot set for the given number of iterations
    """
    c = complex(x, y)
    z = 0.0j
    for i in range(max_iter):
        z = z*z + c
        if (z.real*z.real + z.imag*z.imag) >= 4:
            return i

    return max_iter

@autojit
def numba_create_fractal(Nx, Ny, xmin, xmax, ymin, ymax, max_iter):
    """Create and return a fractal image"""
    image = np.zeros((Ny, Nx), dtype=float)
    dx = (xmax - xmin) * 1. / Nx
    dy = (ymax - ymin) * 1. / Ny
    
    for x in range(Nx):
        rpart = xmin + x * dx
        for y in range(Ny):
            ipart = ymin + y * dy
            color = numba_mandel(rpart, ipart, max_iter)
            image[y, x] = color
    return image

image = numba_create_fractal(300, 200, -2, 1, -1, 1, 20)
plt.imshow(image, cmap=plt.cm.jet)
```

## Part 2: Plotting the data

Display the first line of the data file.

```python
!head -1 asteroids5000.csv
```

Set up `matplotlib` environment.

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

Plot the semi-major axis versus the sine of the inclination angle.

```python
Da = D[:,0]
Di = np.sin(np.radians(D[:,2]))
plt.scatter(Da, Di, s=10, alpha=.3)
plt.xlabel('a')
plt.ylabel('sin(i)')
```

It seemed that many asteroids had semi-major axis from 2 to 3 `AU` and  inclicination angle lower than 0.3. Beyond the axis of 3.2 `AU`, only scarce distributions at 3.5 and 4.0 `AU` were observed. The asteroids with small orbits had relatively lower inclination angles up to 0.1, that increased up to 0.3 until the orbit reached 2.5 `AU`.

Plot a color-magnitude diagram of the asteroids (H vs B-V).

```python
DH = D[:,3]
DBV = D[:,5]
plt.scatter(DH, DBV, s=10, alpha=.3)
plt.xlabel('H')
plt.ylabel('B-V')

cut = 0.78 # picked up manually
plt.axhline(cut)
```

Repeat the orbital parameter plot with two "families" from above in different colors.

```python
plt.scatter(Da, Di, s=10, c=np.where(DBV > cut, 'r', 'b'), alpha=.3)
plt.xlabel('a')
plt.ylabel('sin(i)')
```

The asteroids with higher magnitudes (red) often had smaller orbits. While lower magnitudes (blue) were observed in the entire range, they prevailed in orbits larger than 3 `AU`.

Our plot resembles Fig. 4 of [Parker et al. 2008](http://adsabs.harvard.edu/abs/2008Icar..198..138P) (shown below) in several aspects:

1. A smaller range of inclination angles were observed in semi-major axis lower than 2.5 `AU`.
2. Higher magnitudes (reddish) gathered in smaller orbits.
3. An asteroidal gap was observed at about 2.9 `AU`.

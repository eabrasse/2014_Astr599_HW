```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def mandel(x, y, max_iter):
    """
    Given z = x + iy and max_iter, determine whether the candidate
    is in the mandelbrot set for the given number of iterations
    """
    c = complex(x, y)
    z = 0.0j
    for i in range(max_iter):
        z = z*z + c
        if (z.real*z.real + z.imag*z.imag) >= 4:
            return i

    return max_iter

def create_fractal(Nx, Ny, xmin, xmax, ymin, ymax, max_iter):
    """Create and return a fractal image"""
    image = np.zeros((Ny, Nx), dtype=float)
    dx = (xmax - xmin) * 1. / Nx
    dy = (ymax - ymin) * 1. / Ny
    
    for x in range(Nx):
        rpart = xmin + x * dx
        for y in range(Ny):
            ipart = ymin + y * dy
            color = mandel(rpart, ipart, max_iter)
            image[y, x] = color
    return image
```

```python
def show_fractal(func, *args, **kwargs):
    style = {
        'cmap': plt.cm.jet,
    }
    style.update(kwargs)
    plt.imshow(func(*args), **style)
```

#### CTypes

```python
%%file mb.c
int mandel(double x, double y, int max_iter) {
    int i;
    double zx = 0.0;
    double zy = 0.0;
    double zxx;
    for (i = 0; i < max_iter; i++) {
        zxx = zx*zx - zy*zy + x;
        zy = 2*zx*zy + y;
        zx = zxx;
        if ((zx*zx + zy*zy) >= 4) {
            return i;
        }
    }
    return max_iter;
}

void create_fractal(int* img, int Nx, int Ny, double xmin, double xmax, double ymin, double ymax, int max_iter) {
    int x, y;
    double dx = (xmax - xmin) / Nx;
    double dy = (ymax - ymin) / Ny;
    double rpart, ipart;
    int color;
    for (x = 0; x < Nx; x++) {
        rpart = xmin + x*dx;
        for (y = 0; y < Ny; y++) {
            ipart = ymin + y*dy;
            color = mandel(rpart, ipart, max_iter);
            img[x + y*Nx] = color;
        }
    }
}
```

```python
from ctypes import CDLL, c_void_p, c_int, c_double

def create_fractal_ctypes(Nx, Ny, xmin, xmax, ymin, ymax, max_iter):
    image = np.zeros((Ny, Nx), dtype=np.int32)
    mb = CDLL('mb.dylib')
    mb.create_fractal(image.ctypes.data_as(c_void_p),
                      c_int(Nx),
                      c_int(Ny),
                      c_double(xmin),
                      c_double(xmax),
                      c_double(ymin),
                      c_double(ymax),
                      c_int(max_iter))
    return image
```

```python
%%file mb.f90
    INTEGER FUNCTION mandel(x, y, max_iter)
        IMPLICIT NONE
        
        REAL :: x, y
        INTEGER :: max_iter
                
        INTEGER :: i
        COMPLEX :: z
        
        z = (0, 0)
        DO i = 0, max_iter-1
            z = z*z + cmplx(x, y)
            IF ((real(z)**2 + aimag(z)**2) .GT. 4) THEN
                mandel = i
                RETURN
            END IF
        END DO
        mandel = max_iter
    END FUNCTION mandel
    
    SUBROUTINE create_fractal(img, Nx, Ny, xmin, xmax, ymin, ymax, max_iter)
        IMPLICIT NONE
        
        INTEGER, INTENT(IN) :: Nx, Ny
        INTEGER, DIMENSION(Ny, Nx), INTENT(OUT) :: img
!f2py DEPEND(Ny, Nx) img

        REAL :: xmin, xmax, ymin, ymax
        INTEGER :: max_iter
                
        INTEGER :: x, y, color
        REAL :: dx, dy, rpart, ipart
                
        INTEGER, EXTERNAL :: mandel
        
        dx = (xmax - xmin) / Nx
        dy = (ymax - ymin) / Ny
        
        DO x = 0, Nx-1
            rpart = xmin + x*dx
            DO y = 0, Ny-1
                ipart = ymin + y*dy
                color = mandel(rpart, ipart, max_iter)
                img(y, x) = color
            END DO
        END DO
    END SUBROUTINE create_fractal
```

#### Cython

```python
%%cython -a
import numpy as np
cimport numpy as np
cimport cython

cdef inline int mandel_cython(double x, double y, int max_iter):
    """
    Given z = x + iy and max_iter, determine whether the candidate
    is in the mandelbrot set for the given number of iterations
    """
    cdef double complex c
    c.real = x
    c.imag = y
    cdef double complex z = 0.0j
    
    cdef int i
    for i in range(max_iter):
        z = z*z + c
        zr = z.real
        zi = z.imag
        if (zr**2 + zi**2) >= 4:
            return i

    return max_iter

@cython.cdivision(True)
@cython.boundscheck(False)
def create_fractal_cython(int Nx, int Ny, double xmin, double xmax, double ymin, double ymax, int max_iter):
    """Create and return a fractal image"""
    cdef np.ndarray[np.int32_t, ndim=2] image = np.zeros((Ny, Nx), dtype=np.int32)
    cdef double dx = (xmax - xmin) / Nx
    cdef double dy = (ymax - ymin) / Ny
    
    cdef int x, y
    cdef double rpart, ipart
    cdef int color
    
    for x in range(Nx):
        rpart = xmin + x * dx
        for y in range(Ny):
            ipart = ymin + y * dy
            color = mandel_cython(rpart, ipart, max_iter)
            image[y, x] = color
    return image
```

```python
from numba import autojit
```

My initial idea was to replace `float` with `int32` for better performance and compatibility. But it seems `image` array with `dtype=float` still works pretty well with numba, so it remains unchanged as is.

```python
@autojit
def mandel_numba(x, y, max_iter):
    """
    Given z = x + iy and max_iter, determine whether the candidate
    is in the mandelbrot set for the given number of iterations
    """
    c = complex(x, y)
    z = 0.0j
    for i in range(max_iter):
        z = z*z + c
        if (z.real*z.real + z.imag*z.imag) >= 4:
            return i

    return max_iter

@autojit
def create_fractal_numba(Nx, Ny, xmin, xmax, ymin, ymax, max_iter):
    """Create and return a fractal image"""
    image = np.zeros((Ny, Nx), dtype=float)
    dx = (xmax - xmin) * 1. / Nx
    dy = (ymax - ymin) * 1. / Ny
    
    for x in range(Nx):
        rpart = xmin + x * dx
        for y in range(Ny):
            ipart = ymin + y * dy
            color = mandel_numba(rpart, ipart, max_iter)
            image[y, x] = color
    return image
```

```python
show_fractal(create_fractal_numba, 300, 200, -2, 1, -1, 1, 20)
```

```python
def save_fractal(Nx, Ny, cx, cy, s, max_iter, cmap=plt.cm.jet, dpi=600, filename='fractal.png', func=create_fractal_numba):
    w, h = 1.*Nx/dpi, 1.*Ny/dpi
    fig = plt.figure(figsize=(w, h), dpi=dpi)
    ax = fig.add_axes((0, 0, 1, 1))
    ax.set_axis_off()
    ax.set_xticks([])
    ax.set_yticks([])
    show_fractal(func, Nx, Ny, cx-s, cx+s, cy-s, cy+s, max_iter, cmap=cmap)
    plt.savefig(filename, dpi=dpi)
```

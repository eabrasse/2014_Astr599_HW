## Exercise 1: Cross-validation of photometric redshifts

### Parallel algorithm

Load dataset for each client in the cluster.

```python
%%px
from astroML.datasets import fetch_sdss_specgals
import numpy as np
data = fetch_sdss_specgals()

# put magnitudes in a matrix
X = np.vstack([data['modelMag_%s' % f] for f in 'ugriz']).T
y = data['z']

# down-sample the data for speed
X = X[::10]
y = y[::10]
```

Define a score function for each client.

```python
%%px
from sklearn.ensemble import RandomForestRegressor
from sklearn.cross_validation import cross_val_score

def score(n_estimator, max_depth):
    # define the model with the given value of C
    model = RandomForestRegressor(n_estimators=n_estimator, max_depth=max_depth)

    # compute the scores via cross-validation
    scores = cross_val_score(model, X, y, scoring='r2', cv=3)

    # print the mean of the cross-validation scores
    score = np.mean(scores)
    return score
```

### Parameter grid

```python
from itertools import product
N_ESTIMATOR = range(1, 40)
MAX_DEPTH = range(1, 30)
params = list(product(N_ESTIMATOR, MAX_DEPTH))
```

```python
view.scatter('params', params)
```

### Result

```python
view.execute('scores = [score(*p) for p in params]')
```

```python
scores = view.gather('scores')
```

```python
max(scores)
```

```python
params[scores.index(max(scores))]
```

Best `R2` score is 0.8116 and corresponding values of `n_estimators` and `max_depth` are 37 and 23 respectively.

## Exercise 2: Computing periods of stars

### Parallel algorithm

```python
from astroML.datasets import fetch_LINEAR_sample
lcs = fetch_LINEAR_sample()
```

```python
%%px
from astroML.time_series import lomb_scargle
PERIODS = np.logspace(-2, 0, 10000)

def period(id_):
    time, flux, dflux = lcs[id_].T
    periodogram = lomb_scargle(time, flux, dflux, omega=2*np.pi/PERIODS, generalized=True)
    idx = np.argsort(periodogram)
    return periods[idx[-1]]
```

### Result

```python
view.scatter('ids', lcs.ids)
view.execute('pr = [period(id_) for id_ in ids]')
pr = view.gather('pr')
```

### Plot

```python
g_r = [lcs.get_target_parameter(id_, 'gr') for id_ in lcs.ids]
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

plt.scatter(g_r, np.log(pr), s=10, alpha=0.2, edgecolors='none')
plt.xlabel('g-r')
plt.ylabel('log(P)')
plt.xlim(-1.5, 1.5)
plt.ylim(-3, 0)
plt.show()
```

### Classification: Labeling Photometric Sources

Load the combined RR-Lyrae colors and SDSS standards colors.

```python
from astroML.datasets import fetch_rrlyrae_combined
rrlyrae = fetch_rrlyrae_combined()
```

Plot the data to see how they look like.

```python
def plot_rrlyrae(data):
    X, y = data
    N_plot = 5000
    plt.scatter(X[-N_plot:, 0], X[-N_plot:, 1], s=10, c=y[-N_plot:], alpha=0.2, edgecolors='none')
    plt.xlabel('u-g color')
    plt.ylabel('g-r color')
plot_rrlyrae(rrlyrae)
```

Split the data into train/test set.

```python
from sklearn.cross_validation import train_test_split
rrlyrae_splitting = train_test_split(*rrlyrae, random_state=RS)
```

Define a classification function for benchmark. We're looking for higher precision and recall. It is determined by higher F1 score.

```python
from sklearn.metrics import classification_report, precision_score, recall_score, f1_score

def classify(clf, interval=1, splitting=rrlyrae_splitting):
    X_train, X_test, y_train, y_test = splitting
    s = slice(None, None, interval)
    clf.fit(X_train[s], y_train[s])
    y_pred = clf.predict(X_test)
    vs = (y_test, y_pred)
    print(classification_report(*vs, target_names=['MS star', 'RR Lyrae']))
    return (
        precision_score(*vs),
        recall_score(*vs),
        f1_score(*vs),
    )
```

#### K-Nearest Neighbors

```python
from sklearn.neighbors import KNeighborsClassifier
classify(KNeighborsClassifier(n_neighbors=5))
```

#### Gaussian Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
classify(GaussianNB())
```

#### Linear Discriminant Analysis

```python
from sklearn.lda import LDA
classify(LDA())
```

#### C-Support Vector Classification

Support vector machine didn't work out well in this case. Virtually none was classified to `RR Lyrae'.

```python
from sklearn.svm import SVC
classify(SVC(), interval=5)
```

```python
# explicitly set RBF kernel, even though it is default
classify(SVC(kernel='rbf'), interval=5)
```

#### Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
classify(RandomForestClassifier(random_state=RS))
```

```python
classify(RandomForestClassifier(n_estimators=100, random_state=RS))
```

#### What's the best precision/recall you can obtain for the RR-Lyrae data?

A random forest classifier showed the best result with **80% precision** and **56% recall** (F1=0.6570). K-nearest neighbors also had a decent score (F1=0.6278) with **69% precision** and **57% recall**.

### Regression: Photometric Redshifts

Load SDSS Galaxies with spectral information.

```python
from astroML.datasets import fetch_sdss_specgals
sdss = fetch_sdss_specgals()
```

Split into test/train set.

```python
def split_sdss(data):
    # put magnitudes in a matrix
    X = np.vstack([data['modelMag_%s' % f] for f in 'ugriz']).T
    y = data['z']
    return train_test_split(X, y, random_state=RS)
sdss_splitting = split_sdss(sdss)
```

Define a regression function for benchmark. We're looking for smaller RMSE.

```python
def regress(est, interval=1, splitting=sdss_splitting):
    X_train, X_test, y_train, y_test = splitting
    s = slice(None, None, interval)
    est.fit(X_train[s], y_train[s])
    y_pred = est.predict(X_test)
    
    plt.scatter(y_test, y_pred, s=1, c='k', alpha=0.3, marker=',')
    plt.plot([0, 1], [0, 1], ':k')
    plt.xlim(0, 0.6)
    plt.ylim(0, 0.6)
    
    rms = np.sqrt(np.mean((y_test - y_pred) ** 2))
    return rms
```

#### Ordinary Least Squares Linear Regression

```python
from sklearn.linear_model import LinearRegression
regress(LinearRegression())
```

#### K-Nearest Neighbors

```python
from sklearn.neighbors import KNeighborsRegressor
regress(KNeighborsRegressor())
```

#### $\epsilon$-Support Vector Regression

```python
from sklearn.svm import SVR
regress(SVR(), interval=5)
```

#### Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
regress(RandomForestRegressor(n_estimators=100, random_state=RS), interval=10)
```

#### Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingRegressor
regress(GradientBoostingRegressor(random_state=RS), interval=50)
```

#### What's the best RMS error you can get on the data?

A random forest regressor showed the best result with the lowest RMSE of **0.02317**. K-nearest neighbors was approaching very close with **0.02329** though.

### Implementation

#### 1) Create drumhead_height()

```python
from scipy import *
from scipy.special import jn, jn_zeros

def drumhead_height(n, k, distance, angle, t):
    # only use the last k-th element
    kth_zero = jn_zeros(n, k)[k-1]
    return cos(t) * cos(n*angle) * jn(n, distance*kth_zero)
```

#### 2,3,4) Create plot_Bessel()

```python
from mpl_toolkits.mplot3d import axes3d

def plot_Bessel(t, r0=1.0):
    # (radius, theta): polar space
    radius = linspace(0, r0)
    theta = linspace(0, 2*pi)
    rm, thetam = meshgrid(radius, theta)
    
    # set up 3x3 grid for subplots
    fig = plt.figure(figsize=(16, 10))
    gs = plt.GridSpec(3, 3, wspace=0, hspace=0.1)
    
    # for subplots with different layout and parameters (n, k)
    def plot_drumhead(sp, n, k):
        # create 3D axis of the subplot
        ax = fig.add_subplot(sp, projection='3d')
        
        # (x, y): Cartesian space
        x = rm*cos(thetam)
        y = rm*sin(thetam)
        z = drumhead_height(n, k, rm, thetam, t)
        
        # plot 3D surface with projected contours
        style = {
            'vmin': -0.5,
            'vmax': 0.5,
            'cmap': cm.jet,
            'alpha': 0.4,
        }
        surf = ax.plot_surface(x, y, z, rstride=1, cstride=1, **style)
        ax.contourf(x, y, z, zdir='z', offset=-0.5, **style)
        ax.contour(x, y, z, zdir='y', offset=r0, **style)
        ax.contour(x, y, z, zdir='x', offset=-r0, **style)
        fig.colorbar(surf)
        
        # set up title and labels
        ax.set_title('$u_{{{},{}}}$'.format(n, k))
        ax.set_xlabel('X')
        ax.set_xlim(-r0, r0)
        ax.set_ylabel('Y')
        ax.set_ylim(-r0, r0)
        ax.set_zlabel('Z')
        ax.set_zlim(-0.5, 0.5)

    # lay out multiple subplots
    plot_drumhead(gs[:3,:2], 2, 3)
    plot_drumhead(gs[0,2], 1, 1)
    plot_drumhead(gs[1,2], 3, 3)
    plot_drumhead(gs[2,2], 5, 5)
    
    #HACK: ensure all subplots fit tight in the bounding box (for better image export)
    gs.tight_layout(fig)
```

```python
plot_Bessel(0)
```

#### 5) Generate animation

Save 5 frames for 2 seconds.

```python
f = 5
s = 2.
for i in range(f):
    t = i*(s/f)
    plot_Bessel(t)
    plt.savefig('frame{0:07d}.png'.format(i+1))
    plt.close()
```

Generate animation with [JSAnimation](https://github.com/jakevdp/JSAnimation).

```python
from JSAnimation import IPython_display, HTMLWriter
from matplotlib import animation
from PIL import Image
import glob

# populate filenames for animated frames
filenames = sorted(glob.glob('frame*.png'))

# set up figure and axis
fig, ax = plt.subplots(figsize=(16, 10))
fig.tight_layout()
ax.axis('off')
im = plt.imshow(Image.open(filenames[0]))

# define animation step function
def animate(i):
    image = Image.open(filenames[i])
    im.set_data(image)
    return im,

# set up animation
anim = animation.FuncAnimation(fig, animate, frames=f, interval=s/f*1000, blit=True)
# embed base64-encoded frames directly in the HTML
anim.save('animation.html', writer=HTMLWriter(embed_frames=True))

# show animation in the notebook
anim
```

### RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
```

15 trees, max depth of the trees are 100, with the the square root of features.
Tune the parameters of the classifier:

```python
rfc1 = RandomForestClassifier(n_estimators=10)
rfc1.fit(X_train, y_train)
y_pred = rfc1.predict(X_test)
print(np.sum(y_pred == y_test))
print(np.sum(y_pred != y_test))
```

Calculate the rate of correct prediction (correct predicted divided by totoal number of the prediction samples)

This function return the index of the max value in a list

```python
def max_index(n):
    val = n[0]
    i = 0
    for j in range(1, len(n)):
        if n[j] > val:
            val = n[j]
            i = j
    return i
```

```python
N=50
Correct=[]
for i in range (5,N):
    rfc_temp = RandomForestClassifier(n_estimators=i)
    rfc_temp.fit(X_train, y_train)
    y_pred = rfc_temp.predict(X_test)
    correctTemp=np.sum(y_pred == y_test)
    Correct.append(correctTemp)
print(Correct)
```

```python
plt.plot(Correct)
ax = plt.axes()
plt.xlabel("n_estimators")
plt.ylabel("Number of Correct estimation")
plt.show()
```

```python
n1=max_index(Correct)
```

When n_estimate is n1, the rate of correct prediction is the max. Then, I set the value of this parameter to be n1

```python
rfc = RandomForestClassifier(n_estimators=n1)
rfc.fit(X_train, y_train)
y_pred = rfc.predict(X_test)
```

```python
print(np.sum(y_pred == y_test))
print(np.sum(y_pred != y_test))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred,
                            target_names=['MS star', 'RR Lyrae']))
```

```python
from astroML.datasets import fetch_sdss_specgals
from astroML.datasets import fetch_rrlyrae_combined
from sklearn.cross_validation import train_test_split
```

```python
data = fetch_sdss_specgals()

# put magnitudes in a matrix
X1 = np.vstack([data['modelMag_%s' % f] for f in 'ugriz']).T
y1 = data['z']

# Split into training and testing data
X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1)

print (X1_train.shape)
print (y1_test.shape)
```

### RandomForestRegressor

```python
print (X1_train.shape)
print (y1_train.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor
est = RandomForestRegressor(n_estimators=1)
est.fit(X1_train,y1_train)
y1_pred = est.predict(X1_test)
print y1_pred.shape
```

```python
plt.plot(y1_test, y1_pred, ',k')
plt.plot([0, 1], [0, 1], ':k')
plt.xlim(0, 0.6)
plt.ylim(0, 0.6)
```

Tune the parameter by finding the min rms (it takes very long to run)

```python
rms = np.sqrt(np.mean((y1_test - y1_pred) ** 2))
print(rms)
```

```python
###n_estimators =5
est1 = RandomForestRegressor(n_estimators=5)
est1.fit(X1_train,y1_train)
y1_pred_1 = est1.predict(X1_test)

rms_1 = np.sqrt(np.mean((y1_test - y1_pred_1) ** 2))
print(rms_1)
```

```python
##n_estimators =10
est2 = RandomForestRegressor(n_estimators=10)
est2.fit(X1_train,y1_train)
y1_pred_2 = est2.predict(X1_test)

rms_2 = np.sqrt(np.mean((y1_test - y1_pred_2) ** 2))
print(rms_2)
```

```python
##n_estimators =15
est3 = RandomForestRegressor(n_estimators=15)
est3.fit(X1_train,y1_train)
y1_pred_3 = est3.predict(X1_test)

rms_3 = np.sqrt(np.mean((y1_test - y1_pred_3) ** 2))
print(rms_3)
```

```python
##n_estimators =20
est4 = RandomForestRegressor(n_estimators=20)
est4.fit(X1_train,y1_train)
y1_pred_4 = est4.predict(X1_test)

rms_4 = np.sqrt(np.mean((y1_test - y1_pred_4) ** 2))
print(rms_4)
```

So from the above calculation, when n_estimators is 20, the rms reaches the minimum compared with 1, 5,10 and 15

```python
M=20
closePred=[]
for i in range (1,M,4):
    rfr_temp = RandomForestRegressor(n_estimators=i)
    rfr_temp.fit(X1_train, y1_train)
    y1_pred_temp= rfr_temp.predict(X1_test)
    closeTemp= np.sqrt(np.mean((y1_test - y1_pred_temp ) ** 2))
    closePred.append(closeTemp)
print(closePred)
```

Find the min of the rms

```python
def min_index(n):
    val = n[0]
    i = 0
    for j in range(1, len(n)):
        if n[j]< val:
            val = n[j]
            i = j
    return i
```

Use the index of the min of the rms list

```python
m1=min_index(closePred)
```

```python
est1 = RandomForestRegressor(n_estimators=m1)
est1.fit(X1_train,y1_train)
y1_pred_1 = est1.predict(X1_test)
print y1_pred_1.shape
print np.sqrt(np.mean((y1_test - y1_pred_1) ** 2))
```

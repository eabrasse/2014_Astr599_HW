```python
##define func of sin(x)
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
def func(x):
    return np.sin(x)
```

```python
##The Integrator class according to the template
class Integrator:
    def __init__(self, a, b, n):
        self.a, self.b, self.n = a, b, n
        self.points, self.weights = self.method()
        
    def method(self):
        raise NotImplementedError('no rule in class %s' %self.__class__.__name__)
    
    def integrate(self,func):
        return np.sum(func(self.points)*self.weights)
    
        
```

```python
##Trapezodal method child class
class Trapezoidal(Integrator):
    def _init_(self,a,b,n):
        super(Integrator,self)._init_(a,b,n)
    def method(self):
        h=(self.b-self.a)*1.0/self.n
        x=np.linspace(self.a,(self.b-h),self.n)
        w=np.ones(self.n)*h
        return x,w

### Trial of instance
Trial=Trapezoidal(1,2,10)
Trial.integrate(func)
```

```python
###MidPoint method
class Midpoint(Integrator):
    def _init_(self,a,b,n):
        super(Integrator,self)._init_(a,b,n)
    def method(self):
        h=(self.b-self.a)*1.0/self.n
        w=np.ones(self.n)*h
        x=np.linspace((self.a+h/2.0),(self.b-h/2.0),self.n)
        return x,w
Trial2=Midpoint(1,2,10)
Trial2.integrate(func)
```

```python
##1-cos(x), a test function
def sinIntegrate(x):
    return 1-np.cos(x)
##Trial
sinIntegrate(np.pi)
```

```python
##Compare with the real value by subtracting the real value from both methods
npts=[10,20,50,100,200,300] ##
exact=sinIntegrate(npts)
##setting the integration limit as 2
upper=2

###Trapezoidal Method
Trap=[] ##the integration calculated by Trape
for n in npts:
    Trap.append(Trapezoidal(0,upper,n).integrate(func))
TrapError=Trap-exact ##The error of Trape
    
```

```python
plt.plot(npts, np.abs(TrapError))
plt.title(r'Trapezoid approximation to 1-cos(x)')
plt.xlabel('Number of Bins')
plt.ylabel('Error by Trapezoidal Method')
plt.grid(True)
plt.ylim((0,2));
```

```python
#MidPoint Method
upper=2
Mid=[] ##the integration calculated by Trape
for n in npts:
    Mid.append(Midpoint(0,upper,n).integrate(func))
#print(Trap)
MidError=Mid-exact ##The error of Midpoint
```

```python
plt.plot(npts, np.abs(MidError),c="red")
plt.title(r'Midpoint approximation to 1-cos(x)')
plt.xlabel('Number of Bins')
plt.ylabel('Error by Midpoint Method');
plt.grid(True);
plt.ylim((0,2))
```

```python
plt.plot(npts,abs(TrapError))
plt.plot(npts,abs(MidError))
plt.xlabel('Number of Bins')
plt.ylabel('Error by both Methods');
plt.legend(["Trapezoidal","Midpoint"])
plt.show
```

```python
From above comparison, for the integration of sin(x), there are larger errors in Trapezoidal method when number of bins is small. 
On the other side, when number of bins grows larger, there are larger errors in Midpoint method.
When number of bins grows, the errors will decreas to 0.
```

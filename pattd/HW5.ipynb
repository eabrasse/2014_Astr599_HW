```python
import numpy as np
import matplotlib.pyplot as plt
from pylab import *
```

```python
# Import 3D libraries
from mpl_toolkits.mplot3d import Axes3D
```

```python
from scipy import *
from scipy.special import jn, jn_zeros
```

## Drum head function

```python
def drumhead_height(n, k, r, theta, t):
    u = np.cos(t)*np.cos(n*theta)
    la = jn_zeros(n,k)[k-1]
    return (u * jn(n,la*r))
```

## Bessel plot

```python
def plot_Bessel(t):
    gs = plt.GridSpec(3, 3, wspace=0.2, hspace=0.2)
    fig = plt.figure(figsize=(12, 6))
    
    r = np.linspace(0,1,100)
    theta = np.linspace(0,2*np.pi,100)
    rm, thetam = meshgrid(r, theta)
    xm, ym = meshgrid(r, theta)
    xm = rm*np.cos(thetam)
    ym = rm*np.sin(thetam)
    
    ## Fig 1
    ax1 = fig.add_subplot(gs[:3, :2],projection='3d')
    z = drumhead_height(2,3,rm,thetam,t)
    specs = {'cmap' : 'seismic', 'vmin' : -0.2, 'vmax' : 0.7}
    surf = ax1.plot_surface(xm, ym, z, rstride = 1, cstride = 2, linewidth=0, **specs)
    ax1.set_axis_off() # Remove background axis and grid
    
    ## Fig 2
    ax2 = fig.add_subplot(gs[0, 2],projection='3d')
    z = drumhead_height(0,1,rm,thetam,t)
    specs = {'cmap' : 'PiYG', 'vmin' : -0.2, 'vmax' : 0.7}
    surf = ax2.plot_surface(xm, ym, z, rstride = 1, cstride = 2, linewidth=0, **specs)
    ax2.set_axis_off() # Remove background axis and grid
    
    ## Fig 3
    ax3 = fig.add_subplot(gs[1, 2],projection='3d')
    z = drumhead_height(1,1,rm,thetam,t)
    specs = {'cmap' : 'Spectral', 'vmin' : -0.2, 'vmax' : 0.7}
    surf = ax3.plot_surface(xm, ym, z, rstride = 1, cstride = 2, linewidth=0, **specs)
    ax3.set_axis_off() # Remove background axis and grid
    
    ## Fig 4
    ax4 = fig.add_subplot(gs[2, 2],projection='3d')
    z = drumhead_height(3,1,rm,thetam,t)
    specs = {'cmap' : 'RdYlBu', 'vmin' : -0.2, 'vmax' : 0.7}
    surf = ax4.plot_surface(xm, ym, z, rstride = 1, cstride = 2, linewidth=0, **specs)
    ax4.set_axis_off() # Remove background axis and grid
    return
    
```

```python
# Test Bessel plot at t = 0
plot_Bessel(0)
plt.show()
```

## Produce animation

```python
n = 5 # Produce 5 frame
print('Frame number:')
for i in xrange(0,n,1):
    # 3D plot parameters as function of iterator i
    t = i*4
    # Call plotting function and save figure into file
    plot_Bessel(t)
    plt.savefig('animation_frames/frame{0:01d}.png'.format(i+1));
    plt.close()
    print i,
```

```python
# Import JSAnimation 
from JSAnimation import IPython_display, HTMLWriter
from matplotlib import animation
import Image
import glob

# Choose location of files to use
filenames=sorted(glob.glob('animation_frames/frame*.png'))

#Create figure
fig = plt.figure(figsize=(9,5))
ax = fig.add_axes([0, 0, 1, 1])
ax.axis('off')
im = plt.imshow(Image.open(filenames[0]))

# Create initializatio function
def init():
    im.set_data(Image.open(filenames[0]))
    return im,

# Create animation function
def animate(i):
    image=Image.open(filenames[i])
    im.set_data(image)
    return im,

anim = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=5, interval=100, blit=True)

#set embed_frames=True to embed base64-encoded frames directly in the HTML
anim.save('animation_frames/animation.html', writer=HTMLWriter(embed_frames=True))

# Show animation in ipython notebook
anim
```

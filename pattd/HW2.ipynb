```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

$f(x) = sin(x)$

```python
def func(x):
    return np.sin(x)
```

Parent class

```python
# parent class
class Integrator:
    def __init__(self, a, b, n):
        self.a, self.b, self.n = a, b, n
        self.points, self.weights = self.method()
        
    def method(self):
        raise NotImplementedError('no rule in class %s' %self.__class__.__name__)
        
    # add function integrate
    def integrate(self,func):
        wf = func(self.points)*self.weights
        return np.sum(wf)
```

### Trapezoidal

```python
# create child class Trapezoidal
class Trapezoidal(Integrator):
        
    # Defind trapezoidal rule with n uniform grid
    # n grid points means n-1 spaced panels
    def method(self):
        x = np.linspace(self.a, self.b,self.n)
        weights = np.ones(self.n, dtype=float)*2*(self.b-self.a)/(2*(self.n-1))
        weights[0] = weights[0]/2
        weights[-1] = weights[-1]/2
        return x, weights
```

### Midpoint

```python
# create child class Midpoint
class Midpoint(Integrator):
        
    # Defind midpoint method with n uniform grid
    # n grid points means n-1 spaced panels
    def method(self):
        weights = np.ones(self.n-1, dtype=float)*(self.b-self.a)/(self.n-1)
        x = np.linspace(self.a, self.b,self.n-1) + (self.b-self.a)/(2*(self.n-1))
        return x , weights
```

### Plot

```python
from IPython.html.widgets import interact

def plotintesin(n):
    x = np.linspace(0,4*np.pi,1000)
    ytrap = np.zeros(1000)
    ymid = np.zeros(1000)
    
    for i in np.arange(0,999):
        ytrap[i] = Trapezoidal(0,x[i],n).integrate(func)
        ymid[i] = Midpoint(0,x[i],n).integrate(func)
    
    plt.plot(x,ytrap);
    plt.plot(x,ymid,color='green');
    plt.ylabel("Integral of sin(x) from 0 to x");
    plt.xlabel("x");
    plt.title("Approximated Integration");
    pass

interact(plotintesin, n = (1,100));
```

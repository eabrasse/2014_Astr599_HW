### Part 1: loading and exploring the data

1. Use ``np.genfromtxt`` to load the data from the file.  This is like ``loadtxt``, but can handle missing data.

    - Remember to set the appropriate ``delimiter`` keyword.

2. ``genfromtxt`` sets all missing values to ``np.nan``.  Use the operations we discussed here to answer these questions:

     - How many values are missing in this data?
     - How many complete *rows* are there?  i.e. how many objects have **no** missing values?

3. Create a new array containing only the rows with no missing values.

4. Compute the maximum, minimum, mean, and standard deviation of the values in each column.

```python
#1 Load the data:
asteroids=np.genfromtxt('asteroids5000.csv',delimiter=',')

# 2a Determining number of missing elements /values
x=asteroids[np.isnan(asteroids)]
print("(2a) Total Missing Elements: "+str(x.size))

# 2b Determining number of complete rows
good_asteroids=np.ndarray((0,7))
for row in asteroids:
    if ~(np.isnan(row).any()):
        good_asteroids=np.vstack((good_asteroids,row))
        
#print(good_asteroids.shape)
print("(2b) Number of complete rows: "+str(good_asteroids.shape[0]))

# 3 new array already created as 'good_asteroids'

# 4 Computing Max, min, mean and std deviation
Max=np.amax(good_asteroids, axis=0)
Min=np.amin(good_asteroids, axis=0)
Mean=np.mean(good_asteroids, axis=0)
Std=np.std(good_asteroids, axis=0)
print("(4)\n"+"Max: "+str(Max)+"\nMin: "+str(Max)+"\nMean: "+str(Mean)+"\nStd: "+str(Std))
```

```python
# Kwang Seob Kim
# Amath 500
# HW1

import numpy as np
import urllib2
import csv
import os

#loading data
#got this idea from a colleague and emailing prof vanderplas
url = 'https://raw.githubusercontent.com/jakevdp/2014_fall_ASTR599/master/notebooks/data/asteroids5000.csv'
response = urllib2.urlopen(url)
data = np.genfromtxt(response,delimiter=',')
```

```python
# How many values are missing in this data?

#check which one is nan
print 'There are',np.isnan(data).sum(),'missing values in this data.'
```

```python
#How many complete rows are there? i.e. how many objects have no missing values?
print 'There are', len(data[~np.isnan(data).any(axis=1)]), 'complete rows (no missing values).'
```

```python
#Create a new array containing only the rows with no missing values.
newArray = data[~np.isnan(data).any(axis=1)]
newArray

```

```python
#Compute the maximum, minimum, mean, and standard deviation of the values in each column.
print 'maximum:', np.max(newArray,0)
print 'minimum:', np.min(newArray,0)
print 'mean:', np.mean(newArray, 0)
print 'standard dev:', np.std(newArray,0)
```

```python
%matplotlib inline 
import matplotlib.pyplot as plt

semimajor = newArray[:,0]
inclination = np.sin(np.radians(newArray[:,2]))
plt.scatter(semimajor, inclination,s=8)
plt.xlabel('a')
plt.ylabel('sin(i)')
#What do you notice about the distribution? 
#Two-three data groups are heavily clustered together.
#What do you think could explain this?
#Many asteroids with semi major axis between 2 and 4 have inclination angle of .3 or below.  
```

```python
#Use plt.scatter to plot a color-magnitude diagram of the asteroids (H vs B-V). 
#You should see two distinct "families" of asteroids indicated in this plot. Over-plot a line that divides these.
B= newArray[:,3]
BH= newArray[:,5]
plt.scatter(B,BH, s=5)
plt.xlabel('B')
plt.ylabel('B-H')
plt.axhline(0.78)
```

```python
#Do you see any correlation between color and orbit?
plt.scatter(B,BH, c=np.where(BH > 0.78, 'r', 'y'),s=5)
plt.xlabel('B')
plt.ylabel('B-H')
#Red color (ones with higher magnitude) have more wide spread range of orbit including lower orbit ones. Yellow color ones are more 
#clustered around 8-10 orbit range.
```

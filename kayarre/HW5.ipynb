## 4. ASSIGNMENT: PLOTTING AND ANIMATING THE DRUMHEAD VIBRATIONAL MODES

Whenever you hit a drum, the drumhead vibrates following the two dimensional wave equation in polar coordinates,

$$
\frac{\partial u(r,\theta,t)}{\partial t} = c^2 \nabla^2 u(r,\theta,t) \\[5mm]
u(R_0,\theta,t) = 0
$$

with $R_0 = 1$ the radius of the drumhead and $u(r,\theta,t)$ the elevation of the drumhead. The solution is given in terms of a superposition of its vibrational modes:

$$
u_{n,k}(r,\theta,t) = \cos(t) \cos(n\theta) \mathrm{J}_n(\lambda_{n,k} r)
$$

where $\lambda_{n,k}$ is the $k^{th}$ positive root of $\mathrm{J}_n$ and we assumed $c=1$. Don't worry, if you don't understand the equation, in this assignment we will only plot the solutions step by step.

```python
# Import 3D libraries
import numpy as np
import matplotlib.pyplot as plt 
from mpl_toolkits.mplot3d import Axes3D
import scipy.special as special #jn, jn_zeros
from matplotlib import cm

%pylab inline

```

```python
def k_root(k,n):
    return special.jn_zeros(n,k)[k-1] # m is 0-indexed here

# It's always good to create a function for your plots
def drumhead_height(n, k, r, theta, time):
    return np.cos(time) * np.cos(n * theta) * special.jn(n, k_root(k,n) * r) 
```

```python
from mpl_toolkits.mplot3d import Axes3D

def plot_besssel(t):
    r0 = 1.0
    r = np.linspace(0, 1, 30)
    theta = np.linspace(0, 2*np.pi, 30)
    R, THETA = np.meshgrid(r, theta)
    
    # catesian space
    x_grid = R * np.cos(THETA)
    y_grid = R * np.sin(THETA)
    
    # Create figure
    fig = plt.figure(figsize=(17, 10))
    gs = plt.GridSpec(3, 3, wspace=0, hspace=0.1)
    
    def plot_sub(sub_p, n, k, fs=10):
        axes = fig.add_subplot(sub_p, projection='3d')
        # plot head with different parameters (n, k)
        z_grid = drumhead_height(n, k, R, THETA, t)
        
        # plot 3D surface with projected contours
        specs = {
            'vmin': -0.5,
            'vmax': 0.5,
            'cmap': cm.spectral,
            'alpha': 0.3,
        }
        surf = axes.plot_surface(x_grid, y_grid, z_grid, rstride=1, cstride=1, **specs)
        axes.contourf(x_grid, y_grid, z_grid, zdir='z', offset=-0.5, **specs)
        axes.contour(x_grid, y_grid, z_grid, zdir='y', offset=r0, **specs)
        axes.contour(x_grid, y_grid, z_grid, zdir='x', offset=-r0, **specs)
        fig.colorbar(surf)
        
        # set up title and labels
        axes.set_title(r'$u_{{{0},{1}}}$'.format(n, k), fontsize=20)
        axes.set_xlabel(r'$X$', fontsize=fs)
        axes.set_xlim(-r0, r0)
        axes.set_ylabel(r'$Y$', fontsize=fs)
        axes.set_ylim(-r0, r0)
        axes.set_zlabel(r'$Z$', fontsize=fs)
        axes.set_zlim(-0.5, 0.5)
    # lay out multiple subplots
    plot_sub(gs[:3,:2], 2, 3, 20)
    plot_sub(gs[0,2], 1, 1, 10)
    plot_sub(gs[1,2], 3, 3, 10)
    plot_sub(gs[2,2], 5, 1, 10)

    #ensure all subplots fit bounding box
    gs.tight_layout(fig)
```

```python
plot_besssel(0)
```

```python
# Output a set of images using plot3dsine()
n = 5 # Use 100 to produce bigger animation
frame_step_per_sec = 0.5
print('Frame number:')
for i in xrange(0,n,1):
    # 3D plot parameters as function of iterator i
    t = i * frame_step_per_sec
    # Call plotting function and save figure into file
    plot_besssel(t)
    plt.savefig('animation_frames/Hw_frame{0:07d}.png'.format(i+1));
    plt.close()
    print i,
```

```python

# Import JSAnimation 
from JSAnimation import IPython_display, HTMLWriter
from matplotlib import animation
import Image
import glob

# Choose location of files to use
filenames=sorted(glob.glob('animation_frames/Hw_frame*.png'))

#Create figure
fig = plt.figure(figsize=(10,5))
ax = fig.add_axes([0, 0, 1, 1])
ax.axis('off')
im = plt.imshow(Image.open(filenames[0]))

# Create initializatio function
def init():
    im.set_data(Image.open(filenames[0]))
    return im,

# Create animation function
def animate(i):
    image=Image.open(filenames[i])
    im.set_data(image)
    return im,

anim = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=n, interval=400, blit=True)

#set embed_frames=True to embed base64-encoded frames directly in the HTML
anim.save('animation_frames/HW_animation.html', writer=HTMLWriter(embed_frames=True))

# Show animation in ipython notebook
anim
```

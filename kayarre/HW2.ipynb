```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# part 1
def func(x):
    return np.sin(x)
```

Step 2 and 3:

Integrator class template
Assume uniform grid spacing

```python
# integrator class
class Integrator(object):
    def __init__(self, a, b, n):
        #@param: a initial value
        #@param: b end value
        #@param: n discrete points
        self.a, self.b, self.n = a, b, n
        self.points, self.weights = self.method()
    
    #sum of weight*f(x) from [0,n-1]
    def integrate(self,func):
        f = func(self.points)
        result = self.weights[:-2] * f[:-2] #-2 is n-1 location
        return np.sum(result)
    
    def method(self):
        raise NotImplementedError('no rule in class %s' %self.__class__.__name__)
```

Step 4:
Create trapezoidal child class and method(self) implimentation.

Method(self) creates an array of points x and and an array of weights w based on the trapezoidal rule formula:
$$
\int_{a}^{b} f(x)\, dx \approx \frac{h}{2} \sum_{k=1}^{N} \left( f(x_{k+1}) + f(x_{k}) \right){}= \frac{b-a}{2N}(f(x_1) + 2f(x_2) + 2f(x_3) + \dotsb + 2f(x_N) + f(x_{N+1}))
$$
https://en.wikipedia.org/wiki/Trapezoidal_rule.

```python
#Part 4: Define a child class that has Integrator as parent class
class Trapezoidal(Integrator):
    
    # Initialize child class function 
    def __init__(self, a, b, n):
        super(Trapezoidal, self).__init__(a, b, n)
        
    #Method returns x and weight arrays for Trapezoidal Method    
    def method(self):
        x = np.linspace(self.a,self.b,self.n)
        weights = np.ones(self.n)
        weights[1:-2].fill(2)
        weights = weights*((self.b-self.a)/(2.0*self.n))
        return x, weights
```

```python
trap = Trapezoidal(0.0,np.pi,2000)
```

```python
print(trap.integrate(func))
```

Step 5:

Create a Midpoint class that inherits from Integrator and behaves like Trapezoidal.
$$
\int_{a}^{b} f(x)\, dx \approx \sum_{k=1}^nf(\bar x_k)\Delta x_k
= \frac{b-a}{n}\sum_{k=1}^nf(\bar x_k) = \frac{b-a}{n}(\hat y_1 + \hat y_2 + ... +\hat y_n)
$$
Where $\hat y_i = f(\bar x_k)$, $\Delta x = \frac{b-a}{n}$, and $\bar x_i = \frac{x_{i-1}+x_i}{2}$ 

```python
# Define a child class that has Integrator as parent class
class Midpoint(Integrator):
    
    # Initialize child class function 
    def __init__(self, a, b, n):
        super(Midpoint, self).__init__(a, b, n)
        
    #Method returns proper x, weight arrays for Midpoint Method   
    def method(self):
        #set x to be of (x(i) + x(i+1))/2
        x = np.linspace(self.a,self.b,self.n)
        x_array = [ (x[i-1] + x[i])/2.0 for val,i in enumerate(x) if i > 0]
        np.append(x_array,self.b)
        weights = np.ones(self.n)
        weights = weights*((self.b-self.a)/(self.n))
        return x, weights
```

Step 6:

Function that plots  the integral of f(x) = sin(x) using Trapezoidal Rule for various values of x

```python
#Function takes in precision of integration in steps, n
def plotfunc(n,func, x0, xf):
    input_x = np.linspace(x0,xf,n)
    y = []
    #loop that does the work
    y = [ Trapezoidal(0,x_i,n).integrate(func) for x_i in input_x]
    plt.plot(input_x,y, label="discrete integral")
    plt.xlabel(r"Integral $b$ limit.")
    plt.ylabel(r"Value of integral of $sin(x)$ from 0 to x")
    plt.title(r"Comparison of numerical integration with analytic solution")
    return input_x,y
```

Plot integration function, 1-cos(x) to validate results.
